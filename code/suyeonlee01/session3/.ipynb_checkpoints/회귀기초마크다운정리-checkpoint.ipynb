 # 머신러닝
 머신러닝에는 지도학습, 비지도학습, 강화학습이 있음. 
 ## 머신러닝 개념
 새로운 데이터를 예측하거나 결정을 내릴 수 있도록 하는 기술. 이를 통해 데이터에서 패턴을 찾아내고, 패턴을 바탕으로 새로운 데이터에 대한 예측을 가능하게 함. 
 ### 지도학습
 정답이 있는 데이터로 학습하는 방식. (회귀, 분류)
 ### 비지도학습
 정답이 없는 데이터로 학습하는 방식. (군집화, 클러스터링)

 # **회귀**
 회귀란 데이터에서 패턴을 찾아내어 미래 값을 예측하는 것을 의미함. 통계나 머신러닝의 주요한 개념이며, 연속적인 숫자를 다루는 특징이 있음. 회귀의 종류에는 선형회귀, 비선형회귀, 로지스틱회귀, 릿지회귀 등 다양한 회귀 기법이 존재함. 

 # **선형회귀**
 데이터를 가장 잘 설명하는 회귀선을 찾아내어 예측하는 과정을 말함. 예측하고자 하는 것을 종속 변수(y), 예측을 위해 사용하는 변수를 독립 변수(x)라고 함. 기본가정에는 선형성, 독립성, 등분산성, 정규성이 있음.

 ## 단순 선형 회귀
 데이터를 직선 형태 (**하나의 독립변수**)로 표현하는 경우. 데이터를 가장 잘 설명하는 직선 y=aX+b를 찾는 과정에 해당함. 기울기 a는 회귀계수를, b는 y절편을 의미함. 적절한 회귀계수와 절편을 찾는 과정임. 최소제곱법과 최소자승법을 이용하여 최적의 회귀선을 찾음. 이는 회귀선과 실제 데이터 간의 차이를 최소화하기 위해 잔차 제곱의 합이 최소가 되는 지점으로 최적의 회귀선을 구하는 방식임. 

 ## 다중 선형 회귀
 데이터를 **두 개 이상**의 독립변수로 표현하는 경우. 이 경우에는 3차원 공간에 표현이 됨. 회귀선은 평면임. 단순 선형 회귀와 마찬가지로 최소제곱법을 사용하여 구함. 

 # 상관계수
 상관계수란 상관관계의 정도를 수치로 나타낸 것임. -1과 1사이에 존재함. -1 또는 1에 가까울 수록 두 변수 간의 상관성이 높음을 의미함. 상관계수가 음수라면 음의 상관관계가, 양수라면 양의 상관관계가 존재함. 0인 경우 선형 상관관계가 없음. 

 # VIF지수
 분산 팽창 인수, 회귀 모델의 결정계수 R square를 사용하여 계산 됨. 1인 경우 상관관계가 전혀 없음을 의미. 5 미만인 경우 다중공선성 문제가 없다고 간주. 5 초과인 경우 그 징후가 있을 수 있으므로 주의를 요함. 10을 초과하는 경우 다중공선성이 심각하다고 간주 됨. 

 # 다중공선성 대처 방법
 1. 변수 제거
 2. 변수 변환
 3. 규제 선형 모델 활용
 4. PCA(주성분분석) : 데이터의 차원을 축소하는 데 사용되는 통계적 기법. 고차원 데이터에서 중요한 정보를 최대한 보존하면서 데이터의 복잡성을 줄이는 것을 목표로 함. (데이터의 차원이 높으면 분석과 시각화가 어려움.)

# 성능평가지표
## 평균제곱오차(MSE)
=Mean Squared Error. 회귀 분석에서 모델의 예측값과 실제 관측값 사이의 오차 제곱의 평균을 의미. 모델의 예측 성능을 평가하는 지표 중 하나임. 
## 평균절대오차(MAE)
=Mean Absolute Error. 절대값 오차의 평균을 의미. 

# 변수 유성 평가
## t검정
독립변수의 회귀계수가 0인지 아닌지로 가설 설정. 회귀분석에서의 t-검정은 각 독립 변수의 회귀계수가 유의한지를 검정하는데 사용됨. 유의성이 크다면 종속 변수에 영향을 주는 변수라고 판단함. t통계량에 따른 p-value가 0.05보다 작으면 해당 변수는 유의하다고 판단함.
