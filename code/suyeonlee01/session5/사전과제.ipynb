# 1. 지도 학습과 비지도 학습의 차이에 대해 설명해주세요.

머신러닝: 인공지능의 한 분야. 컴퓨터가 스스로 학습할 수 있도록 도와주는 알고리즘이나 기술을 개발하는 분야. 머신러닝에서 컴퓨터는 알고리즘을 이용하여 데이터를 분석하고, 분석결과를 스스로 학습한 후, 이를 기반으로 판단이나 예측을 하게 됨. 

지도 학습: 문제와 정담을 모두 알려주고 공부시키는 방법. 정답이 있는 데이터를 사용함. 예측값을 이미 만들어둔 정답과 같아지도록 기계를 학습시킴. ex. 회귀, 분류

비지도 학습: 답을 가르쳐주지 않고 공부시키는 방법. 정답이 없는 데이터 사용. 데이터 속의 패턴 또는 각 데이터 간의 유사도를 기계가 학습하도록 하는 것. (스스로 패턴 파악) ex. 군집화

강화학습: 보상을 통해 상을 최대화, 벌을 최소화하는 방향으로 공부시키는 방법. 시행착오를 반복하여 정답을 찾는 과정. 


# 2. 회귀와 분류의 차이에 대해 설명해주세요.

회귀: 데이터가 연속형 변수를 예측하기 위해 사용될 때
분류: 데이터가 범주형 변수를 예측하기 위해 사용될 때
즉, 예측하고자 하는 값인 종속변수가 연속형이면 회귀를, 범주형이면 분류를 사용함. 

# 3. 분류 모델의 네 가지 종류와, 각 모델이 무엇인지 간단하게 정리해주세요.
1. 로지스틱 회귀: 독립 변수의 선형 조합에 로지스틱 함수를 적용하여 출력값을 0에서 1 사이로 변환해주는 것을 의미함. 이진분류 문제를 해결하는 데 적합함. 해당 이메일이 스팸인지 아닌지 등을 판별할 수 있음. 

2. 결정 나무: 조건에 따라 데이터를 분류하며, 최종적으로 데이터가 순수한 label의 집합으로 구성될 때까지 분류를 반복하는 모델. 

3. 서포트 벡터 머신: 클래스를 분류할 수 있는 다양한 경계선 중 최적의 라인을 찾아내는 알고리즘. 

4. KNN: 데이터로부터 거리가 가까운 k개의 다른 데이터 레이블을 참조하여 분류하는 알고리즘.

# 4. 분류 평가 지표에는 무엇이 있는지 작성해주세요.
1. 혼동 행렬: 분류 모델의 예측 결과를 정확한 예측과 잘못된 예측으로 구분하여 나타낸 표.

2. F1-Score: 머신러닝 모델의 성능을 평가하는 주요지표. 조화평균을 사용하는 이유는 정밀도와 재현율 간의 균형을 효과적으로 평가하기 위함임.

3. ROC / AUC Curve: 분류 모델의 성능을 평가하는 중요한 도구임. 

4. 다중 분류 평가지표: 각 클래스별로 구한 평가 지표를 클래스의 샘플 수로 가중 평균하여, 전체적인 모델 성능을 평가하는 지표. 


# 5. 하이퍼파라미터 최적화가 무엇인지 쓰세요.
하이퍼파라미터: 모델 학습 과정에 반영되는 값으로, 학습 시작 전에 사용자가 직접 설정하는 변수.
하이퍼파라미터 최적화: 적절한 하이퍼파라미터를 찾아 모델 성능을 향상시키는 것.
