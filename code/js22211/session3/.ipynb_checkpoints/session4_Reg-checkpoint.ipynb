# 회귀
## 머신러닝
* 비지도 학습 : 정답이 없는 데이터로 학습 
* 지도 학습 : 정답이 있는 데이터로 학습
* 머신 러닝 : 새로운 데이터를 예측하거나 결정을 내릴 수 있도록 하는 기술. 데이터에서 패턴 찾고 패턴 기반 예측
## 모델링
* 훈련 데이터 : 모델을 학습시킬때 사용하는 데이터
* 테스트 데이터 : 학습된 모델을 검증하기 위해 사용하는 데이터
전체 데이터를 두개의 부류로 나누는 이유는? 예측을 얼마나 잘 수행하는지 공정하게 평가하기 위해서
## 회귀
### def
1. 데이터에서 패턴을 찾아내 미래값을 예측하는 것<br>
2. 통계, 머신러닝의 주요 개념
3. 연속적인 숫자를 다룬다. <br>
### kind
선형, 비선형, 로지스틱, 릿지, 라쏘, 다항 <br>
### 선형 회귀
* 선형 회귀란 데이터를 가장 잘 설명하는 회귀선을 찾아 예측하는 과정.
* 보고 싶은 것을 종속변수, 예측을 위해 사용하는 변수를 독립 변수라고 한다.
### 종류
1. 단순선형회귀 : 데이터를 직선 형태로 표현
2. 다중선형회귀 : 데이터를 2개 이상의 독립 변수를 사용하여 표현
### 단순선형회귀 분석
#### 단순이란?
* 하나의 독립 변수
* 직선
* 기울기 (회귀 계수)와 절편 찾기
#### 기울기와 절편을 어떻게 찾을 수 있을까? 최소 제곱법과 최고 자승법
* Sum of Squared Residuals (잔차제곱의 합) 사용. 최소가 되는 지점으로 최적의 회귀선을 구하는 방식
* key point : 회귀선과 실제 데이터와의 차이를 최소화하기 위한 시도이다.
* 최소제곱법으로 힘들때는 경사하강법을 사용한다.
### 다중선형회귀분석
#### 다중이란?
* 독립 변수가 2개 이상인 경우 선형회귀
* 독립 변수 x_i가 2개인 경우 아래와 같이 3차원 공간에 표현된다. 회귀선은 평면.
* 최소제곱법 사용. 관측치와 평면간의 차이가 잔차
#### 다중공선성
독립 변수들 간에 강한 상관관계가 있을 때 발생하는 문제이다. 즉, 회귀 분석에서 여러 독립 변수가 있을때, 이들 독립 변수 중 일부가 서로 강하게 연관되어 있으면 예측 모델의 신뢰성과 정확성에 문제가 생길 수 있다. 
##### 다중공선성 확인방법
1. 상관계수 : hitmap, corr() 함수를 사용하여 쉽게 확인 가능
2. VIF지수 : [ 1 : 다른 변수와 상관관계 없음, 1이상 5미만 : 다중공선성 문제 X, 5이상 10미만 : 다중공선성 문제 p up, 10 ~ : 심각 & 회귀 모델 불안정 ]
##### 다중공선성 대처방법
1. 변수 제거
2. 변수 변환
3. 규제 선형 모델 사용 : 릿지, 라쏘, 엘라스틱넷
4. PCA(주성분 분석) : 데이터 차원 줄이기
### 모델 평가법
1. MSE
2. RMSE
3. MAE
#### 변수 유성 평가
* 회귀계수가 0이라면 의미 없는 변수, 이 계수가 0인지 아닌지로 가설을 세운다.
* t- 검정은 각 독립 변수의 회귀계수가 유의한지를 검정하는데 사용된다.
* 유의성이 크다면 종속 변수에 영향을 주는 변수라고 판단
* p-value가 0.05보다 작다면 해당 변수는 유의
