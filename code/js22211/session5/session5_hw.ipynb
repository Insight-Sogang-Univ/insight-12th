[ 다중 선형 회귀 ]

- contents
    
    다중 선형 회귀 모형은 단순 선형 회귀모형의 확장으로 연속형 종속 변수 하나에 설명 변수가 둘 이상인 모형을 말한다. 설명변수가 늘어나므로 추가 검토해야 할 문제는 설명 변수 정보 중첩으로 발생하는 다중공선성 문제와 너무 많은 설명 변수를 포함해 발생하는 차원의 저주 문제가 있다. 
    
    1. id를 제외한 모든 독립 변수를 사용하여 다중 선형 회귀분석을 한다. 
        
        → 그 후 선별한 독립 변수들로 다시 다중 선형 회귀분석을 해 본다 : 그냥 노가다로 다 해보고 그 중 가장 괜찮은 걸 고르기 
        
    2. 모든 설명 변수를 선택하고 다중 선형 회귀 모델을 학습시킨후 통계 장표를 확인한다. 
        
        → 모든 독립 변수들의 p-value가 유의 수준보다 작아 통계학적 유의성이 확인된다 . R-square 값을 통해 모델의 성능 파악도 가능
        
    3. 하지만 다중 선형회귀 분석에서는 다중공선성을 확인해야 한다. → 설명 변수가 많아 독립변수끼리 중첩이 발생한다.
    4. 모든 설명변수에서 한 5개의 설명 변수로 축소한 후 회귀 분석 진행 → 통계학적 유의성도 확인 & 다중공선성 문제도 줄어듦.

[ 회귀모형의 가정진단 ]

- contents
    
    회귀 모형은 반응 변수와 설명 변수의 선형 관계, 오차에 대한 독립성, 정규성, 등분산성을 전제로 한다. 앞서 확인한 모형의 유의성과 계수의 유의성이 확보되었더라도 오차에 대한 가정을 만족하지 않으면 다른 대안을 찾아야 한다 .
    
    정규성 확인법 : Q-Q 플롯, 잔차 패턴, 샤피로 윌크 검정
    
    독립성 검정 : 잔차 그래프, ACF를 사용
    
    등분산성 : 잔차 그래프를 통해 오차의 등분산성 가정을 만족시킨다.
