# 데이터 분석 프로세스
- 데이터 수집 -> 탐색(**EDA**: 파악) -> **전처리**(: 손질) -> 모델링


# EDA
## 개념
- **데이터**를 여러 방식으로 **파악**
- by 패턴 발견(시각화) or 특이성 확인 or 가설 검정(통계, 그래프)

## EDA: 미리 고려할 방향성
- **대상**: 일변량(패턴) / 다변량(변수 간 관계)
- **종류**: 시각화 / 비시각화

## EDA 유형
### 일변량
- **비시각화**: 분포 확인
- **시각화**: 전체적으로
### 다변량
- **비시각화**: 변수 간 관계
- **시각화**: 변수 간 관계 전체적으로

# 데이터 확인
## 1. 데이터 읽고 쓰기
- 절대 / 상대 경로
- 입력(read_) / 출력(to_)
- csv / xlsx 자주

## 2. 데이터셋 파악
- *프레임 -> 변수(column) -> 분포(그래프)*
- 변수 유형: 범주형(명목:단순 분류/순서) / 수치형(이산:정수/연속:소수)


# 전처리
## 결측값 처리 = 누락
*확인: NAN / ? / 0*
- **삭제**: 행(레코드)/열(변수), 전체(유효성)/부분(관리비용)
  *결측값 무작위일 때만!!!!*
- **대체**: 평균값, 최빈값 활용 -> 일괄(모든 변수)/유사 유형(범주형 변수)

## 이상치 처리 = 범위에서 많이 벗어난
*확인: 통계 / 시각화 / Z-score / Tukey Fences*
- **전체 삭제**: human error&관측치 많음
- **대체**: human error&관측치 적음 ->평균 or 예측모델 등
- **변수화**: 자연 발생 -> 이상값 파악(종속만 outlier) -> 특성 추가
- **리샘플링**: 자연 발생 -> 이상값 분리(설명, 종속 0utlier) -> 각각 모델 만듦

## 변수 가공(피처 엔지니어링) = 변수 형태 변형&처리
*for 기존 데이터 보다 유용하게*
- **레이블인코딩(0~n-1)/원핫인코딩(0/1)**: 텍스트 -> 숫자로
- **구간화**: 연속 -> 이산: 원칙x
- **변환**: 기존 변수 -> 다른 변수: 원칙x
- **스케일링**: 변수 값 범위 비슷하게: 표준화(평0분1, 정규분포)/MinmaxScaler(0~1,데이터 비율 같게)

# 시각화
*그래프, 차트, 그림, 도표 -> 패턴, 관계, 추세*
*Matplotlib, Seaborn 활용*

## 파라미터 = 매개변수
프로그래밍된 함수의 입력값

## 그래프
- boxplot: 사분위수, 이상치
- countplot: 범주, 빈도
- histplot: 수치, 구간별 빈도
- distplot, kdeplot: 히스토그램 곡선으로
- barplot, pointplot: 범주 x에 따른 수치 y 평균
- scatterplot: 산점도(두 변수 간의 관계)
- regplot: 산점도+회귀선
- catplot: 수치, 범주의 관계
- pieplot: 부분, 전체 간의 비율
- heatmap: 변수 간 상관계수
- violinplot: boxplot + distplot, kdeplot
- pairplot: 여러 변수 간 산점도 한번에
