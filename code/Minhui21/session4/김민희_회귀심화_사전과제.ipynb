# 선형 회귀
## 다중선형회귀의 수식적 이해
- 선형회귀식 = **설명변수의 가중평균** + 상수항 + 오차항
- 설명변수*가중치 -> **y를 얼마나 잘 설명**하는지 나타냄
  
## 다중선형회귀의 기본가정
- 신뢰성 향상을 위한 기본가정
- **선형성**: 종속변수와 설명변수 간의 관계=선형적
- **오차항의 평균=0**: white noise의 평균=0
- **독립성**: 설명변수 서로 선형독립적 <-> 다중공선성
- **등분산성**: 오차항의 분산 일정해야 <-> 이분산성:횡단면 자료에서 자주 나타남
- **오차항=자기상관 X**: 오차항 공분산 항상 0 <-> 자기상관: 시계열 자료에서 자주 나타남
- 정규성: 오차항=정규분포 따름 -> 필수 x

## 회귀분석 평가방법
- **시각화**: 데이터 전반 잘 요약 but 객관적 비교 어려움 -> **통계지표** 활용!

### 통계지표
- **R-square**:회귀선에 의해 설명되는 편차(SSR)/총 편차(SST)=1-회귀선에 의해 설명되지 않는 편차(SSE)/총 편차(SST) -> 모델이 데이터를 **얼마나 잘 설명하는지** 측정 -> 0~1: **1**에 가까울수록 모델 잘 설명
- **Adjusted R-squared**: 설명변수 개수가 늘어나면 당연히 R-squared(설명력) 늘어남 -> 당연함 but 오버피팅 문제 => so **변수 개수 증가에 덜 민감하도록 조정**
- **AIC,BIC(SC)**: 정보기준 / **낮을수록 좋아** -> AIC: 복잡성에 대한 페널티 적음 / BIC(SC): 더 엄격. 데이터 양에 따라 강한 페널티


# 비선형 회귀
- **다항식 회귀모델**: 선형 모델의 변수를 다항식으로 변형 -> 선형 회귀 사용 -> 회귀선 fit하게 그림 but 너무 많은 변수 -> 오버피팅 가능성 -> 릿지, 라쏘 규제 이용 가능
- **지수/로그 회귀모델**: 증감 형태가 지수적, 그 역인 경우: 기존 변수를 지수/로그로 변형 -> 선형 회귀 사용
