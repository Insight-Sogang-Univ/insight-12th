# 기술 통계
## 중심 위치 척도
- **평균**: .mean: 총합//개수 -> 이상치에 민감
- **중앙값**: .median: 가운데 위치 -> *이상치에 덜 민감*
- **최빈값**: .mode: 가장 자주 -> *이산형*
**-> skewness 확인**
  
## 변동성 척도
- **사분위수**: .percentile( , 퍼센트): 하위 25%/50%/75%/100%
- **IQR**: 25~75% -> 이상치: Q1,3 +- 1.5*IQR
- **범위**: .max-.min
- **분산**: .var: 편차 제곱 합 -> 데이터 얼마나 퍼져있는가
- **표준편차**: .std: 분산에 루트 -> *단위 일치성(분산보다 자주 씀)*
- **변동계수**: .std//.mean*100 -> 두 표본 집단 비교할 때 기준 같게 만들어줌
  
## 연관성 척도 - 두 변수
- **공분산**: .cov: 함께 어떤 방향으로, 얼마나 변화? -> *선형 관계만 체크*
- **상관계수**: [].corr: 공분산 표준화 -> -1~1 사이 / **인과관계와 다름**

*모집단: 전체 / 표본집단: 임의추출된 n명 -> 표본으로 모집단 추정*
*표본 많은 게 유리*

## 확률분포
: 확률변수가 특정한 값을 가질 확률 -> 함수
- 확률변수:어떤 사건의 결과 -> 숫자 대응
- 이산 / 연속: 이항, 포아송... / 정규, t, 카이제곱...

## 확률분포 종류
- **정규분포**: 평균 기준 대칭 / 표준편차 커지면 평평 / 평균, 표준편차 -> *모집단 분산 알 때*
- **t분포**: 표본 평균의 분포 / 정규분포st but 두꺼운 꼬리 / 자유도 커지면(=표본 크기 커지면, =더 많은 독립적인 정보 사용할수록) 표준정규분포에 수렴 -> *모집단 분산 모를 때*
- **카이제곱분포**: 0에서 시작, 오른쪽 긴 꼬리 / 자유도 증가 -> 정규분포 가까워짐 -> 독립성, 적합도 검정


# 추론 통계: 가설 검정
## 가설 검정의 절차
*가설 -> 유의 수준 -> 검정통계량 -> 기각/채택*

### **가설 설정**
- **귀무(H0)**:~는 차이가 **없다** / **기각**하고자 하는 명제 / 지지될 때 위험 적음
- **대립(H1)**:~는 차이가 **존재** / **주장**하고자 하는 명제 / 지지될 때 위험 큼

### 오류 -> 귀무가설(H0)
- **1종 오류(알파=유의수준)**: 실제 참 but 기각 -> *더 큰 문제*
- **2종 오류(베타)**: 실제 거짓 but 채택

### **유의수준 설정**
- 보통 **0.05**=5%
- 오류 허용, 감수 정도 / 신뢰성 판단

### **검정통계량 산출**:
- **z값**: 표준정규분포 -> 30 이상 표본 크기 가설 검정, 모집단 평균에 대한 가설 / |z값| 크면 평균으로 멀리 떨어져 있다, 임계치는 z=1.96:0.05, z=2.58:0.01
- **t값**: t분포, 작은 표본의 불확실성 반영 -> 작은 표본 크기 가설 검정, 모집단 평균에 대한 가설, 두 표본 평균 간의 차이 검정 / |t값| 크면 모집단 평균으로부터 멀리 떨어져 있다, 임계치는 t분포표. 같은 유의 수준이면 t값>z값&자유도 높아지면 z값에 수렴
- **f값**: f분포 -> 두 표본 분산 사이의 차이, 분산 분석에서 여러 그룹 간 평균 차이 검정 / f값이 1에 가까우면 두 표본 분산 서로 비슷

### **가설 기각/채택 판단**
- **유의확률(p-value)<유의수준(알파)**: 귀무가설 **기각**

## 검정 방법
### 수치-수치: (피어슨)상관분석
- 시각화(산점도)보단 **수치화(상관분석)** -> spst.pearsonr
- 유의성 검정! H0: 두 변수 간 선형관계 존재 x
- *결과: (상관계수, p-value)*
- **일반적으로 |상관계수|:0, 0.3, 0.5이 기준 , p-value<0.05 기각**
- 선형관계만, 직선 기울기 고려x

### 범주-수치: F-test, T-test
- **시각화**: *sns.barplot* / *검은 선=95% 신뢰구간* -> **두 평균의 차이 큼&신뢰구간 거의 겹치지 않음 => 대립가설 지지**
- **범주2-수치화**: **T-test** / *spst.ttest_ind* / H0(귀무가설): 두 그룹의 평균 차이x -> 기각 유무로 주장 / *결과: TtestResult(t값,p값,자유도)* -> **|t값|이 2보다 크면 차이 크다고 판단 => 귀무가설 기각 & p값 0.05 기준** / 단측 검정은 나누기 2
- **범주3이상-수치화**: **F-test** / (집단 간 분산//집단 내 분산)으로 계산 / H0(귀무가설): 여러 그룹의 평균 차이x / *spst.f_oneway* / **f값이 2~3이상 차이 크다고 판단 => 귀무가설 기각 & p값 0.05 기준** / 튀는 값 하나라도 있으면 차이 크다고 판단함

### 범주-범주: 카이제곱-test
- **시각화**: *mosaic* / 모자이크 플롯 -> **범주별 차이 명확 => 유의미한 관계 / 비슷 => 독립**
- **수치화**: **Chi2-test** / H0(귀무가설): 여러 그룹의 특정 여부는 관련이 없다 / [(관측빈도-기대빈도)^2//기대빈도]로 계산 -> **기대빈도와 관측빈도 차이 크면 => 유의미한 관계일 확률 높음, 자유도의 약 2배 이상 -> 의미있는 차이** / 자유도 = (x변수 자유도)*(y변수 자유도) <- 범주형 변수의 자유도 = 범주의 수-1 => 마지막엔 자동으로 결정되니까! / *pd.crosstab -> spst.chi2_contingency* / *결과: (카이제곱 통계량, p값, 자유도, 기대빈도)*
