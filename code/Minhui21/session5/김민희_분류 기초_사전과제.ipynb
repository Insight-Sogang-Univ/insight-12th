# 지도 학습과 비지도 학습의 차이
- **지도**: **정답 있는** 데이터로 학습 -> **예측값 => 정답**
- **비지도**: **정답 없는** 데이터로 학습 -> **데이터 패턴, 데이터 간 유사도**

  
# 지도학습 중 회귀와 분류의 차이
**회귀**: **연속형** 변수 예측 -> 독립변수를 기반으로 **종속 변수를 잘 설명하는 함수** 구하는 것
**분류**: **범주형** 변수 예측 -> 기존 데이터가 어떤 레이블에 속하는지 패턴 인지 -> **새롭게 관측된 데이터 레이블 판별**


# 분류 모델: 4가지
## 로지스틱 회귀
- **이진 분류** 문제: 출력값 특정값 이상 1, 이하 0
- 목표: **특정 클래스에 속할 확률** 추정
- **시그모이드 함수**(출력값 0~1, s자 형태) 이용 -> 가중치 sigmiod(wx+b): w는 기울기와 비슷! b는 y절편과 비슷!
- 결과=모델이 예측한 확률->*odds*(=p(A)/1-p(A)) 활용 -> **비선형적**인 결과를 **선형** 결과로 해석.
  
## 결정 나무
- **조건**에 따라 데이터 분류 -> **분류 반복** until 데이터=>순수한 **label집합**
- **CART 알고리즘**: 데이터셋 임계값 기준으로 두 child로 나눔(최대한 동일한 객체로 모음)
  -> 임계값 기준: **불순도**(서로 다른 클래스 섞여있는 정도) **낮춤** = **지니 계수 낮춤**
- 파라미터 설정, 차이점 시각화, Prunning(가지치기) 고려하면서!!!
  
## 서포트 벡터 머신(SVM)
- 클래스 분류 경계선 중 최적 라인 찾는 알고리즘
- 최적 라인=**margin(선과 각 점의 거리)이 가장 큰 선**
  
## KNN
- 가정: **비슷한 특성** 데이터 = 서로 **가까이** 존재
- **거리가 가까운 k개**의 다른 데이터 **레이블을 참조**하여 분류
- 과정: 데이터 준비(**선행학습 과정x**) -> k값(가장 가까운 이웃 개수,홀수로) 설정 -> 거리 계산 ->가장 가까운 k개의 이웃 선택 -> 분류
  
# 분류 평가 지표
## 혼동 행렬: 이진
- 예측 성공 여부(T/F)(실제값과 예측값 일치 여부)+예측값(P/N)=(참/거짓): p/n 먼저 판단하고 실제값과 같은지 비교해 t/f
- **정확도**: T_/all -> 예측 정확도 높음
- **정밀도**: TP/P -> "FP" / 거짓을 참으로 예측한 정도 낮음
- **재현도**(recall): TP/TP+FN -> "FN" / 참을 거짓으로 예측한 정도 낮음
- 정밀도와 재현도는 모델의 종류와 역할에 따라 중요도 달라질 수 있음
- **Threshold는 재현도와 반비례 / 정확도와 정비례**
- **정밀도&재현도 그래프** 활용 -> 두 value **만나는** 지점 => Threshold 결정 -> **예측 오류 최소화**
  *Threshold: 모델 참 거짓 가르는 분류 기준*
  
## F1-score: 이진
- **정밀도&재현율의 조화평균**: 정밀도 재현율 간의 균형 평가
- **머신러닝** 모델 성능 평가
  
## ROC/AUC Curve: 이진
- **ROC**: 얼마나 분류가 잘 되었는가! -> TPR, FPR 그래프에 표현
- **AUC**: ROC와 x축 사이 적분값 -> 0~1 1에 가까울수록 좋아
- **분류** 모델 성능 평가
  
## 다중 분류 평가 지표
- 위 이진 분류 평가 지표 클래스별 점수 -> 적절히 평균 내는 것
- **Macro** average: 모든 라벨 유사한 중요도 / 클래스별로 구한 평가 지표의 **평균**
- **Weighted** average: 샘플 많은 라벨에 중요도 / 클래스별로 구한 평가 지표의 **가중평균** -> 가중치: 빈도 높은 클래스
- **Micro** average: 전체 성능 / **모든 클래스의 예측 결과 합**
