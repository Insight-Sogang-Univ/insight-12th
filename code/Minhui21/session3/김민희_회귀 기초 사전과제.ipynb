# 머신러닝
*데이터에서 **패턴** 찾아냄 -> 새로운 데이터 **예측**, 결정 내릴 수 있도록 하는 기술*
- **지도학습**: 정답 o 데이터로 학습 ex. 회귀, 분류
- **비지도학습**: 정답 x 데이터로 학습 ex. 군집화/클러스터링
- **강화학습**: reward를 통해 학습 ex. model based, model free


# 모델링: 머신러닝 모델 생성
- 절차: 전체 데이터 -> 훈련데이터 / 테스트 데이터 분리 for 공정 평가 -> **훈련데이터로 모델 학습** -> **테스트 데이터로 반복 성능 평가** -> 최종모델 생성


# 회귀란?
*데이터에서 패턴 찾아냄 -> 미래 값 예측*
- 연속된 숫자를 다룸

## 선형회귀
*데이터를 가장 잘 설명하는 **회귀선y=ax+b**을 찾아 **예측**하는 과정*
- **독립변수x**를 통해 **종속 변수y**를 예측

## 단순선형회귀분석
*데이터를 **하나의 독립변수=직선**로 표현*
- y=ax+b -> 적절한 **a=회귀계수, b=y절편**을 찾는 과정

### 최적의 회귀선: 회귀선과 실제 데이터와의 차이 최소화! 
- **잔차의 제곱합 최소**: **최소 제곱법, 최소자승법**

## 다중선형회귀분석
*데이터를 **두 개 이상의 독립 변수**로 표현*
- 최소제곱법 사용 가능
  
### 다중선형회귀분석에서 다중공선성 체크 필수!
*독립 변수들끼리 상관계수 크면 -> 회귀분석의 정확도 낮아짐*
- **상관계수, VIF지수** 높으면 다중공선성 존재한다고 판단

### 다중공선성 대처 방법
- 변수제거 / 변수 변환 / 규제 선형 모델 활용(모델 복잡도 낮추기) / PCA=주성분분석(중요한 정보는 보존하면서 독립변수를 합쳐 *데이터의 차원 줄이기*)

  
# 모델 평가방법-모델 성능 및 유의성 평가
- **성능 평가 지표**: 평균 *제곱* 오차(MSE)->이상치 민감 / 평균 *절대* 오차(MAE)
- **변수 유성 평가**: t검정 -> 각 독립 변수의 회귀계수가 유의한지(0이 아닌지) / p-value<0.05일 때 유의
