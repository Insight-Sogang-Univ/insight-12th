1. EDA?

탐색적 데이터 분석의 과정.

 -> 데이터 분석(모델링)을 위해 데이터를 여러 방식으로 파악하는 모든 과정!

 -> 데이터 분석 과정에 해당 데이터에 대한 탐색과 이해가 전제되어야 한다는 것!

 -> 이 단계에서 데이터를 잘못 해석하거나, 문제를 올바르게 정의하지 못한다면 분석 난항!
 
EX) 시각화 / 특이성 확인 / 통계 및 가설 검정

- EDA 대상

  일변량 : EDA를 통해 한 번에 파악하려는 변수가 한 개 / 데이터 설명, 패턴 찾는 것이 목적

  다변량 : EDA를 통해 한 번에 파악하려는 변수가 여러 개 / 여러 변수 간 관계 찾는 것이 목적

 - EDA 종류
   
   시각화 : 차트, 그림 활용 데이터 확인 (대략적인 형태를 파악할 때)
   
   비시각화 : 주로 Summary Statistics를 통해 데이터 확인 (정확한 값이 필요할 때)

 - EDA 유형 : EDA 대상과 종류에 따라 4가지로 구분한 것
   
   일변량 시각화 : 전체적으로 살펴보기

   일변량 비시각화 : 데이터 분포 확인

   다변량 시각화 : 변수 간 관계를 전체적으로 살펴보기

   다변량 비시각화 : 변수 간 관계를 확인

2. 데이터 확인

데이터 읽기 및 쓰기 : 데이터를 분석하려면 데이터가 필요

* 데이터를 불러오는 것이 읽기, 새로운 데이터를 만드는 것이 쓰기

 - 절대경로 / 상대경로

   절대경로 : 처음부터 시작하여 목적지까지 전체적인 경로

   ex)C:\Users\owner\Desktop\t\insight-12th\code\YOUNGJIN-CHO-1

   상대경로 : 현재 작업 디렉토리 기준, 상대적으로 표현되는 경로

   ex) ./ : 현재 디렉토리, ../: 최상위 디렉토리 등

 - 데이터 입출력
   
   데이터 파일의 형식에 따라 데이터를 읽고 쓰는 함수가 다름!

   ex) read_csv, to_csv / read_json, to_json 등

   -> CSV와 EXCEL을 가장 많이 다룸!

   cf) CSV : comma separated value 의 약자, 데이터를 쉼표로 구분하는 텍스트 파일, 데이터 크기가 작고 압축에 용이

   cf2) EXCEL : 행과 열이 데이터프레임의 행, 열로 일대일 대응

   두 형식 모두 pandas를 통해 DataFrame 형식으로 읽어올 수 있음

 - 데이터셋 파악하기

   * 데이터프레임 보기 By head(n)
   * 데이터 변수(데이터프레임의 column = feature) 확인 (변수 정의, 유형, 데이터 형식 확인)

   cf) 데이터의 종류 : 범주형 / 수치형

   범주형 : 몇 개의 범주로 나누어진 데이터 (명목형 / 순서형)

   수치형 : 숫자로 표현되는 데이터 (이산형 / 연속형)

   * 데이터 분포 확인 By 산포도, 박스플롯 등 -> 전처리 아이디어 얻기!

3. 전처리

데이터 분석(모델링)을 위해 데이터를 적절한 방식으로 손질하는 과정

데이터 분석의 정확도는 데이터 품질에 의해 좌우

 -> 누락 데이터, 중복 데이터 등의 오류 수정, 분석 목적에 맞게 변형!

 - 결측값 처리

결측값 : 데이터 수집 과정에서 측정되지 않거나 누락된 데이터

결측값은 변수 간 관계를 왜곡 시키고, 정확성을 낮춤 + 시각적 표현에도 문제!

 * 결측값 확인 : 데이터프레임 상 NaN, ?, 0 등의 값으로 나타나는 결측값 확인

By info(), value_count(dropna = False), isnull(), notnull(), replace() 등

 * 결측값 처리
   1. 삭제 : 데이터가 있는 행 또는 열 삭제 By dropna() (결측값이 포함된 레이블 제거)
   - 행 삭제 : 분석 대상의 관측값 제거
   - 열 삭제 : 분석 대상이 갖는 특성 삭제
   - 전체 삭제 : 결측값이 발생한 모든 관측치 삭제 -> 관측치가 줄어 유효성 문제 발생 가능
   - 부분 삭제 : 모델에 포함시킬 변수 중 결측값이 발생한 모든 관측치 삭제 -> 모델에 따라 필요한 변수 다름, 관리 비용 증가
     
   *** 삭제는 결측값이 무작위로 발생했을 때 진행! 결측값이 무작위로 발생한 것이 아닌데 삭제할 경우 왜곡된 모델 생성 가능성 있음

   2. 대체 : 결측값을 다른 값으로 대체 (평균값, 최빈값 등 활용) By fillna()
    - 일괄 대체 : 모든 변수를 일괄적으로 같은 값으로 대체
    - 유사 유형 대체 : 범주형 변수를 활용, 유사한 범주에 따라 다른 값으로 대체
  
 - 이상치 처리

   이상치 : 관측된 데이터 범위에서 많이 벗어난 값

   이상치를 처리해주지 않으면 데이터 분석에 큰 영향, 알맞은 처리 진행해야 함
   
   * 이상치 확인

   a. 통계를 통해 확인 By describe()

   b. 시각화를 통해 확인 By BoxPlot

   c. Z-score를 통해 확인 : 평균으로부터 얼마의 표준편차만큼 벗어나 있는가

   d. Tukey Fences를 통해 확인 : 사분위 범위를 기반으로 판단, 데이터의 중간 50%범위만 측정 (상/하위 25% 제외)

   * 이상치 제거
     
   a. 전체 삭제 : 이상 값이 Human error에 의해 발생한 경우, 해당 관측치 삭제, 단순 오타 / 주관식 설문 등의 비현실적 응답, 데이터 처리 과정에서의 오류

   b. 다른 값으로 대체 : 절대적 관측치의 숫자가 작은 경우, 단순 삭제를 통해 이상치를 제거하면 관측치의 절대량이 작아져 신뢰성의 문제 발생

    -> 평균값 또는 예측 모델을 통한 예측값으로 대체

   c. 변수화 : 이상값이 자연 발생한 경우 단순 삭제나 대체를 통해 수립된 모델은 현상을 제대로 설명하지 못할 수 있음, 찬찬히 이상값을 파악하는 것이 중요 -> 특이점 또는 특징 찾기!

   ex) 전문직종 종사 여부를 통한 연봉의 이상치 변수화

   d. 리샘플링 : 자연발생한 이상값을 처리하는 또 다른 방법, 해당 이상값을 분리해 모델을 만드는 방법, 별다른 특이점이 없을 경우 케이스를 분리하여 분석하기!

 - 피처(feature) 엔지니어링 (변수 가공)

   해결하고자 하는 문제를 컴퓨터가 잘 이 해할 수 있도록 피처(변수)들의 형태를 변형하거나 적절하게 처리하는 과정

   cf) 피처(feature) : 입력 변수 = column

   도메인 지식과 기존의 변수를 사용, 기존의 데이터에 정보를 추가하는 일련의 과정을 피처 엔지니어링이라고 함

   기존의 데이터를 보다 유용하게 만드는 법

   도메인 지식 여부에 따라 변수 가공의 결과와 그 유의성이 달라지기도 함

   1. 레이블인코딩 / 원핫인코딩

      기계는 숫자만을 이해, 텍스트 -> 숫자 변환 과정 필요함

      레이블 인코딩(Lable Encoding) : 범주형 변수를 0~N-1까지의 숫자로 변환

      원핫인코딩(One-Hot Encoding) : 범주형 변수를 이진 벡터로 변환

   2. 구간화
  
      연속 데이터를 일정한 구간으로 나누어 분석하는 것이 효율적인 경우 有

      원칙이나 규칙이 따로 있는 것 X, 필요 영역에 따라 분석가의 도메인 지식 활용하여 수행

      Using pd.cut(데이터프레임['컬럼명'], bins=[나누 기준 리스트], labels=[지정할 label])

   3. 변환

         기존의 피처를 다른 피처로 변환하여 변수 추가

         정해진 원칙 X, 분석가의 의도에 따라 다양한 데이터 생성 가능

   4. 스케일링
     
         서로 다른 변수 값 범위를 일정한 수준으로 맞추는 작업

         숫자의 상대적 크기 차이로 인한 결과의 왜곡을 방지하기 위해!

         정규화(Normalization) 작업 수행 Using 로그함수, 제곱근 등

         라이브러리 활용 (from sklearn.preprocessing import StandardScaler, MinMaxScaler, QuantileTransformer, RobustScaler)

         a. StandardSscaler (표준화)

         각 피처의 평균을 0, 분산을 1로 변경 -> 모든 피처들이 같은 스케일을 가짐

         정규분포를 따른다고 가정하는 기술에 적합

         b. MinMaxScaler

         모든 피처가 0과 1 사이에 위치하도록..

         데이터가 서로 다른 비율의 속성으로 구성되어 있을 떄 같은 비율로 속성 맞춤

         연산 속도 향상, 알고리즘 최적화에 효과적
      

4. 시각화

   데이터나 정보를 시각적인 형태로 표현하는 과정 또는 결과물

   숫자와 텍스트로 표현되는 정보를 그래프, 차트, 그림, 도표 등 시각적 요소로 변환 -> 데이터 패턴, 관계 추세 파악

   Matplotlib (Numpy 기반 플로팅 라이브러리) / Seaborn (matplotlib 기반 통계 데이터 시각화 라이브러리) 활용

   데이터가 범주형인지, 수치형인지 / 결측값 및 이상치가 있는지 확인 -> 그래프를 어떤 것을 사용할지 결정

 - 파라미터 : 프로그래밍된 함수의 입력값, '키워드' = '인자' 형식으로 전달

   * 다양한 그래프



  
   (1) boxplot(상자 수염 그림) : 사분위수와 이상치를 보여주는 그래프

   (2) countplot, histplot

   countplot : 범주형 변수의 빈도수 확인

   histplot : 도수분포표를 그래프로, 수치형 변수의 구간별 빈도수

   (3) distplot, kdeplot(커널밀도추정 그래프) : 히스토그램을 연속적으로 곡선으로 연결한 그래프

   (4) barplot, pointplot

   barplot : 범주형 데이터 값 x에 따른 수치형 데이터 값 y의 평균 값 제공

   pointplot : 막대 그래프와 모양만 다르고 동일한 정보를 제공

   (5) scaatterplot(산점도 그래프), regplot(회귀선이 추가된 산점도 그래프)

   scatterplot : 두 변수 간 관계를 시각화하기 위해 사용되는 그래프

   regplot : 두 개의 연속 변수 사이의 산점도를 그리고 회귀선을 함께 나타내는 그래프

   (6) catplot : 수치형 데이터, 범주형 데이터의 관계를 볼 때 사용

   (7) pieplot : 데이터의 부분과 전체 간 비율을 표현하는 그래프, 비율을 강조하기 위해 사용( + 모든 데이터가 합쳐져 전체를 이루는 경우 효과적)

   (8) heatmap : 변수 간 상관계수를 직관적으로.. corr() 메서드 활용 변수 간 상관계수 구하고 이를 히트맵에 표현

   ex) sns.heatmap(df.corr(), annot = True) # annot은 상관계수 표시 여부

   (9) violinplot : 박스플록, 커널밀도추정 함수 그래프를 합쳐놓은 형태

   ex) sns.violinplot(x = '', y = '', data = '')

   (10) 여러 그래프 한 번에 찍기, pairplot : 여러 변수 간 산점도를 한 번에 보여주는 그래프
