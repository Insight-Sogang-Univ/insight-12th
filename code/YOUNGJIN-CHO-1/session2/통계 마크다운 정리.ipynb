1. 통계란?

   데이터를 수집, 분석, 해석하는 과정을 통해 패턴, 관계, 변동 등을 이해하고 결론을 도출하는  학문 및 방법론

   -> 우리 주변에서 발생하는 다양한 현상과 정보를 수집하고, 이를 분석하여 패턴을 찾아내거나 미래를 예측하는 데 도움을 줌

 - 기술통계학
   
   데이터의 주요 특성을 요약하고 설명하는 방법을 다루는 분야
   
   EX) 평균, 중앙값, 표준편차 등

   -> 측정이나 실험을 통해 수집한 자료를 정보로써 전환하기 위해 요약 및 정리하여 그 특성을 기술하는 절차에 관한 개념

 - 추론통계학

   수집한 데이터를 기반으로 미래의, 혹은 조사하지 못한 전 정보를 추론하고 결론을 도출하는 방법을 다루는 분야

   전체 집단에서 일부를 뽑아 만든 표본 집단을 통해 전체 집단에 대한 정보를 추측하는 과정

   EX) 통계적 가설검정, 신뢰구간
   

2. 기술통계 (수치 기술 통계)

- 중심 위치 척도
 
a. 평균 : 데이터의 총합을 데이터의 개수로 나눈 값, 데이터의 중심 경향성을 나타냄 / 평균을 총합치를 사용하므로 이상치에 민감 -> series.mean()

b. 중앙값 : 데이터 값을 크기 순서대로 나열했을 때 가운데 위치한 값, 이상치에 덜 민감, 데이터의 중심 경향성을 나타내는 대안적 지표 -> series.median()

c. 최빈값 : 데이터에서 가장 자주 나타나는 값, 주로 이산형 데이터에 사용, 데이터의 분포를 이해하는 데 도움 -> series.mode()

-> 이들을 활용해 자료의 개형(왜도 등)을 확인할 수 있음

cf. 왜도 : 데이터 분포의 비대칭성을 나타내는 통계적 지표 -> series.skew()

- 변동성 척도

a. 사분위수(quartile) : 데이터의 표본을 네 개의 동일한 크기로 나눈 값

np.percentile(arr, 25)

np.percentile(arr, 50)

np.percentile(arr, 75)

Q1 : 하위 25% (하위 사분위수) / Q2 : 하위 50% (중앙값) / Q3 하위 75% = 상위 25% (상위 사분위수)

b. 사분위간 범위 (IQR) : Q3에서 Q1값을 제외한 범위, 데이터의 중앙 50%가 퍼져 있는 범위 -> 이상치 탐지에 유용

아래쪽 이상치 : Q1 - 1.5*IQR보다 작은 값
위쪽 이상치 : Q3 + 1.5*IQR보다 큰 값

c. 범위 : 데이터의 최대값과 최소값의 차이를 나타내는 지표, 데이터의 전체 변동 범위를 알려줌np.max(arr) - np.min(arr)

d. 분산 : 데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지를 나타내는 측정 지표, 분산이 높으면 데이터가 많이 퍼져있음을 의미 -> 편차제곱의 평균으로 구함

e. 표준편차 : 데이터 값들이 평균으로부터 얼마나 퍼져있는지를 나타내는 중요한 통계적 지표, 값이 높으면 데이터가 많이 퍼져있음을 의미

표준편차를 사용하는 이유? 분산의 단위는 관측값의 제곱이므로 대상 데이터와 일치하지 않는 문제가 발생, 제곱근을 취해 단위를 일치시켜야 함 -> 단위 일치성의 문제에서 자유로운 표준편차를 사용, 데이터 해석 및 비교가 직관적으로 가능해짐

f. 변동계수 : 데이터의 절대적인 변동성이 아닌 상대적인 변동성을 나타내는 통계적 지표

표준편차를 해당 데이터의 평균으로 나눈 값으로 계산

np.std(arr) / np.mean(arr) * 100

변동계수를 사용하는 이유? 표준편차는 단위가 존재, 평균값이 다른 두 표본집단을 비교하거나 단위가 다른 두 데이터를 비교할 때 적합하지 않음 -> 표준편차를 평균으로 나눠주면 단위가 사라져 집단의 퍼진 정도를 동일한 기준으로 비교할 수 있음

 - 연관성 척도

a. 공분산 : 두 변수가 함께 어떤 방향으로 변화하는지와 그 크기를 표현하는 데 사용

np.cov(df['col1'], df['col2'])
 np.cov(df['col1'], df['col2'])[0][1

b. 상관계수 : 두 변수 간의 선형적인 관계를 나타내는 통계적 지표

-1과 1 사이의 값을 가짐, 상관관계(수치적 연결)와 인과관계(한 변수가 다른 변수를 발생)는 다르다는 점에 주의!

공분산을 각 변수의 표준편차로 나누어 -1과 1 사이로 표준화한 것

Corr (X,Y)이 양수면 양의 상관관계

3. 추론 통계1 (모집단 분포 추정)

모집단과 표본 집단

모집단 : 표본이 배경으로 하는 전체의 통계 단위 (모수 : 모집단의 특성을 나타내는 지표)

표본 집단 : 모집단에서 임의의 방법으로 일부를 통계의 자료로 선택한 부분

 -> 모집단과 모수를 정확하게 알 수 없지만, 표본과 표본 통계량을 통해 모집단 추정 가능

 대수의 법칙 : 표본의 크기가 커질수록 표본 평균이 모평균에 근사하는 현상, 데이터 많을수록 좋다!

 중심극한정리 : 표본의 크기가 크면 표본평균의 분포가 모집단의 분포와 상관없이 정규분포에 가까워짐

 - 확률 분포

확률변수가 특정한 값을 가질 확률을 나타내는 함수

확률변수? 어떤 사건이나 실험의 가능한 결과를 숫자로 대응시키는 함수

이산확률변수 : 확률변수가 이산적인 값을 가지는 경우, 유한하거나 셀 수 있는 값 ex) 베르누이분포, 이항분포, 포아송분포 등

연속확률변수 : 확률변수가 연속적인 값을 가지는 경우, 무한히 많은 값의 범위 ex) 정규분포, T분포, 카이제곱분포 등

 - 확률분포 종류

a. 정규분포 : 종모양, 평균 기준으로 대칭, 표준편차가 클수록 곡선이 평평 (평균 μ, 표준편차 σ)

b. T분포 : 종모양, 평균 기준으로 대칭, 정규분포보다 꼬리가 두꺼움, 표본으로 모평균을 추정할 때 사용, 표본의 크기가 작을 떄 정규분포보다 정확한 결과 제공 (자유도 df)

cf. 자유도? 얼마나 많은 정보가 독립적으로 변할 수 있는가를 나타내는 숫자 / 독립적인 정보의 수

t분포의 자유도는 표본의 크기에 비례 -> 자유도가 높을수록 t분포는 정규분포에 가까워짐(통계적 분석의 정확도 상승) But 너무 많은 자유도를 가지면 오버피팅(새로운 데이터에 대한 일반화 성능 저하되는 현상)의 문제 발생 가능

c. 카이제곱분포 : 0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭, 자유도가 증가함에 따라 중심이 오른쪽으로 이동, 대칭적으로 변화, 독립적인 표준 정규분포를 따르는 변수들의 제곱 합, 독립성 검정, 적합도 검정 등에 사용 (자유도 df)

4. 추론통계 2 (가설 검정)

 - 가설 검정의 절차

 가설 설정 -> 유의 수준 설정 -> 검정 통계량 산출 -> 가설 기각/채택 여부 판단

a. 가설 설정 : 귀무가설 H0와 대립가설 H1 설정

귀무가설은 주로 기각하고자 하는 명제로 설정 =  대립가설은 내가 주장하고자 하는 명제로 설정 -> 1, 2종 오류를 활용해 설정하기도 함

일반적으로 '~~와 ~~는 차이가 존재한다' 라는 명제

1종 오류 : 귀무가설이 실제로 참이었는데 기각할 확률

2종 오류 : 귀무가설이 실제로 거짓이었는데 기각하지 못할 확률

1종 오류를 더 치명적인 것으로 파악 -> 이를 활용하여 귀무가설 및 대립가설 설정 가능

b. 유의수준 설정 : 연구자가 결과를 판단할 떄 허용할 수 있는 오류의 최대치, 주로 0.05, 5%로 설정

유의수준은 연구자가 어떤 결정이 얼마나 신뢰할 만한지를 판단하는 기준

유의수준이 5%라는 것의 의미? 귀무가설이 실제로 참인데도 불구하고, 5%의 확률로 귀무가설을 잘못 기각할 수도 있다는 것, 연구자는 5%의 오류를 감수하며 귀무가설을 기각할 준비가 되어 있다는 뜻

c. 검정통계량 산출 : 가설 검정의 결과를 판단하는 데 사용되는 값

데이터의 분포나 특징에 따라 T값, F값 등 어떤 통계량을 사용할지 정해야 함

Z-score : 표준정규분포를 사용하여 얻은 값, 특정 데이터가 평균으로부터 얼마나 떨어져 있는지를 표준편차 단위로 나타냄, Z의 절댓값이 클 경우 비정상적이거나 이상치일 가능성이 있다는 것

T-score : T분포를 사용하여 얻은 값, 작은 표본의 크기에 대한 가설 검정에 주로 사용, 작은 표본에서 발생할 수 있는 불확실성을 잘 반영함, T값의 절댓값이 크면 헤당 데이터가 모집단 평균으로부터 멀리 떨어져 있으며, 관측된 표본 평균이 모집단 평균과 유의미한 차이가 있을 가능성이 높다는 것을 나타냄

F-score : F분포를 사용하여 얻은 값, 두 표본 분산 사이의 차이를 비교하는 데 사용, 두 개의 독립된 카이제곱 분포된 변수들의 비율로 정의, 1에 가까울수록 두 표본의 분산이 서로 비슷하다는 것을 의미(두 표본의 퍼진 정도가 비슷하다는 것)

d. 가설 기각/채택 판단

유의확률? 가설검정에 쓰일 수 있는 또 따른 척도, 검정통계량으로 유도 가능, 유의확률이 유의수준보다 작을 시 귀무가설 기각

H0 : 두 변수 간 선형 상관관계가 존재하지 않는다.

H1 : 두 변수 간 선형 상관관계가 존재한다.

abs(r) < 0.5 -> 두 변수 간 상관관계가 중간

abs(r) > 0.3 -> 두 변수 간 상관관계가 강함

+ p-value는 보통 0.05 이하를 귀무가설 기각의 기준으로 사용

범주가 2개 -> T-test 사용 (t값은 두 평균 간 차이를 표준 오차로 나눈 값)

H0 : 두 그룹의 평균에는 차이가 있다.

H1 : 두 그룹의 평균에는 차이가 없다.

cf. 만약 두 그룹의 평균에 차이가 없다는 것을 주장하고 싶다면? T-test 검정법 자체가 두 변수 간 차이를 기준으로 '얼마나 멀리 떨어져 있는지'를 체크하므로 차이가 없다는 가설을 주장할 수 없음 -> 간접적으로 귀무가설을 기각하지 못하였다! 라고 주장할 수 있음

spst.ttest_ind(a, b) -> 결과는 TtestResult(statistic = t-값, pvalue = p값, df = 자유도)의 형태

cf) ind : independent ; 독립적인 집단 / ttest-rel 에서 rel : related ; 같은 그룹이 다른 상태로..

만약 단측검정을 하고 싶다면? p-value를 2로 나누어주기 (T분포는 좌우 대칭이므로)

범주가 3개 이상 -> 분산분석(anova) 사용(F-test의 일종)

F-통계량은 위에서 언급된 것처럼 분모 그룹과 분자 그룹의 분산 비

F값은 집단 간 분산/집단 내 분산으로 계산

H0 : 두 그룹의 평균에는 차이가 있다.

H1 : 두 그룹의 평균에는 차이가 없다.

spst.f_oneway(A, B, C) -> 결과는 F_onewayResult(statistic = f값, pvalue = p값)

** anova는 전체 평균 대비 각 그룹 간 차이가 있는지만 알 수 있음, 모두가 서로 다르다가 아닌 하나가 튀고 나머지가 비슷한 상황에서도 귀무가설 기각할 수 있음 (-> 사후분석을 진행하여 차이가 있다는 것을 파악한 후 어떤 범주가 튀는지를 파악하는 작업)
