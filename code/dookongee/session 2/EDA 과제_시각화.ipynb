# <span style="background-color:#E6E6FA"> regplot</span>

▶ 회귀선이 추가된 산점도 그래프 <br>
▶ 두 개의 연속 변수 사이의 산점도를 그리고 선형회귀분석에 의한 회귀선을 함께 나타냄

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x='Age',y='Fare',data=data,ax=ax1) ## 회귀선 있는 버전
sns.regplot(x='Age',y='Fare',data=data, ax=ax2, fit_reg=False) ## 회귀선 없는 버전

plt.show()
```

### (10) 나머지 칸에 본인이 파악하고 싶은 정보를 자유롭게 시각화하고, 인사이트를 작성해보세요.

Q. 알아낸 것은? <BR>
A1. 연령대 분포 그래프를 보면 Pclass 1에는 비교적 10세 이하의 아동의 인구가 적고 50 ~ 60세의 인구들의 분포는 많은 것으로 보인다. Pclass1은 가장 비싼 등급이다. 많은 돈이 드는 만큼 가족여행으로는 Pclass1 등급을 선택할 가능성이 낮아서 10세 이하의 아동이 적은 것으로 파악할 수 있고, 50 ~ 60세의 인구들은 돈을 더 내더라도 더 편한 등급에 머무는 것을 선호하는 경향이 있다고 볼 수 있다.<BR>
A2. Family는 SibSp와 Family를 더한 값이 0보다 큰 값들이고 Alone은 더한 값이 0인 경우이다. 즉 Alone은 가족 없이 혼자 탑승한 경우이다. 그래프르 보면 Family의 경우 가족과 동승한 Family의 경우에서는 생존 여부가 비슷하지만, 가족과 동승하지 않고 혼자 탑승한 경우는 생존하지 못한 경우가 생존한 경우보다 더 많았다. 가족이 없이 혼자 탑승한 경우 생존자 수가 더 많을 것으로 예상했는데 반대의 결과가 나왔다. 일반적으로 노인이나 여성, 어린아이 보다는 젊은 층의 남성이 혼자 탑승하는 경우가 더 많고, 젊은 남성의 경우 노인, 여성, 어린아이보다 구조 순서가 뒤로 밀리기 때문에 다른 사람들을 도와주다가 더 많이 죽었거나, 배의 침몰 사실을 알려줄 사람이 없어서 늦게 확인했기 때문에 생존자보다 사망자 수가 더 많은 것으로 볼 수 있다.<BR>
A3. Embarked는 승선항으로 S 승선항에서 탑승한 사람들의 수가 가장 많은 것으로 보아 배가 S항에서 출발을 했거나, C,Q항보다 S항의 규모가 훨씬 큰 것이라고 유추해 볼 수 있다.

```python
# 자유롭게 코드 작성
f, ax = plt.subplots(2, 2, figsize=(12, 4))
data['Family']=data['SibSp']+data['Parch']
data['Family_onehot']=data['Family'].apply(lambda x:'Alone' if x==0 else 'Family')
sns.countplot(data=data,x='Embarked',ax=ax[0,0])
sns.scatterplot(data=data,x="Age",y="Pclass",alpha=0.3,ax=ax[0,1]) # alpha: 0에 가까울수록 투명, 1에 가까울수록 불투명
sns.countplot(data=data,x='Family_onehot',hue='Survived',ax=ax[1,0])
sns.barplot(data=data,x="Pclass",y="Survived",ax=ax[1,1])
plt.tight_layout() #플롯 간의 간격을 자동으로 조정해서 보기 좋게 만들어준다.
plt.show()
```
