# 통계

통계란 데이터를 수집, 분석, 해석하는 과정을 통해 데이터의 패턴을 파악하고 결론을 도출하는 것을 의미한다. 

## 기술통계

데이터의 평균, 중앙값, 표준편차 등이 해당되며 데이터의 기본적인 특성을 파악할 수 있다. 기술통계는 숫자 형태로 데이터를 요약하는 "수치 기술 통계"와 표와 그래프로 데이터를 시각화하는 "표와 그래프 기술 통계"로 구분된다.

## 중심 위치 척도
- 평균: 데이터를 모두 더한 값을 전체 데이터 개수로 나눈 값이다. 데이터의 총 합을 사용하므로 1,2,5,7,3,100처럼 100의 이상치가 존재하는 경우 평균 값이 유의미한 정보를 제공하지 못한다.
- 중앙값: 데이터를 크기 순서대로 정렬했을 때 가운데 위치한 값이다. 이상치가 존재하더라도 데이터의 중심 경향성을 나타낼 수 있다.
- 최빈값: 데이터에서 자주 등장한 값으로, 등장 횟수가 가장 많은 데이터를 의미한다. 연속형 데이터보다는 이산형 데이터에서 주로 사용된다.

평균(mean), 중앙값(median), 최빈값(mode)을 통해 데이터 분포의 개형(Skewness)를 확인할 수 있다. Skewness는 분포의 비대칭성을 설명하는 통계적 개념으로 데이터가 평균값을 기준으로 얼마나 치우쳐 있는지를 나타낸다. 
Positive Skewness의 경우 데이터의 많은 부분은 왼쪽에 집중돼 있지만 소수의 큰 값들이 오른쪽에 존재한다. ex) 소득 분포, 부동산 가격
Negative Skewness의 경우 데이터의 많은 부분은 오른쪽에 집중돼 있지만 소수의 작은 값들이 왼쪽에 존재한다. ex) 시험 성적 분포
Symmetrical Distribution의 경우 데이터의 좌우가 균등한 경우이다. ex) 키나 체중의 신체 값

## 변동성 척도
- 사분위수: 데이터를 4개의 동일한 크기로 나눈 값으로 하위 25%에 속한 값은 Q1, 하위 50%에 속한 값은 Q2, 하위 75%에 속한 값은 Q3이다. Q2는 중앙값에 해당한다.
- 사분위간 범위(IQR): Q3에서 Q1을 제외한 값으로, 데이터의 중앙 50%가 속한 범위이다. IQR지수를 사용해서 이상치를 탐지할 수 있다.
- 범위: 데이터의 최대값과 최소값의 차이로, 전체 데이터의 변동 범위를 알 수 있다.
- 분산: 데이터가 얼마나 퍼져있는지를 나타내는 지표로, 분산 값이 크면 데이터가 많이 퍼져있는 것을 의미한다.
- 표준편차: 분산에 제곱근을 적용한 값이다. 분산과 마찬가지로 데이터가 얼마나 퍼져있는지를 나타낸다. 분산은 데이터를 제곱하여 사용하므로 단위가 일치하지 않는다. 따라서 분산에 다시 제곱근을 취한 표준편차를 사용하여 단위를 일치시킬 수 있다.
- 변동계수: 표준편차를 평균으로 나눈 값으로 표준편차가 평균에 비해 얼마나 큰지를 나타낸다. 변동계수를 사용하면 서로 다른 단위를 가진 데이터들 간의 데이터 퍼짐 정도를 비교할 수 있다. 

## 연관성 척도
- 공분산: 두 변수의 관계를 나타내는 값이다.
  - Cov(X,Y)가 양의 값: X가 증가할 때 Y도 같이 증가한다.
  - Cov(X,Y)가 음의 값: X가 증가할 때 Y는 감소한다.
  - Cov(X,Y)가 0: X와 Y사이에 관계가 없다.
- 상관계수: 두 변수 간의 선형적인 관계를 나타내는 지표이다. 공분산을 표준편차로 나눠서 -1과 1사이의 값으로 표준화한 것이다. 

## 추론 통계

수집한 데이터를 통해 미래의 정보를 추론하고 결론을 도출하는 방법이다. 모집단에서 일부를 추출해서 표본집단을 만들고 이것을 통해 전체 집단에 대한 정보를 추측하는 과정으로 통계적 가설검정과 신뢰구간이 해당된다. 

## 모집단 분포 추정
- 모집단은 전체 데이터 집합을 의미하고, 표본집단은 모집단에서 일부를 뽑아서 만든 소규모 집단을 의미한다. 
- 성인 평균 키를 구하는 예시에서 우리나라 모든 성인의 키를 조사할 수 없기 때문에 표본집단을 사용하여 N명의 키를 추출하고 N명의 통계량을 통해 전체 모집단의 데이터를 추정할 수 있다.
- 표본집단을 통해 모집단을 추정하는 것은 "대수의 법칙"에 의해 가능하다. 대수의 법칙이란 표본의 크기가 커질수록 표본 평균이 모평균에 근사하는 현상이다. 따라서 데이터가 많을 수록 표본집단을 통해 추정한 모집단이 실제 모집단에 더 근접해진다. 

### 확률 분포
- 확률분포는 확률변수가 특정한 값을 가질 확률을 나타내는 분포이다. 확률분포는 이산확률분포와 연속확률분포로 구분된다.
- 이산확률분포는 확률변수가 이산적인 값, 즉 유한한 값을 가지는 경우이다. ex) 주사위의 눈(1,2,3,4,5,6)
  - 베르누이 분포, 이항분포, 포아송 분포 등이 해당된다.
- 연속확률분포는 확률변수가 연속적인 값을 가지는 경우이다. ex) 사람의 키(170, 170.1, 170.0001, 170.00242 ...)
  - 정규분포, 지수분포, 베타분포, 균일분포 등이 해당된다. 

## 가설검정
- 가설 설정->유의수준 설정->검정 통계량 산출->가설 기각/채택 판단
- 가설 설정: 귀무가설과 대립가설을 설정한다.
  - 귀무 가설(H0): 주로 기각하고자 하는 명제로 설정한다. '~는 차이가 없다'라는 명제로 '집단 A와 B의 평균은 차이가 없다'와 같이 설정한다.
  - 대립 가설(H1): 내가 주장하고자 하는 명제로 설정한다. '~는 차이가 존재한다'라는 명제로 '집단 A와 B의 평균은 차이가 존재한다'와 같이 설정한다.
- 1종, 2종 오류
  - 1종 오류: 귀무가설이 실제로 참이었는데 기각될 확률. 즉 실제로는 차이가 없는데도 차이가 있다고 잘못 결론을 내리는 경우이다. 
  - 2종 오류: 귀무가설이 실제로 거짓인데 기각하지 못할 확률. 즉 실제로는 차이가 있는데도 차이가 없다고 잘못 결론을 내리는 경우이다.
  - ex) 귀무가설: 신약의 효과는 기존 약과 차이가 없다.
        대립가설: 신약의 효과는 기존 약과 차이가 있다.
        1종 오류: 신약의 효과가 기존 약과 차이가 없지만 신약이 기존 약에 비해 더 효과적이라고 잘못 결론을 내리는 경우.
        2종 오류: 실제로 신약이 기존 약보다 더 효과적이지만 신약과 기존 약의 효과가 차이가 없다고 잘못 결론을 내리는 경우.
- 유의수준: 연구자가 결과를 판단할 때 허용할 수 있는 오류의 최대치이다.
  - 유의수준이 0.05라는 것은 "귀무가설이 실제로 참인데도 5%의 확률로 잘못 기각할 수 있다"로 5%의 오류를 감수하며 귀무가설을 기각하겠다는 것을 의미한다.
  - 신약 예시에서, 유의수준이 0.05라면 신약이 기존 약과 효과가 차이가 없더라도 5%의 확률로 신약이 더 효과적이라고 결론을 내릴 가능성을 수용하는 것이다.
- 검정 통계량 산출: 가설 검정의 결과를 평가하는데 사용하는 값으로 Z-Score, T-Score, F-Score로 구분된다.
  - Z-Score: 표준 정규분포를 통해 얻은 값으로 큰 표본 크기의 경우에 주로 사용한다. Z의 절댓값이 크면 해당 데이터가 평균으로부터 멀리 떨어져 있는 것을 의미하며 때로는 이상치일 가능성이 있다.
  - T-Score: T분포를 통해 얻은 값으로 작은 표본 크기의 경우에 주로 사용한다. T값은 모집단의 평균에 대한 가설을 검정하거나 두 표본 평균 간의 차이를 검정할 때 사용한다. T의 절댓값이 크면 해당 데이터가 모집단 평균으로부터 멀리 떨어져 있는 것을 의미하며, 표본 평균이 모평균과 유의미한 차이가 있을 가능성이 높다는 것을 의미한다.
  - F-Score: F분포를 통해 얻은 값으로 두 표본 분산 사이의 차이를 비교할 때 사용한다. F값이 1에 가까우면 두 표본의 분산이 서로 비슷하다(데이터의 퍼짐 정도가 비슷하다)는 것을 의미한다.
- 가설 기각/채택 판단
  - 유의확률: 유의확률이 유의수준보다 작으면 귀무가설을 기각한다. 유의확률이 유의수준보다 크면 귀무가설을 기각할 수 없다. 

### 여러 검정 방법
- 수치형-수치형
  - 시각화: 산점도를 통해 관계를 볼 수 있지만 데이터가 많으면 효율적이지 못하다.
  - 수치화: 두 변수 간 피어슨 상관계수를 확인한다. 피어슨 상관계수의 절댓값이 0.3보다 작으면 두 변수 간의 상관관계가 약함, 0.5보다 작으면 상관관계가 중간, 0.5보다 크면 상관관계가 강함을 의미한다. 
- 범주형-수치형
  - 시각화: barplot을 통해 두 평균의 차이가 크고 신뢰구간이 거의 겹치지 않으면(두 집단의 평균이 유의미하게 다를 가능성이 높다) 대립가설을 지지할 가능성이 높다고 판단할 수 있다.
  - 수치화(변수가 2개): T-test를 사용한다. t값은 절댓값2를 기준으로, 2보다 크면 두 집단 간의 평균 차이가 크다는 것을 의미한다. 2보다 작으면 두 집단의 차이가 통계적으로 유의미하지 않을 수 있다는 것을 의미한다.
  - 수치화(변수가 3개): 변수가 3개라면 2개씩 묶어서 T-test를 3번 진행하거나 분산분석(ANOVA)를 사용한다. 
- 범주형-범주형
  - 시각화: 모자이크 플롯을 통해 눈으로 봤을 때도 범주별 차이가 명확하게 보인다면 범주간의 유의미한 관계가 있을 가능성이 크다고 판단한다.
  - 수치화: 기대빈도와 관측빈도를 사용한다.
- 수치형-범주형
  - 시각화: kde플롯을 통해 수치형이 변할 때 범주형이 어떻게 변하는지를 확인할 수 있다. 서로 관계가 없을 때는 kde플롯의 개형이 비슷하게 나타난다.
