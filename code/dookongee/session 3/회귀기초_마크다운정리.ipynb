# 인공지능


인공지능은 크게 AI, ML, DL로 구분된다. AI와 ML에서 data를 기반으로 학습했으면 ML, 그렇지 않으면 AI이다. ML과 DL에서 DNN(Deep Neural Network)를 활용하여 학습했으면 DL, 그렇지 않으면 ML에 해당한다.<BR>
그림에서 1번 영역에는 Rule-Based Algorithm이, 2번 영역에는 결정트리, 선형회귀, 퍼셉트론, svm 등이, 3번 영역에는 CNN, RNN, GAN 등이 해당된다.

# 머신러닝
이때 머신러닝은 4가지 분류로 나눌 수 있다.
### 지도학습
정답이 있는 데이터로 학습하는 방식으로, 인공지능에게 문제와 정답을 함께 제공해서 학습하도록 한다. 지도학습에는 회귀와 분류가 해당된다. 
### 비지도 학습
정답이 없는 데이터로 학습하는 방식으로, 인공지능이 문제만 알고 정답을 모르는 상태로 학습하는 것을 의미한다. 비지도학습에는 군집화가 해당된다.
### 자기지도 학습
데이터는 많은데 정답을 알고 있는 데이터가 너무 적은 상황에서 정답을 모르는 데이터의 정답을 만드는 것은 비용이 너무 비싸다. 따라서 원래 풀려고 했던 문제 대신에 다른 문제를 인공지능이 새롭게 정의해서 풀고 그것을 통해 원래 문제를 푸는 방식이다. <BR>
고양이를 분류하는 문제에 대해서, 고양이 사진의 특징을 학습시키도록 새롭게 문제를 정의하고, pre-trained된 모델을 transfer learning 해서 원래의 문제를 풀도록 한다. 
### 강화학습
이세돌9단과 알파고의 대국에서 알파고의 알고리즘에 관한 것으로, 인공지능에게 행동의 연속을 학습시키는 것이다. <br>
ex) 오목에서 흑돌 5개를 두면 승리하므로 이때 큰 보상을 준다. 그러면 인공지능은 흑돌 5개를 연속으로 두기 위해 흑돌 4개를 연속으로 두는 상황을 만들고, 4개를 연속으로 두는 상황을 위해 3개를 연속으로 두는 상황을 만들고, ... 이런 식으로 흑돌 5개를 뒀을 때의 상황에서부터 앞으로 이동하며 행동의 연속을 학습시키는 방식이다.

# 모델링
모델링은 일련의 과정을 통해 머신러닝 모델을 생성하는 것이다. <br>
### 모델링 과정
1. 전체 데이터를 train, test, valid 데이터셋으로 분류한다.
2. train 데이터셋으로 모델을 학습시키고 valid 데이터셋으로 학습이 잘 됐는지를 검증하며 하이퍼 파라미터를 조정하며 overfitting 되지 않는 선에서 적절히 학습을 중단시킨다.
3. 학습이 완료된 모델에 대해서 test 데이터셋으로 학습의 정도를 평가한다.
### 데이터를 train과 test로 쪼개는 이유
데이터를 train과 test로 구분하지 않고 한 번에 학습시키면 모델이 데이터를 학습하는 것이 아니라 데이터의 패턴을 단순히 암기하여 성능이 좋아질 가능성이 있기 때문에 학습용 데이터와 검증용 데이터는 반드시 분류해야 한다.

# 회귀
회귀는 입력과 출력 간의 관계를 파악하는 것이다. 관계를 학습하여 처음 보는 입력에 대해서도 적합한 출력을 얻는 것을 목표로 한다.<br>
ex) 키와 몸무게의 관계를 파악해서 처음 보는 키에 대해서도 적합한 몸무게를 출력하여 키에 대한 몸무게를 예측하는 모델을 만든다.<br>
회귀에는 선형회귀, 비선형 회귀, 로지스틱 회귀, 릿지 회귀, 라쏘 회귀, 다항 회귀 등이 존재한다.

# 선형회귀
선형회귀는 입력x와 출력y간의 관계를 가장 잘 설명할 수 있는 y=ax+b의 직선을 찾는 것을 목표로 한다. <br>
이때 예측하고자 하는 것이 y(종속 변수), 예측을 위해 사용하는 것이 x(독립 변수)라고 한다. 
### 단순 선형회귀, 다중 선형회귀
단순 선형 회귀: 데이터를 하나의 독립 변수로 표현하는 경우이다.<br>
다중 선형 회귀: 데이터를 2개 이상의 독립 변수로 표현하는 경우이다.
### 선형 회귀 기본 가정
선형 회귀를 사용하기 위한 조건으로는 선형성, 독립성, 등분산성, 정규성이 존재한다.

# 단순 선형 회귀 분석
하나의 독립변수만 존재하는 선형 회귀이다.<br>
입력x와 출력y의 관계를 가장 잘 설명하는 y=ax+b의 직선을 찾는 과정이다.<br>
즉 관계를 잘 설명하는 기울기(a)와 y절편(b)의 값을 찾는 것이 단순 선형 회귀의 목표이다.
### 최적의 회귀선 찾는 방법
적절한 a,b는 loss를 최소화할 때의 a,b값이다. 이때 loss는 모델의 예측(y^)과 실제 정답(y)의 차이이다. 즉 예측값과 실제값의 차이가 loss이다. <br>
이때 단순히 y-y^을 loss로 설정하면 y-y^이 +2와 -2로 나타나는 경우는 +2와 -2의 합이 0이므로 y값을 정확히 예측하지 못했음에도 좋은 모델로 간주된다. 따라서 |y-y^|처럼 절댓값을 씌운 것을 최소로 하는 것을 목표로 한다. <br>
그러나 절댓값만 씌우게 되면 오류에 대해서 덜 민감하다. y=|x|를 살펴보면 2의 오류에 대해서는 2만큼, 4의 오류에 대해서는 4만큼의 loss를 보인다. 그러나 y=x^2의 함수를 보면 2의 오류에 대해서는 4만큼, 4의 오류에 대해서는 16만큼의 loss를 보이므로 더 큰 오류에 대해서 더 민감하게 작동한다. 따라서 일반적으로 (y-y^)^2을 loss로 설정하고 이것을 최소화하도록 학습한다. 

# 경사하강법(gradient descent)
현재의 a,b 위치에서 loss를 줄이는 방향으로 학습하기 위해 경사하강법을 사용한다. <br>
경사하강법에서 gradient는 미분값으로 함수의 x=a에서의 기울기값에 해당한다. 함수의 특성 상 gradient를 줄이는 방향으로 a와 b를 조정하면 전체 loss 함수의 값이 작아지게 된다. (y=x^2의 함수에서 x좌표가 양수일때와 음수일때 모두 기울기가 작아지는 방향으로 나아가면 y절편이 0에 가까워지며 값이 작아진다.) <br>
경사하강법을 사용하여 P(k)가 k에서의 a와 b 값이라고 할때, P(k+1)=P(k)-gradient를 통해 loss를 줄이는 방향으로 학습해간다. 이때 너무 큰 보폭으로 학습하는 것을 방지하기 위해 gradient에 learning rate를 곱해서 사용한다. 

# 다중 선형 회귀 분석
독립변수가 2개 이상인 경우의 선형회귀이다.<br>
만약 독립변수가 2개라면 단순 선형 회귀 분석에서의 회귀선은 평면이 된다. <br>
따라서 단순선형 회귀분석에서와 마찬가지로 예측값에 대해서 평면과의 차이를 줄여나가는 방향으로 학습을 진행한다. 
### 다중 공선성
다중 공선성은 독립 변수들 간에 상관관계가 큰 경우 발생한다. <br>
다중 공선성이 높은 경우 어떤 독립 변수가 종속 변수에 얼마나 영향을 미치는지를 정확히 구분하기 어렵기 때문에 어떤 변수의 영향을 반영해야 할지 불확실해진다. 따라서 회귀 분석의 정확도가 낮아진다. 
### 다중 공선성 확인-상관계수
다중 공선성은 상관계수를 통해 확인할 수 있다. <br>
상관계수는 변수 간의 상관관계의 정도를 수치로 나타낸 것으로 -1과 1 사이의 값이다.<br>
-1이나 1에 가까울수록 두 변수의 상관성이 높다. 0은 두 변수의 상관 관계가 없음을 나타낸다.<br>
상관계수는 heatmap이나 corr()함수, pairplot을 통해 확인할 수 있다.
### 다중 공선성 확인-VIF 지수
VIF 지수는 분산 팽창 인수로 회귀모델의 결정 계수 R square를 사용하여 계산된다.<br>
- VIF=1: 해당 독립 변수는 다른 변수와 상관관계가 전혀 없다.
- VIF<5: 일반적으로 다중 공선성 문제가 없다고 간주한다.
- VIF>5: 다중 공선성의 징후가 있을 수 있다.
- VIF>10: 다중 공선성이 심각하다고 간주한다. 이경우 해당 변수를 제거하거나, 변수 간의 상관관계를 줄이기 위한 조치가 필요하다.
### 다중 공선성 대처 방법
1. 변수 제거(변수 선택법): 독립변수로 사용할 변수를 선택한다.
2. 변수 변환: 변수들을 더하거나 빼서 새로운 변수를 생성한다.
3. 규제 선형 모델 활용: 릿지, 라쏘, 엘라스틱넷 등의 방법을 통해 모델의 복잡도를 줄인다.
4. PCA: 데이터의 차원을 축소하여 데이터의 복잡성을 줄인다.
### PCA 과정
1. 기존의 데이터셋에서 평균을 빼서 centering을 한다.
2. centering된 데이터셋을 표준편차로 나누어서 표준화시킨다.
3. 공분산 행렬의 eigenvector와 eigenvalue를 계산한다. N차원으로 축소하는 경우 eigenvalue을 내림차순으로 정렬했을 때, 가장 큰 값부터 N개의 값을 사용한다.
4. 데이터를 주성분 공간으로 투영시킨다.
5. 데이터 표준화를 해제하고 주성분 공간에서 투영된 데이터를 원래 데이터 공간으로 변환한다. 이때 데이터의 주요 정보는 주성분 방향을 따르는 데이터로 남게 된다.고 판단.

# 모델 평가 방법
내가 만든 회귀 모델의 성능을 평가해야 성능을 향상시키기 위한 조치가 필요한지를 결정할 수 있다.

## 성능평가지표

#### 평균 제곱 오차(MSE)
<BR>
모델의 예측값과 실제 관측값 차이의 오차 제곱의 평균.

#### 평균 절대 오차(MAE)
<BR>
모델의 예측값과 실제 관측값 사이의 절대 오차의 평균.

#### MSE vs MAE
MSE는 큰 오차에 대해 민감하게 반응하지만 MAE는 모든 오차를 동일하게 취급한다. (위의 "최적의 회귀선 찾는 방법"에서 설명)<BR>
MSE는 이상치에 민감하게 반응하지만, MAE는 이상치가 존재해도 전반적인 예측 성능을 유지할 수 있다.다.

## 설명력 평가 지표

#### 결정 계수
<BR>
결정계수는 R^2으로 모델이 설명하는 데이터의 총 변동(평균과의 차이) 중에서 설명된 비율을 나타낸다.<BR>
R^2은 0과 1사이의 값으로, 1에 가까울수록 모델이 데이터를 잘 설명한다고 할 수 있다.<BR>
- SST(총 변동성): 데이터가 평균으로부터 얼마나 벗어나는지. 실제값과 평균값 사이의 차이에 대한 제곱합이다.
- SSR(설명된 변동성): 모델이 예측한 값이 평균으로부터 얼마나 벗어나는지. 예측값과 평균값 사이의 차이에 대한 제곱합이다.
- SSE(잔차 변동성): 모델이 예측하지 못한 오차의 변동성. 실제 값과 예측값 사이의 차이에 대한 제곱합이다. 

## 변수 유성 평가

#### t검정
각 독립변수의 회귀계수가 유의한지를 검정하는데 사용한다. <br>
유의성이 크다면 종속변수에 영향을 주는 변수라고 판단하고, p-value가 0.05보다 작다면 해당 변수는 유의하다고 판단한다.
