# 1. 지도 학습과 비지도 학습의 차이

### 지도학습
지도학습이란 컴퓨터를 학습시킬 때 문제와 정답을 같이 제공해서 학습시키는 것이다. <BR>
지도학습의 방식은 모델이 예측한 예측값이 실제 정답과 최대한 같아지도록 학습시키는 것이다. <BR>
지도학습의 예시로는 회귀와 분류가 있다. <br>
ex) 분류: 컴퓨터에게 동물 사진을 제공했을 때, 이것이 고양이인지 강아지인지 분류하는 모델. 
### 비지도학습
비지도학습이란 컴퓨터를 학습시킬 때 정답을 제공하지 않고 학습시키는 것이다. <br>
비지도학습의 방식은 제공된 데이터들에 대해서 패턴이나 유사도를 모델이 학습하도록 하는 것이다. <br>
비지도학습의 예시로는 군집화가 있다.<br>
ex) k-clustering: 여러가지 데이터를 제공했을 때, 유사한 데이터끼리 묶어서 k개의 군집으로 분류한다.

# 2. 회귀와 분류의 차이


### 회귀
회귀는 주어진 데이터를 기반으로 정답을 잘 맞추는 함수를 찾는 문제이다. <br>
ex) 주택 평수, 역까지의 거리, 지역의 평균 소득, 학교와의 거리 등의 데이터를 사용하여 주택 가격을 예측하는 모델. <br>
회귀는 예측하고자 하는 값이 연속형 변수일 때 사용한다. <br>
ex) 사람의 키와 몸무게 정보를 통해 새로운 키가 입력으로 들어왔을 때 그 사람의 적절한 몸무게를 예측한다. 예측하려는 값인 몸무게는 62kg, 62.1 kg, 62.01kg ... 의 연속형 변수이기 때문에 회귀를 사용한다.
### 분류
분류는 데이터가 어떤 레이블에 속하는지의 패턴을 학습하고 새롭게 관측된 데이터에 대해 그 데이터가 속한 레이블을 판별하는 문제이다.<br>
ex) 메일의 제목, 메일의 내용 등의 정보를 사용하여 해당 메일이 스팸 메일인지 정상 메일인지 분류하는 모델. <br>
분류는 예측하고자 하는 값이 범주형 변수일 때 사용한다. <br>
ex) BMI 수치, 흡연 여부 등의 정보를 통해 해당 사람의 당뇨 여부(true/false)를 예측한다. 주어진 데이터로 당뇨가 True일 때와 False일 때의 특징을 학습하고 새로운 test 데이터에 대해 알맞은 범주로 분류하도록 학습한다. <br>
분류는 binary classification과 multiclass classification으로 나뉜다.
- binary classification: 스팸 메일인가? True/False 처럼 어떤 기준에 대해서 참/거짓을 분별한다.
- multiclass classification: 손글씨 이미지와 그것이 어떤 숫자를 나타낸 것인지에 대해 학습한 후, 새로운 손글씨 입력에 대해서 0~9중 어떤 숫자인지를 분별한다.

# 3. 분류 모델 4가지
-> 로지스틱 회귀, decision tree, support vector machine, KNN

### 로지스틱 회귀
로지스틱 회귀는 독립변수의 선형 회귀에 로지스틱 함수를 적용하여 출력값을 0과 1 사이로 변환해준 것이다. <br>
이진분류에 대해서 y=ax+b의 직선으로 분류 문제를 풀려고 하면 제대로 동작하지 않는다. 그 이유는 아래의 이미지와 같이 특정 시점까지는 0이던 데이터들이 어느 시점에는 1로 분류가 되므로 1과 1 분류의 중앙값 근처에서 급격한 이동이 발생한다. 따라서 직선보다는 곡선이 분류에 더 유리하다. 또한 이진분류의 경우 예측값이 0과 1 사이의 값들이지만, y=ax+b의 직선을 사용하면 무한대까지의 값을 가진다. 

 <br>
시그모이드 함수는 아래의 그림과 같이 0과 1사이의 값을 가지는 S자 형태의 곡선이다. 시그모이드 함수를 사용하면, 출력값이 특정 값 이상이면 1, 특정값 이하면 0으로 설정하여 0과 1의 이진분류를 수행할 수 있다. 

<br>
인공지능은 sigmoid함수가 제공된 데이터를 최대한 잘 분류하도록 하는 가중치 w와 편향 b를 구하는 것을 학습 목표로 한다.<br>

로지스틱 회귀를 수학적으로 표현하면 logit을 linear regression한 것이다. logit은 odds에 log를 취한 것이다. odds는 승산으로 "사건 A가 일어날 확률 / 사건 B가 일어날 확률"이다. 즉 해당 이미지가 강아지인 것을 q라고 한다면 해당 이미지가 강아지인 것에 대한 odd는 q/(1-q)로 표현이 된다. logit은 이것에 log를 취한 것이므로 $log\left(\frac{q}{1 - q}\right)$가 된다. 이때 logit을 l이라고 하고 식을 나타내면 $l = \log\left(\frac{q}{1 - q}\right)$이다. 양 변에 $e$를 지수로 취하면 $e^l = \frac{q}{1 - q}$가 된다. 이 식은 $e^{-l} = \frac{1}{q}-1$로 변형이 가능하고, 양 변에 1을 더하면  $1+e^{-l} = \frac{1}{q}$가 된다. 마지막으로 q에 대해 역수를 취하면 $q=\frac{1}{1+e^{-l}}$이 된다. 이때 $\frac{1}{1+e^{-l}}$가 sigmoid함수이고, l은 logit이다. 따라서 로지스틱 회귀는 sigmoid에 logit을 통과시켜 확률이 출력되도록 한 것이다.

### decision tree
decision tree란 각 구간마다 조건에 의해 데이터를 분류하고, 최종적으로 데이터가 순수한 label의 집합으로 구성될때까지 분류를 반복하는 모델이다. <br>
decision tree에서는 CART알고리즘을 사용하는데 데이터셋을 임계값을 기준으로 두개의 child로 나누는 것을 의미한다. <br>
분할 기준으로는 지니 불순도와 엔트로피를 사용한다. 이때 불순도란 분류하려는 데이터 집합에서 서로 다른 클래스가 섞여 있는 정도를 의미한다. <br>
지니 불순도란 특정 node에서 임의로 선택된 샘플이 잘못 분류될 확률을 나타낸다. 지니 불순도(Gini)는 $Gini=1-\sum_{i=1}^{n} P_i^{2}$으로 표현할 수 있다. 이때 $P_i$는 클래스 i의 비율을 의미한다. ex) 특정 feature에 대해 class 1: 50%, class 2: 30%, class 3: 20%로 분류됐다면, 이때의 gini 계수는 $1-(0.5^{2}+0.3^{2}+0.2^{2})=0.62$로 표현된다. <br>
decision tree는 이러한 gini 불순도 값을 최소화하는 방향으로 분할을 진행한다.

### support vector machine
support vector machine은 클래스를 분류할 수 있는 다양한 경계선 중에서 최적의 라인을 찾아내는 알고리즘이다. 즉 margin을 최대하화는 결정 경계를 학습한다.<br>
margin이란 결정경계와 support vector 사이의 거리를 의미한다. <br>
단순히 데이터들을 분류하는 것에 그치지 않고, margin을 최대화하도록 하여 결정경계가 특정 데이터에 더 가깝게 형성되지 않도록 한다.

#### KNN
KNN은 K-Nearest-Neighbor의 약자로 데이터로부터 거리가 가까운 k개의 다른 데이터 레이블을 참조하여 분류하는 알고리즘이다. KNN은 다음과 같은 방식으로 작동한다.
1. 데이터를 준비한다.
2. K값을 설정한다. K는 가장 가까운 이웃의 개수를 의미하며 보통 홀수로 설정한다.
3. 새로운 데이터가 주어지면 새로운 데이터와 다른 데이터들 간의 거리를 유클리드 거리나 맨해튼 거리를 통해 계산한다.
4. 계산된 거리 중에서 가장 가까운 K개의 데이터를 선택한다.
5. K개의 이웃 중에서 가장 많이 등장하는 클래스를 예측 결과로 한다.

장점: 훈련이 필요하지 않고 정보 손실이 없다.<BR>
단점: KNN은 사전학습한 모델을 사용하는 것이 아니라 그때그때 계산하는 방식이므로 쿼리를 처리하는데 시간이 오래 소요되고 데이터 간의 거리를 기반으로 한 모델이기 때문에 이상치에 큰 영향을 받는다.

# 4. 분류 평가 지표

### 혼동 행렬
혼동 행렬이란 분류 모델의 예측 결과를 정확한 예측과 잘못된 예측으로 구분하여 나타낸 표를 의미한다.<br>
예측 성공 여부와 예측값에 대해서 4가지로 구분된다.<br>
- TP(True Positive): 참인 데이터를 참으로 예측
- TN(True Negative): 거짓인 데이터를 거짓으로 예측
- FP(False Positive): 거짓인 데이터를 참으로 잘못 예측
- FN(False Negative): 참인 데이터를 거짓으로 잘못 예측

TP, TN, FP, FN을 사용하여 모델 평가 지표를 도출할 수 있다.
- Accuracy(정확도): 모든 가능한 예측 중 참인 비율
- Precision(정밀도): 참이라고 예측한 경우 실제 참인 비율
- Recall(재현도): 실제로 참인 경우 중 참으로 예측하는 비율

Accuracy: $\frac{TP+TN}{TP+TN+FP+FN}$로 모델이 입력된 데이터에 대해 얼마나 정확하게 예측하는지를 나타낸다. <br>
- 정답 레이블의 비율이 불균형하면 모델의 정확도를 신뢰할 수 없다. (ex) 신용카드 회사에서 고객이 신용카드 연재를 할지 예측하는 모델을 만들 때 실제로 연체하는 고객이 전체의 5%밖에 안된다면, 모델이 그냥 "모든 고객은 연체하지 않는다"라고 예측해도 정확도가 95%나 된다.<br>
Precision: $\frac{TP}{TP+FP}$로 거짓을 참으로 판단한 정도를 알 수 있다. Precision이 높을수록 거짓을 참으로 예측한 정도가 낮다.<br>
Recall: $\frac{TP}{TP+FN}$으로 참을 거짓으로 판단한 정도를 알 수 있다. Recall이 높을수록 참을 거짓으로 예측한 정도가 낮다. <br>

분류할 때, 확률에 기반해서 threshold 값이 넘으면 참, 넘지 않으면 거짓으로 판단한다. threshold를 조정하면 Precision과 Recall을 조정할 수 있다.
- threshold 낮춤: 참으로 예측하는 경우가 늘어나므로 recall이 증가하고 precision이 감소한다.
- threshold 높임: 참으로 예측하는 경우가 줄어드므로 recall이 감소하고 precision이 증가한다.

따라서 recall과 precision 그래프에서 두 그래프가 만나는 지점을 threshold로 정하면 예측 오류를 최소화할 수 있다. 

### F1-score
F1-score는 precision과 recall의 조화평균을 계산한 것이다. <br>
F1-score=$\frac{2}{\frac{1}{Precision}+\frac{1}{Recall}}$=$\frac{2 \times Precision \times Recall}{Precision+Recall}$이다. <br>
F1-score를 사용하면 모델이 한쪽에 치우치지 않고 Precision과 Recall을 동시에 고려한 성능을 평가할 수 있다.

### ROC/AUC Curve
ROC curve: 얼마나 분류가 잘 되었는가를 보여주는 그래프<br>
ROC curve는 x축을 FPR로 하고, y축을 TPR로 하여 분류가 얼마나 잘 되었는지를 나타낸다. 다음 그림에서와 같이 y=x로 기울기가 1일 때가 random classifier로 50%의 성능을 보이는 경우이고, ROC curve 그래프가 좌상단으로 붙을수록 이진분류기의 성능이 좋다고 판단한다. <br>
 

AUC curve: ROC와 x축 사이의 면적(적분값)<br>
AUC curve는 ROC curve 그래프 아래의 면적을 나타낸 것으로 0에서 1 사이의 값을 가지며 1에 가까울수록 분류 성능이 좋다.

### 다중 분류 평가 지표
다중분류의 경우 앞서 다룬 이진분류 평가 지표를 사용해서 클래스별로 점수를 구하고 이것을 적절히 평균내서 사용할 수 있다.<br>
- Macro average: 클래스별로 구한 평가 지표 평균. 모든 클래스에 동등한 가중치를 부여한다.
- Weighted average: 클래스별로 구한 평가 지표 가중 평균. 각 클래스별로 구한 평가 지표를 클래스의 샘플 수로 가중 평균하여 모델 성능을 평가한다. 빈도가 더 높은 클래스에 더 큰 가중치를 부여한다.
- Micro average: 모든 클래스의 예측 결과를 더하여 전체적인 성능을 평가하는 지표. 모든 클래스에 동등한 가중치를 부여한다.

모든 라벨이 유사한 중요도를 가진다면 macro average 값을 참고하고 샘플이 많은 라벨에 중요도를 두고 싶으면 weighted average 값을 참고한다. 라벨 상관없이 전체적인 성능을 평가하고 싶을 때는 micro average 값을 참고한다.

# 5. 하이퍼 파라미터 최적화란?

하이퍼 파라미터는 모델 학습 전에 사용자가 직접 설정하는 변수로, epoch 수, decision tree의 max_depth 등이 있다. <br>
하이퍼 파라미터 최적화란 model의 성능을 향상시키기 위해 적절한 하이퍼 파라미터를 찾는 것이다. 최적화 과정은 다음과 같다.
1. 하이퍼 파라미터 탐색 범위 설정
2. 평가 지표 계산 함수 정의
3. 1단계에서 샘플링한 하이퍼 파라미터 값을 사용하여 검증 데이터로 정확도 평가
4. 위의 단계를 반복하며 정확도 결과를 보고 하이퍼 파라미터 값의 범위를 좁혀나간다.

하이퍼 파라미터 최적화는 grid search, random search, bayesian optimization의 방법을 사용한다. 
- grid search: 정해진 범위에서 하이퍼 파라미터를 모두 순회. 최적해를 정확히 찾을 수 있지만 시간이 너무 오래 걸린다.
- random search: 정해진 범위에서 하이퍼 파라미터를 무작위로 탐색. grid search보다는 속도가 빠르지만 정확도가 떨어진다. 
- bayesian optimization: 사전 정보를 바탕으로 하이퍼 파라미터 값을 확률적으로 추정하며 탐색. 여러 개의 하이퍼 파라미터들에 대해서 Aquisitian Function을 적용했을 때 가장 큰 값이 나올 확률이 높은 지점을 찾아낸다.
