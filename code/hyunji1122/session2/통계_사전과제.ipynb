# 통계_사전과제

## 1. 통계란
: 데이터를 수집, 분석, 해석하는 과정을 통해 패턴, 관계, 변동 등을 이해하고 결론을 도출하는 학문 및 방법론 <BR>
: 기술 통계와 추론 통계로 이루어짐
#### - 기술 통계학
: 측정이나 실험을 통해 수집한 자료로 정보로써 전환하기 위해 요약하고 정리하여 그 특성을 기술하는 절차에 관한 개념    
#### - 추론 통계학
: 전체 집단에서 일부를 뽑아 만든 표본 집단을 통해 전체 집단에 대한 정보를 추측하는 과정

## 2. 기술 통계
#### 2.1 중심 위치 척도
> - 평균(mean)    
>   : 데이터 값의 총합을 데이터의 개수로 나눈 값    
>   : 데이터의 중심 경향성    
>   : 이상치에 민감할 수 있음

> - 중앙값(median)  
>   : 데이터 값들을 크기 순서대로 나열했을 때 가운데 위치한 값    
>   : 이상치에 덜 민감 
>   : 데이터의 중심 경향성을 나타내는 대안적인 지표

> - 최빈값(mode)    
>   : 데이터에서 가장 자주 나타나는 값   
>   : 주로 이산형 데이터에서 사용됨

> - 왜도(Skewness)   
>   : 데이터 분포의 비대칭성을 나타내는 지표   

#### 2.2 변동성 척도
> - 사분위수(quartile)    
>   : 데이터의 표본을 네 개의 동일한 크기로 나눈 값    
>   : Q1 = 하위 25퍼센트에 위치한 값 / Q2 = 중앙값 / Q3 = 상위 25퍼센트

> - 사분위간 범위(IQR)    
>   : Q3에서 Q1값을 제외한 범위    
>   : 이상치 탐지에 유용

> - 범위(range)    
>   : 데이터의 최대값과 최솟값의 차이를 나타내는 지표 -> 데이터 전체 변동 범위 알려줌

> - 분산(variance)    
>   : 데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지를 나타내는 측정 지표    
>   : 편차 제곱의 평균

> - 표준편차(standard deviation)    
>   : 데이터 값들이 평균으로부터 얼마나 퍼져 있는지를 나타냄    
>   : 분산의 제곱근

> - 변동계수(coefficient of variation)    
>   : 데이터의 상대적인 변동성을 나타내는 통계적 지표

#### 2.3 연관성 척도
> - 공분산(covariance)    
>   : 두 변수가 함께 어떤 방향으로 변화하는지와 그 크기 표현    
>   : Cov(X,Y)>0 이면 양의 관계 /  Cov(X,Y)<0 이면 음의 관계 / Cov(X,Y)=0 이면 관계X

> - 상관계수(correlation)    
>   : 두 변수 간의 선형적인 관계를 나타내는 통계적 지표    
>   : 공분산을 각 변수의 표준편차로 나누어 표준화 한 것    
>   : Corr(X,Y)>0 이면 양의 상관관계 / Corr(X,Y)<0 이면 음의 상관관계

## 3. 추론 통계 1 (모집단 분포 추정)
- 모집단과 표본집단 -> 모집단과 모수를 정확하게 알 수 없지만 표본과 표본 통계량을 통해 모집단 추정 가능
- 대수의 법칙 : 표본의 크기가 커질수록 표본 평균이 모평균에 근사
- 중심극한정리 : 표본의 크기가 크면 분포가 정규분포에 가까워짐

#### 3.1 확률분포
: 확률변수가 특정한 값을 가질 확률
: 이산확률분포와 연속확률분포로 구성
> - 이산확률분포
>   : 베르누이 / 이항 / 기하 / 음이항 / 초기하 / 포아송

> - 연속확률분포
>   : 정규분포 / T분포, F분포 / 감마분포 / 지수분포 / 카이제곱 / 베타분포 / 균일분포

#### 3.2 확률분포 종류
-정규분포
> 형태 : 종모양 / 평균을 기준으로 대칭 / 표준편차가 클수록 평평    
> 매개변수 : 평균, 표준편차    
> 특징 : 가장 잘 알려져 있음 / 많은 자연 현상과 사회 현상이 정규분포를 따름

-T분포
> 형태 : 종모양 / 정규분포보다 꼬리 두꺼움    
> 매개변수 : 자유도    
> 특징 : 표본으로 모평균 추정할 때/ 표본 크기가 작을 때 사용

-카이제곱분포
> 형태 : 0에서 시작해서 오른쪽으로 긴 꼬리 / 자유도 증가할 수록 정규분포에 가까워짐      
> 매개변수 : 자유도    
> 특징 : 독립적인 표준 정규 분포를 따르는 변수들의 제곱 합 / 독립성 검정, 적합도 검정에 사용


## 4. 추론통계 2 (가설 검정)

#### 4.1 가설 검정의 절차
: 가설 설정 - 유의 수준 설정 - 검정통계량 산출 - 가설 기각/채택 판단

-가설 설정
> 귀무가설 H0과 대립가설 H1 설정    
> 귀무가설 => 기각하고자하는 명제 / 대립가설 => 주장하고자 하는 명제
> - 1종 오류
>   : 귀무가설이 실제로 참인데 기각할 확률
> - 2종 오류
>   : 귀무가설이 실제로 거짓인데 기각하지 못할 확률

-유의 수준 설정
> 유의수준 : 연구자가 결과를 판단할 때 허용할 수 있는 오류의 최대치    
> ex) 유의수준=0.05 -> "귀무가설이 실제로 참인데도 불구하고 5%의 확률로 귀무가설을 잘못 기각할 수 있다"

-검정통계량 산출
> 검정통계량 = 가설 검정의 결과를 판단하는 데 사용되는 값    
> 데이터에 따라 어떤 통계량을 사용할지 정해야함

-가설 기각/채택 판단
> 유의확률(p-value)    
> : 가설 검정에 쓰이는 또 다른 척도. 검정통계량에서 유도 가능
