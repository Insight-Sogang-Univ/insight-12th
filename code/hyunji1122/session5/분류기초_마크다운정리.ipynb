## **1. 지도 학습과 비지도 학습**

### **지도 학습**
  
  - 문제와 정답을 모두 알려주고 공부시키는 방법 -> 정답이 있는 데이터 사용
  - 예측값을 이미 만들어둔 정답과 같아지도록 기계 학습
  - ex) 회귀, 분류
<BR><BR>
### **비지도 학습**
  
  - 답을 가르쳐주지 않고 공부하는 방법 -> 정답이 없는 데이터 사용
  - 데이터 속의 패턴 또는 각 데이터 간의 유사도를 기계가 학습하도록 하는 것
  - ex) 군집화

=> *지도 학습은 정답이 주어진 상태에서 학습하는 알고리즘, 비지도 학습은 정답이 주어지지 않은 상태에서 학습하는 알고리즘!*


## **2. 회귀와 분류**


### **회귀**    
  : 주어진 데이터(X)를 기반으로 정답(Y)를 잘 맞추는(fit)하는 함수.    
  ex) 주택 가격 예측 -> 종속변수 = 주택 가격, 독립변수 = 주택 평 수, 역까지의 거리, 지역 평균 소득
### **분류**    
  : 기존 데이터가 어떤 레이블에 속하는지 패턴을 알고리즘으로 인지한 뒤에 새롭게 관측된 데이터에 대한 레이블을 판별하는 것    
  ex) 대출상환여부(0/1) 예측 -> 종립변수 = 대출상환여부(0/1), 독립변수 = 고객의 특성, 대출의 특성


=> *예측하고자 하는 값(종속 변수)이 연속형이면 회귀, 범주형이면 분류 사용*


## **3. 분류 모델의 대표적인 종류**

### 1) **로지스틱 회귀**    
  - 독립 변수의 선형 조합에 로지스틱 함수(시그모이드 함수)를 적용하여 출력값을 0에서 1사이로 변환해주는 것
  - 이진 분류 문제를 해결하는 데 적합
  - **시그모이드 함수**
    > 시그모이드 함수 : 출력이 0과 1사이의 값을 가지면서 S자 형태로 그려지는 함수
    > 
    > 입력값이 커지면 1에 수렴, 입력값이 작아지면 0에 수렴    
    > 출력값이 0부 1까지의 값을 가짐    
    > -> *출력값이 특정값 이상이면 1, 특정값 이하면 0으로 정하면 이진분류 문제를 풀기 위해 사용 가능*
<BR><BR>

### 2) **결정 나무**
   - 조건에 따라 데이터를 분류하며, 최종적으로 데이터가 순수한 label의 집합으로 구성될 때까지 분류를 반복하는 모델
    
     > - **Root Node**: Decision Tree의 시작이 되는 노드
     > - **Edge**: 노드와 노드를 연결하는 길목
     > - **Leaf Nodes**: Tree의 가장 마지막 노드로, 모델에서 label에 해당
     > - **Height(depth)**: Tree의 깊이로, 클수록 tree의 구조는 복잡해짐
     > - **Level**: 노드의 절대적 위계, Root node의 level = 0, leaf node의 level = height - 1
     > - **Parent**: 상대적으로 높은 위계의 노드
     > - **Child**: 상대적으로 낮은 위계의 노드

   - 실제 학습 시 고려해야 할 것들: **모수 설정, 차이점 시각화, Prunning**
     - **parameter 설정**   
       : `min_samples_split`, `min_samples_leaf`, `min_weight_fraction_leaf`, `max_leaf_nodes`, `max_features`
     - **시각화**    
       : 분류가 잘 이루어졌는지 확인
     - **Prunning(가지치기)**    
       : 불필요한 노드 지우기
       
       -> 노드가 너무 많아지면 과적합이 될 확률이 높으므로 prunning을 통해 하부 트리를 제거하여 일반화 성능을 높일 수 있음

### 3) **서포트 벡터 머신(SVM)**
- 클래스를 분류할 수 있는 다양한 경계선 중 최적의 라인을 찾아내는 알고리즘
 
- 명확하게 분류할 수 있는 데이터 집단에 뛰어난 성능을 보이며 고차원 공간에서도 효과적으로 사용 가능

### 4) **KNN**
- 데이터로부터 거리가 가까운 k개의 다른 데이터 레이블을 참조하여 분류하는 알고리즘
  
- 계산 순서:
  - **데이터 준비**: 각 데이터는 특징 벡터와 레이블로 이루어짐
  - **K값 설정**: K는 가장 가까운 이웃의 개수를 나타내며 보통 홀수개로 설정
  - **거리 계산**: 새로운 데이터(예측하려는 데이터)가 주어지면 이 값과 기존 모든 데이터 간의 거리를 계산
  - **가장 가까운 K개의 이웃 선택**: 계산된 거리 중에서 가장 작은 거리 값을 가진 K개의 데이터를 선택
  - **분류하기**: K개의 이웃 중 가장 많이 등장하는 클래스가 예측 결과가 됨 
- 장점 : 훈련 필요X, 정보 손실X
- 단점 : 쿼리를 처리하는 데에 시간이 오래 소요, 이상치에 큰 영향을 받음




    


## **4. 분류 평가 지표**

### 1) **혼동 행렬**
: 분류 모델의 예측 결과를 정확한 예측과 잘못된 예측으로 구분하여 나타낸 표(행렬)

->  TP, TN = 예측을 올바르게 한 경우 / FN, FP = 예측을 잘못한 경우

### 2) **F1-Score**
: 정밀도와 재현율의 조화 평균

*조화평균 사용하는 이유-> 정밀도와 재현율 간의 균형을 효과적으로 평가하기 위해!*

### 3) **ROC/AUC Curve**
#### **ROC Curve**
: 얼마나 분류가 잘 되었는가를 보여주는 그래프

- TPR : 참인 것들 중에 참이라고 예측한 비율
- FPR : 거짓인 것들 중에 참이라고 잘못 예측한 비율

#### **AUC Curve**
: ROC와 x축 사이의 면적(적분값)

-> 모델의 성능을 숫자로 나타낼 수 있음. 0에서 1사이의 값을 가지며, 1에 가까울수록 분류 성능이 좋은 것에 해당!



## **5. 하이퍼파라미터 최적화**
#### **하이퍼파라미터**
: 모델 학습 과정에 반영되는 값으로, 학습 시작 전에 사용자가 직접 설정하는 변수    
ex) decision tree의 `max_depth`, random forest의 `n_estimators`
#### **하이퍼파라미터 최적화**
: Tuning을 거쳐 적절한 하이퍼파라미터를 찾음으로써 Model 성능을 향상시키는 것
##### **[최적화 과정]**
1) 하이퍼파라미터 탐색 범위 설정
2) 평가 지표 계산 함수 정의
3) 1단계에서 샘플링한 하이퍼파라미터 값을 사용하여 검증 데이터로 정확도 평가
4) 위 단계를 특정 횟수 반복하며 정확도 결과를 보고 하이퍼파라미터의 범위를 좁힘
