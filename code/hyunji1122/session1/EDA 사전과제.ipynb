## 0. 데이터 분석 프로세스
: 데이터 수집 -> 데이터 탐색 -> 데이터 전처리 -> 데이터 모델링

> 데이터 수집
> : 다양한 형식으의 데이터 수집

> 데이터 탐색(EDA)
> : 데이터를 여러 가지 방식으로 파악

> 데이터 전처리
> : 데이터 분석을 위해 데이터를 적절한 방식으로 손질(결측치, 이상치 처리)

> 데이터 모델링
> : 정보 추출을 위해 모델 구축

## 1. EDA의 개념
: EDA = 데이터를 여러 가지 방식으로 파악하는 모든 과정
> <EDA 진행 시 고려해야할 점>
> - EDA 대상 : 일변량 vs 다변량
> - EDA 종류 : 시각화 vs 비시각화(수치적으로 파악)
> - EDA 유형 
>   * 일변량 비시각화
>   * 일변량 시각화
>   * 다변량 비시각화
>   * 다변량 시각화

## 2. 데이터 확인
- 데이터 읽기/쓰기
> (1)절대경로와 상대경로    
>> 절대경로 : 처음부터 시작하여 목적지까지 전체적인 경로 ex)C:\Users\사용자 이름\OneDrive\Desktop    
>> 상대경로 : 현재 작업하고 있는 디렉터리 기준으로 상대적으로 표현되는 경로        
>      
> (2)데이터 입출력    
>> 데이터 파일의 형식에 따라 데이터를 읽고 쓰는 함수가 다름    
>    
> (3) CSV와 EXCEL
>> CSV 파일 : 쉼표로 구분하고 있는 텍스트 파일. 가장 널리 사용됨
>> EXCEL 파일 : 행과 열이 데이터프레임의 행, 열로 일대일 대응
>>  -> 두 형식 모두 pandas를 통해 DataFrame 형식으로 읽어올 수 있음
- 데이터셋 파악하기
> (1) 데이터 프레임 보기
>> head(n) : 데이터 프레임 상위 n개 데이터 확인
>
> (2) 데이터 변수 확인    
>> 변수 = 데이터 프레임의 column = feature    
>> 변수 유형: 질적.범주형/양적.수치형으로 구분(-> 트렌드가 보이면 연속형, 트렌드 안보이면 이산형)
> 
> (3) 데이터 분포 확인
>> : 단변수 분석일 경우 원시 데이터 변수들의 그래프를 통해 분석 가능

## 3. 전처리
- 결측값 처리
> (1) 결측값 확인
> > info(), value_counts(dropna=False), isnull(), notnull(), replace()
> 
> (2) 결측값 처리
> > 삭제: dropna() --> 행삭제/열삭제//전체삭제/부분삭제    
> > 대체: fillna() --> 일괄 대체 or 유사 유형 대체
- 이상치 처리
> (1) 이상치 확인
>> - 통계를 통해 확인: describe()    
>> - 시각화를 통해 확인: BoxPlot
>> - Z-score을 통해 확인: Z값이 음수면 평균 이하, 양수면 평균 이상    
>> - Tukey Fences을 통해 확인: 사분위 범위를 기반으로 두가지 경우에 이상치라고 판단->(Q1-(1.5*IQR)미만, Q3+(1.5*IQR)초과
>
> (2) 이상치 제거
>> - 전체 삭제 : 이상값이 Human error에 의해 발생한 경우
>> - 다른 값으로 대체 : 절대적인 관측치의 숫자가 작은 경우
>> - 변수화 : 이상값이 자연 발생한 경우
>> - 리샘플링 : 이상값을 분리해서 모델 제작
- 피처 엔지니어링(변수 가공)    
 : 해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 피처들의 형태를 변형하거나 적절하게 처리하는 과정
> (1) 레이블인딩, 원핫인코딩 : 텍스트로 주어지는 값을 숫자로 바꾸는 작업
>> Lable Encoding: 범주형 변수를 0부터 N-1까지의 숫자로 변환    
>> One-Hot Encoding: 범주형 변수를 이진 벡터로 변환
> 
> (2) 구간화: 연속적인 값을 일정한 구간으로 구분
> > ```python
> > pd. cut(데이터프레임['컬럼명'], bins=[나누는 기준 리스트], labels=[지정할 label])
> > ```
>
> (3) 변환: 기존의 피처를 변환해 새로운 피처 생성
>> 정해진 원칙이 있는 방법이 아니어서 분석가의 데이터 특성에 대한 이해도에 따라 다양한 데이터 생성 가능
>
> (4) 스케일링: 서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업    
>> - StandardScaler(표준화) : 각 feature의 평균을 0, 분산을 1로 변경--> 모든 피처들이 같은 스케일 갖게됨    
>> - MinMaxScaler : 모든 feature가 0과 1사이에 위치하게 만드는 것. 데이터가 서로 다른 비율의 속성으로 구성되어 있을 때, 같은 비율로 속성을 맞춤

## 4. 시각화
: 숫자와 텍스트로 표현되는 정보를 그래프, 차트, 그림, 도표 등의 시각적 요소로 변환    
-> 데이터 패턴, 관계, 추세 등 쉽게 파악 가능    
- 파라미터

> : 프로그래밍된 함수의 입력값

- 다양한 그래프

> boxplot
> : 사분위수와 이상치 보여줌
   
> countplot, histplot    
> : countplot -> 범주형 변수의 빈도수를 확인하는 그래프    
> : histplot -> 수치형 변수의 구간별 빈도수를 보여주는 그래프    

> distplot, kdeplot    
> : 히스토그램을 연속적으로 곡선으로 연결한 그래프    

> barplot, pointplot
> : barplot -> 범주형 데이터 값 x에 따른 수치형 데이터 값 y의 평균 값 제공
> : pointplot -> 막대 그래프와 모양만 다르고 동일한 정보 제공

> scatterplot, regplot
> : scatterplot -> 두 변수 간의 관계 시각화
> : regplot -> 두개의 연속 변수 사이의 산점도를 그리고 회귀선을 함께 나타냄

> catplot
> : 수치형 데이터와 범주형 데이터의 관계를 볼 때 사용

> pieplot
> : 데이터의 부분과 전체 간의 비율을 표현하는 그래프. 주로 비율 강조를 위해 사용됨

> heatmap
> : 변수간 상관계수를 직관적으로 볼 수 있는 그래프

> violinplot
> : 박스 플롯과 커널밀도추정 함수 그래프를 합쳐놓은 그래프

> pairplot
> : 여러 그래프 한번에 설정
