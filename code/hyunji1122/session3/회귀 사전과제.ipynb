# 회귀 사전과제

## 머신러닝

- 비지도학습: 정답이 없는 데이터로 학습하는 방식
- 지도학습: 정답이 있는 데이터로 학습하는 방식
- 머신러닝: 새로운 데이터를 예측하거나 결정을 내릴 수 있도록 하는 기술

## 모델링

- 훈련데이터: 모델을 학습시킬 때 쓰이는 데이터
- 테스트 데이터: 학습된 모델을 검증하기 위해 사용하는 데이터

## 회귀
-> 데이터에서 패턴을 찾아내어 미래 값을 예측하는 것    
-> 선형회귀, 비선형회귀, 로지스틱회귀, 릿지회귀, 라쏘회귀, 다항회귀 등 다양한 기법이 존재

### 선형회귀

: 데이터를 가장 잘 설명하는 회귀선을 찾아 예측하는 과정    
: 예측하고자 하는 것->종속변수, 예측을 위해 사용하는 것->독립변수


#### 단순선형회귀분석

- 하나의 독립변수만 존재하는 선형 회귀
- 데이터를 가장 잘 설명하는 직선 y=ax+b을 찾는 과정에 해당
> <최소 제곱법, 최소자승법>
> - 잔차의 제곱의 합이 최소가 되는 지점으로 최적의 회귀선을 구하는 방식
> - 회귀선과 실제 데이터와의 차이를 최소화하기 위한 시도

<BR>

#### 다중선형회귀분석

-> 독립 변수가 2개 이상인 경우의 선형회귀

##### 다중공산성
- 회귀 분석에서 독립 변수들 간에 상관관계가 큰 경우 발생
- 다중공산성이 높을 경우 어떤 독립 변수가 종속 변수에 얼마나 영향을 미치는지 정확하게 구분하기 어렵기 때문에 중요

##### 다중공산성 확인 방법
- 상관계수    
  : 상관관계의 정도를 수치로 나타낸 것       
  : -1과 1사이에 존재하며 양수면 양의 상관관계, 음수면 음의 상관관계, 0은 선형상관관계 없음을 뜻함    
  : 히트맵 혹은 corr()함수를 통해 확인 가능
- VIF지수
  : 분산 팽창 인수
  : VIF가 높으면 다중공산성이 존재한다고 판단

##### 다중공산성 대처 방법
- 변수 제거 : 독립변수로서 사용할 변수를 선택하는 방법
- 변수 변환 : 변수들을 더하거나 빼서 새로운 변수 생성
- 규제 선형 모델 활용 : 릿지, 라쏘, 엘라스틱넷 등의 방법을 통해 복잡도를 줄이는 방법
- PCA(주성분분석): 데이터의 차원을 축소하는 데 사용되는 통계적 기법

#### 모델 평가방법
- 모델 성능 및 유의성 평가

> 평균 제곱 오차(MSE)
> 
> :회귀 분석에서 모델의 예측값과 실제 관측값 사이의 오차 제곱의 평균을 의미
> : 모델의 예측 성능을 평가하는 지표 중 하나

> 평균 절대 오차(MAE)
> 
> : 회귀 분석에서 모델의 예측값과 실제 관측값 사이의 절대값 오차의 평균을 의미
> : 모델의 예측 성능을 평가하는 지표 중 하나

- 변수 유성 평가

> t검정    
> : 각 독립 변수의 회귀계수가 유의한지를 검정하는데 사용됨    
> : 유의성이 크다면 종속 변수에 영향을 주는 변수라고 판단
