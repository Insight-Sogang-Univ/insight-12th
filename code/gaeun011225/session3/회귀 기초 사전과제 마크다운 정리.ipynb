## <font color="green">회귀란?
* 데이터에서 패턴을 찾아내어 미래 값을 예측하는 것
* 선형회귀, 비선형회귀, 로지스틱회귀, 릿지회귀 등 다양한 회귀기법 존재

#### <font color="green"><선형회귀>
* 데이터를 가장 잘 설명하는 회귀선을 찾아 예측하는 과정
* 종속변수(y) : 예측하고자 하는 것 / 독립변수(x) : 예측을 위해 사용하는 변수
* <u>**단순 선형 회귀**</u> : 데이터를 직선 형태로 표현하는 경우 / <u>**다중 선형 회귀**</u> : 데이터를 두 개 이상의 독립변수를 사용해 표현하는 경우

#### <font color="green"><단순선형회귀분석>
하나의 독립변수만 존재하는 선형 회귀로 데이터를 가장 잘 설명하는 y=aX+b을 찾는 과정임.
* 최적의 회귀선을 찾는 방식) <u>**최소 제곱법, 최소 자승법**</u>
  : 잔차의 제곱의 합이 최소가 되는 지점으로 최적의 회귀선을 구하는 방식

#### <font color="green"><다중선형회귀분석>
독립변수 x_i가 2개인 경우 3차원 공간에 표현이 되고 회귀선은 평면이 되는 선형 회귀
+ 다중선형회귀에서 중요사항) <u>**다중공산성**</u>
  : 회귀 분석에서 독립 변수들 간에 상관관계가 큰 경우 발생, 다중공산성이 높은 경우 어떤 독립변수가 종속변수에 얼마나 영향을 미치는지 구분하기가 어려움
  
1. <u>**다중공산성 확인 방법**</u>
  : 상관계수 / VIF지수(분산 팽창 인수)

2. <u>**다중공산성 대처 방법**</u>
* 변수 제거 : 독립변수로서 사용할 변수 선택
* 변수 변환 : 변수들을 더하거나 빼서 새로운 변수 생성
* 규제 선형 모델 활용 : 릿지, 라쏘, 엘라스틱넷 등의 방법을 통해 모델의 복잡도를 줄임
* PCA(주성분분석) : 고차원 데이터에서 중요한 정보를 최대한 보존하면서 복잡성을 줄임(데이터 차원의 축소)
  

#### <font color="green"><모델 평가방법>
+ <u>성능 평가 지표</u>
  1. 평균 제곱 오차(MSE)
     : 모델의 예측값과 실제 관측값 사이의 오차 제곱의 평균
  2. 평균 절대 오차(MAE)
     : 모델의 예측값과 실제 관측값 사이의 절대값 오차의 평균
     
+ <u>변수 유성 평가</u>
  1. t검정
     : 각 독립변수의 회귀계수가 유의한지를 검정하는데 사용 / 유의성이 크면 종속변수에 영향을 주는 변수라고 판단 / 보통 p-value가 0.05보다 작으면 해당 변수는 유의하다고 판단
