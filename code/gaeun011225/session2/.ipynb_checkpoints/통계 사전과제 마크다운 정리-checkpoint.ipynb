## 기술통계와 추론통계

### 기술 통계학
+ 측정이나 실험을 통해 수집한 자료를 정보로써 전환하기 위해 요약해 특성을 기술하는 절차에 관한 개념
+ 주로 평균, 중앙값, 표준편차 등의 개념

#### (1) 중심 위치 척도
+ 평균 : 데이터 값의 총합을 데이터의 개수로 나눈 값으로 데이터의 *중심경향성*을 나타냄
+ 중앙값 : 데이터 값들을 크기 순서대로 나열할 때 가운데 위치한 값
+ 최빈값 : 데이터에서 가장 자주 나타나는 값 / 주로 이산형 데이터에서 사용됨

#### (2) 변동성 척도
+ 사분위수 : 데이터의 표본을 네 개의 동일한 크기로 나눈 값
  + Q1) 하위 25퍼센트에 위치한 값
  + Q2) 하위 50퍼센트에 위치한 값
  + Q3) 하위 75퍼센트에 위치한 값 = 상위 25퍼센트에 위치한 값
+ 사분위간 범위(IQR) : Q3에서 Q1 값을 제외한 범위로 데이터의 중앙 50%가 퍼져있는 범위
+ 범위 : 데이터의 최대값 - 데이터의 최소값
+ 분산 : 데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지를 나타내는 지표 / = 편차 제곱의 평균
+ 표준편차 : 분산과 마찬가지로 데이터 값들이 평균으로부터 얼마나 퍼져 있는지를 나타내는 지표 / = 분산의 제곱근 / 표준편차를 사용해야 데이터의 단위를 일치시킬 수 있음
+ 변동계수 : 데이터의 상대적인 변동성을 나타내는 지표 / = 표준편차를 데이터의 평균으로 나눈 값

#### (3) 연관성 척도
+ 공분산 : 두 변수가 함께 어떤 방향으로 변화하는지와 그 크기를 나타내는 척도
  + Cov(X,Y)>0 : X와 Y는 양의 관계
  + Cov(X,Y)<0 : X와 Y는 음의 관계
  + Cov(X,Y)=0 : X와 Y는 양도 음도 아닌 관계 / 선형관계는 없으나 비선형적인 관계가 존재할 수 있음
+ 상관계수 : 두 변수 간의 선형적인 관계를 나타내는 통계지표 / -1~1 사이의 값을 지님
  + Corr(X,Y)>0 : 양의 상관관계
  + Corr(X,Y)<0 : 음의 상관관계



### 추론 통계학 1(모집단 분포 추정)
#### (1) 확률분포
+ 확률분포 : 확률변수가 특정한 값을 가질 확률을 나타내는 함수
+ 확률변수 : 어떤 사건이나 실험의 결과를 숫자로 대응시키는 함수
+ 이산확률분포와 연속확률분포 존재

#### (2) 확률분포의 종류
+ 정규분포
  : 종 모양 / 평균을 기준으로 대칭 / 표준편차가 클수록 곡선이 평평해짐
+ T분포
  : 종 모양 / 평균을 기준으로 대칭 / 자유도가 커짐에 따라 표준정규분포 N(0,1)에 수렴 / 표본으로 모평균을 추정할 떄 사용
  + 자유도 : 얼마나 많은 정보가 독립적으로 변할 수 있는가를 나타내는 숫자
+ 카이제곱분포
  : 자유도가 증가함에 따라 중심이 오른쪽으로 이동하고 정규분포에 가까워짐 / 주로 독립성, 적합도 검정에 사용됨


### 추론 통계학 2(가설 검정)
#### (1) 가설 검정의 절차
: 가설 설정 -> 유의수준 설정 -> 검정통계량 산출 -> 가설 기각/채택 판단
1. 가설설정
   귀무가설 H0과 대립가설 H1을 설정
   + 귀무가설 : 주로 기각하고자 하는 명제로 설정("~는 차이가 없다") / 현재 통용되고 있는 가설
   + 대립가설 : 주로 내가 주장하고자 하는 명제로 설정("~는 차이가 존재한다") / 새롭게 주장하고 싶은 가설
   + 1종 오류 : 귀무가설이 실제로 참이었는데 기각할 확률
   + 2종 오류 : 귀무가설이 실제로 거짓인데 기각하지 못할 확률
     
2. 유의수준 설정
   + 유의수준 : 정해진 수치를 얼마나 벗어났을 때 귀무가설을 기각할 것인가를 판단하는 기준
     
3. 검정통계량 산출 : 검정통계량은 가설 검정의 결과를 판단하는 데 사용되는 값
   + 데이터의 분포나 특징에 따라 T값,F값,Z값 등 어떤 통계량을 사용할 지 정해야함
   
4. 가설 기각/채택 판단 : 유의확률(p)가 유의수준(a)보다 작을 시에 귀무가설을 기각함

#### (2) 여러가지 검정 방법
1. 수치형-수치형) (피어슨)상관분석
2. 범주형-수치형) F-test(ANOVA), T-test
3. 범주형-범주형) Chi2-test
