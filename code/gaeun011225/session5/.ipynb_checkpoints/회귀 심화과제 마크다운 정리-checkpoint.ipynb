## 회귀 심화 클론코딩을 통해 배운 점

+  pandas 데이터프레임 df의 기술 통계를 확장해, 각 열에 대해 중간값, 결측치 수, 왜도, 첨도를 추가로 계산하여 데이터프레임에 저장하는 방법을 알 수 있었습니다. 이렇게 추가로 원하는 통계정보를 얻을 수 있다는 점이 흥미로웠습니다.
+  종속변수인 price의 왜도가 크기 때문에 자연로그를 취해 분포를 정규분포에 더 가깝게 만드는 이유가 궁금해서 찾아보니 회귀 분석에서 중요한 가정 중 하나인 정규성 가정을 충족시키기 위함이고, 선형관계를 확보해 회귀모델이 더 잘 작동하도록 위함을 알 수 있었습니다.
+  date2(집이 매각된 연도)와 yr_built(집이 지어진 연도)의 기간 차이를 비교해서 sold-built_years 라는 새로운 변수를 추가하는 과정을 보면서 이런 식으로 유용한 변수를 뽑아내는 거구나 하고 알 수 있었습니다.
+  heatmap을 통해 변수 간 상관관계를 확인하는 과정에서 데이터프레임에 문자열이 존재해서 상관계수를 구할 수 없다는 오류를 접했습니다. 데이터프레임을 살펴보니 date 칼럼이 문자열로 남아있어서 해당 오류를 접했음을 인지하고 date칼럼을 삭제한 후 상관계수를 구했습니다.
+  상관계수가 높은 독립변수들을 뽑아내서 데이터 분포를 확인하고, 다중공산성 발생 가능성을 고려해 상관계수가 가장 높은 하나의 변수를 뽑아내는 과정을 경험했습니다. 그 과정 속에서 다중공산성을 미리 방지하는 과정의 중요성을 다시 한번 알 수 있었습니다.
+  단순 선형회귀 분석을 해본 다음에, 여러 독립변수를 사용해 다중 선형회귀 분석을 진행했습니다. 회귀분석 하는 과정을 자주 반복하다보니 독립/종속변수를 선택하고 회귀모형을 돌리고 결과를 확인하는 과정에 조금은 익숙해졌음을 느꼈습니다. 회귀분석 결과를 해석하는 방법은 조금 어려워서 더 반복하며 연습해야겠습니다!
+  다중 선형 회귀분석에서 모든 변수의 경우의 다중공산성을 확인했기에, 독립변수 일부를 뽑아내 다중 선형회귀 분석을 진행하니 다중공산성이 줄어든 것을 확인할 수 있었습니다. 이러한 과정에서 회귀분석 과정에서 늘 다중공산성의 발생 가능성을 우선시하여 생각해야함을 알 수 있었습니다.
+  또한, 저번 세션때 배운 선형 회귀모델의 기본 가정들을 확인했습니다. 정규성 가정을 확인하기 위해 Q-Q 플롯, 잔차 패턴, 샤피로-윌크 검정을 사용했고 독립성을 확인하기 위해서는 잔차 그래프와 ACF, 등분산성을 확인하기 위해서는 잔차 그래프를 사용했습니다. 이런 가정들을 확인하기 위해 다양한 검증방법이 존재함을 알 수 있었습니다.
