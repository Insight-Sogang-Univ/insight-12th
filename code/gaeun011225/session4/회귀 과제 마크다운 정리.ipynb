## 회귀 과제 마크다운 정리

+ 제대로 된 데이터 분석과정을 거치기 전에 결측값의 존재를 확인하는 부분을 계속 인지하지 못했는데 이번 과제를 통해 다시 한번 결측값을 찾는 과정을 경험할 수 있었습니다. 이번 데이터에서는 모든 열과 행에 결측값이 존재하지 않았지만 앞으로 결측값이 존재하는 다양한 데이터를 접하게 될테니 처리하는 방법 등을 잘 익혀놔야 겠다고 생각했습니다.
+ 네 가지 변수('mnth', 'weekday', 'season', 'weathersit')에 대해 더미 변수를 만들기 전에, 이 변수가 수치형 데이터처럼 보이지만 실제로는 범주형 데이터라는 점을 인지하고 범주형 데이터로 변환하는 과정이 흥미로웠습니다. 범주형 데이터 타입으로 변환해야만 pandas의 get_dummies() 같은 함수를 통해 각 범주에 대해 적절한 더미 변수를 생성할 수 있기 때문에 범주형 데이터로 바꿔주고 의미없는 대소관계가 생기지 않도록 하는 과정이 중요함을 다시 한번 인지할 수 있었습니다.
+ 수치형 변수와 범주형 변수를 각각 구분해 시각화하는 부분도 재밌었습니다. 확실히 시각화하는 방법을 사용하니 어떤 변수간에 연관성이 있는지 한 눈에 확인할 수 있다는 점이 큰 장점으로 느껴졌습니다.
+ 해당 과제를 하면서 RFE모델을 처음 접해보았다습니다. 궁금해서 해당 모델에 대해 더 찾아보니 RFE는 모델의 성능을 유지하거나 향상시키면서 중요한 특성(피처)만 남기는 것이 목적이고 불필요하거나 중요하지 않은 특성들을 제거해 모델의 복잡도를 줄여 더 해석하기 쉬운 모델을 만든다는 점이 큰 장점이라고 느껴졌습니다. 실제로 코드를 수행하니 선택된 피쳐와 선택되지 못한 피쳐가 구별되는 것이 신기했습니다.
+ 이번 과제에서도 VIF값을 확인해 각 변수의 다중공산성을 확인했습니다. 값이 높은 변수부터 제거하는 과정에서 VIF가 가장 높았던 변수인 temp가 아닌 변수 hum을 제거하는 것이 재밌었습니다. 온도가 자전거 대여사업에 중요요소인 것을 고려하는 것처럼 단순히 VIF값만 보는 것이 아니라 다양한 정보를 고려해 변수를 제거해야 하는 것임을 알 수 있었습니다.
+ 이후에 p-value값을 통해서도 변수를 제거하는 과정을 거치니 다양한 변수들 사이에서 종속변수에 유의미한 영향을 끼치는 독립변수를 파악하는 과정이 조금 어렵게 느껴졌습니다. 하지만, 앞으로 더 많은 데이터셋을 다루면서 제 능력치를 키워야겠다고 생각할 수 있었습니다!
+ 회귀 교육세션때 test 데이터를 사용해 값을 예측하는 과정을 잘 따라가지 못했었는데 이번 과제를 통해 다시 한번 더 이해할 수 있었습니다. 마지막 부분에서 결정계수 값을 확인해 해당 모델이 데이터 세트를 잘 설명할 수 있다는 점을 알 수 있었습니다.
+ 이번 과제를 통해 데이터 전처리와 분석 후 최종모델을 설정하고 평가하기 위해서는 어떠한 과정을 거쳐야 하는지 알 수 있었습니다. 생각보다 더 길고 난이도가 높은 과정에 놀랐지만 처음부터 쉬운 건 없다는 것을 다시 한번 느꼈습니다. 앞으로 이 과정을 잘 익혀나가면서 더 많은 데이터를 다뤄보고 싶다는 생각이 들었습니다.
