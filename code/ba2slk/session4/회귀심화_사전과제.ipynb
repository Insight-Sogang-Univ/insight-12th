# 복습
## 회귀란?
- 연속적인 데이터에서 패턴을 찾아내는 통계적 기법.
- 선형회귀, 비선형 회귀, 다항회귀, ...
- 데이터 요약, 예측, 시계열 모델링 및 변수 간 인과관계 발견
- R-Squre, Adjusted R-squre, AIC, BC(SC)

### 단순선형회귀분석 vs 다중선형회귀분석
=> 독립 변수(설명 변수) 개수에 차이가 있을 뿐 본질은 같다.

### 그럼 최적의 회귀선을 찾는 방법은? 최소제곱법, 최소자승법
=> 잔차(회귀선-관측값)제곱의 합이 최소가 되는 회귀선 찾기

# **선형 회귀**
## 다중선형회귀: 수식적 이해
선형회귀식 => 종속변수 y(우리가 알고 싶은 대상)에 대한 설명 변수(feature)의 **가중 평균**<br>
즉, 설명변수에 가중치를 곱해 더함으로써 y를 얼마나 잘 설명하는 지 나타냄.

## 다중선형회귀: 기본 가정
: 분석의 신뢰성과 가능성을 보장하기 위해서 반드시 충족 여부를 판단해야 함.
### 1. 선형성
### 2. 오차항(white noise)의 평균은 0이다.
### 3. 독립성
: 각각의 설명 변수가 서로 선형독립적이어야 함. (즉, 다중공선성이 없어야 함.)
### 4. 등분산성
: 오차항의 분산이 일정해야 한다.
- 그렇지 않은 경우는 이분산성(heterosceasticity)가 있다고 함. 횡단면 자료에서 많이 나타남.
- 잔차의 도표화, 검정(Brown-Forsythe, Breusch-Pagan) -> 등분산성 확인
### 5. 오차항은 자기상관 되어 있지 않다.
: 오차항의 공분산은 항상 0이어야 한다.
- 그렇지 않은 경우 자기상관이 있다고 함. 시계열 자료에서 많이 나타남.
- 더빈-왓슨 검정, ACF, PACF 함수 찍기
### 6. 정규성
- 오차항이 정규분포를 따른다.
- 위배되어도 다중선형회귀분석 결과에 큰 영향을 주지는 않음.
- 샤피로-윌크 검정, 자퀴-베라 검정, Q-Q plot 찍기

## 7. 회귀 분석 평가 방법
=> 회귀선을 데이터와 함께 **시각화** 하거나 **통계지표** 활용

### 시각화
: 직관적으로 두 회귀선 중 어떤 것이 해당 데이터를 잘 요약하는 지 비교하기는 어려움 -> 통계지표까지 활용

### 통계 지표
1. 결정 계수 (R-squre, R^2)<br>
: 설명된 비율/총 변동(평균과의 차이)<br>
- 모델이 데이터를 얼마나 잘 설명하는지 측정하는 지표. 0과 1사이 값임.
- 1에 가까울수록 데이터를 잘 설명함.

2. 조건된 결정 계수(Adjusted R^2)
: R^2는 데이터와 큰 관련성이 없는 변수를 추가해도 증가함. => 따라서 변수의 개수 증가에 덜 민감하도록 조정된 지표임.<br>
- R*2 식에 (n-1)/(n-k)를 곱함. 변수가 K개 증가하면 전체 값은 감소함.

3. AIC, BIC(SC)
- 정보기준. 값이 **낮을수록** 좋음
- AIC: BIC에 비해 복잡성에 대한 패널티가 비교적 작음.
- BIC: AIC보다 더 엄격한 기준으로 데이터의 양에 따라 강한 패널티 부여

# **비선형 회귀**
## 다항식 회귀모델
: 선형 모델의 feature를 다항식으로 만들어서 선형 회귀 사용<br>
- 장점: 선형 모델에 비해 회귀선을 잘 fit하도록 그릴 수 있음.
- 단점: 너무 많은 feature를 이용하면 과적합 문제 발생


## 지수/로그 회귀모델
: 기존 식에 지수/로그를 사용해 변형시킨 후 선형 회귀 사용.
