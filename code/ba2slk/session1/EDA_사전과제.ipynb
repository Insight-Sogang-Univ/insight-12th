## **1. EDA(Exploratory Data Analysis, 탐색적 데이터 분석)란?**
데이터 분석(모델링)을 위해 데이터를 여러 가지 방식으로 파악하는 모든 과정. 뜯어보기!
- 시각화 -> 패턴
- 데이터 특이성 확인
- 통계 & 그래프로 가설 검정

즉, **주어진 데이터에 대해 알아보는 것**. EDA를 통해 얻은 데이터에 관한 이해, 판단이 향후 분석에 큰 영향을 주기 떄문에 매우 중요한 과정임.

### 미리 고려할 방향성?
>**EDA 대상(일변량/다변량)**
>>1. univariate(일변량)
: EDA를 통해 한 번에 파악하려는 변수가 1개. 데이터 설명 및 패턴 찾는 것이 목표.
>>2. Multi-variate(다변량)
: EDA를 통해 한 번에 파악하려는 변수가 여러 개. 변수 간 관계 찾기. 여러 변수 간의 관계를 파악하기 전에 개별적으로 데이터를 파악하는 것이 오류 대비 면에서 좋음.

>**EDA 종류(시각화/비시각화)**
>> 1. Graphic(시각화) -> 차트, 그림 (데이터 형태를 한 눈에 파악)
>> 2. NonGraphic(비시각화) -> Summary Statistics (정확한 값에 초점)

>**EDA유형**
: 위 4가지를 조합하면 됨.
>> 일변량 비시각화: 주어진 데이터의 distribution  확인<br>
>> 일변량 시각화: 주어진 데이터를 전체적으로 파악<br>
>> 다변량 비시각화: 둘 이상의 변수 간 관계 확인<br>
>> 다변량 시각화: 둘 이상의 변수간 관계를 전체적으로 파악<br>

## **3. 전처리**
: 데이터 분석(모델링) 전, 데이터를 적절한 방식으로 손질하는 과정. 누락 데이터, 중복 데이터 등 오류 수정. 

- 데이터 정제: 결측값(누락된 값) 보완. 이상값(튀는 값) 제거
- 데이터 변환: 일관성 확보 및 중복 최소화
- 데이터 필터링: 오류 제거 및 보정을 통한 데이터 품질 향상
- 데이터 통합: 유사한 성질의 데이터 연계 & 통합.

### 결측값 처리
결측값 = 데이터 수집 과정에서 측정되지 않거나 누락된 데이터 -> 변수간 관계 왜곡 -> 모델 정확성 하락 -> 시각화에 영향

1. 결측값(Nan, ?, 0) 확인: pandas 함수로 확인 가능

- info(): 데이터 프레이 요약. 유효한 값의 개수를 보여줌.
- value_counts(dropna=False): 각 열의 결측값 포함 전체 데이터 확인.
- isnull(): 누락 데이터면 True. e.g) df.isnull().sum()
- notnull(): 유효 데이터면 True.
- replace(): NaN이 아닌 결측치를 NaN으로 변환. 
 
3. 결측값 처리<br>
<데이터가 있는 열 또는 행 삭제>

> 행 삭제 -> 관측값(레코드) 제거<br>
> 열 삭제 -> 분석 대상의 특성(변수) 삭제

> 전체 삭제: 결측값이 발생한 모든 관측치 삭제 -> 간편 vs 모델 유효성 저하
> 부분 삭제: 모델에 포함시킬 변수 중 결측값이 발생한 모든 관측치 삭제 -> 관리 비용 증가

❗결측값이 무작위로 발생한 경우 사용. 그렇지 않은데 삭제하면 모델 왜곡 가능성

<결측값을 다른 값으로 대체>
: 평균값, 최빈값 활용

>일괄 대체: 모든 변수에 대해서 일괄적으로 같은 값으로 대체.

>유사 유형 대체: 유사한 범주에 따라 다른 값으로 대체

### 이상치 처리
이상치 = 관측된 데이터의 범위에서 많이 벗어난 값

1. 이상치 확인 <br>

a. 통계를 통해 확인 -> describe()<bR>
b. 시각화를 통해 확인 -> BoxPlot<br>
c. Z-score -> 해당 데이터가 평균으로부터 얼마의 표준 편차만큼 벗어나 있는지<br>
    - z < 0 : 평균 이하<br>
    - z > 0 : 평균 이상

> 데이터가 정규 분포를 따르면 이상치 판단 기준이 됨.

d. Tukey Fences<br>
: 사분위 범위 (IQR)을 기반으로, 두 가지 경우에 이상치라고 판단<br>
    - Q1 - (1.5 * IQR) 미만<br>
    - Q3 + (1.5 * IQR) 초과

❓IQR은 데이터의 중간 50% 범위를 측정 == 데이터 상위 25%와 하위 25%를 제외한 중간 범위를 의미

2. 이상치 제거 <br>

a. 전체 삭제: Human Error에 의해 발생하면 관측치 삭제<br>
b. 다른 값으로 대체: 관측치의 절대적 숫자가 작으면, Human Error에 의해 이상값이 발생했더라도 삭제 대신 평균 등으로 대체 OR 결측값 때처럼 다른 변수들을 사용해서 예측 모델을 만들고, 이상값을 예측, 해당 값으로 대체<br>
c. 변수화: 자연 발생한 이상값의 경우 -> **파악**하여 상황에 맞는 판단을 내림 <br>
d. 리샘플링: 자연발생한 이상값에 대해서 해당 값을 분리해서 새로운 모델을 만듦. <br>

### 피처 엔지니어링 (변수 가공)
: 해결하고자 하는 문제를 컴퓨터가 잘 이해할 수 있도록 피처(변수)들의 형태를 변형 또는 적절하게처리하는 과정. feature = 입력 변수 = column

-> 데이터 전처리의 마지막 단계. 기존의 데이터에 '정보'를 추가. 도메인 지식 여부가 가공 결과와 유의성에 차이 불러옴.

> 피처 엔지니어링 방식
>1. 레이블 인코딩, 원핫인코딩: 텍스트 -> 숫자
>2. 구간화(binning): 연속적인 값 -> 일정 구간
>3. 변환: 기존의 피처 -> 새로운 비처
>4. 스케일링: 변수의 값 범위를 일정 수준으로 맞추는 작업

#### 레이블 인코딩 vs 원핫 인코딩
레이블 인코딩: 범주형 변수 -> 0 ~ N-1 숫자로 변환<br>
원핫 인코딩: 범주형 변수를 이진(0, 1) 벡터로 변환

#### 구간화
연속 데이터 -> 일정한 구간으로 나눔. 분석가의 도메인 지식 극적으로 활용. `pd.cut()`

#### 변환
기존의 피처 -> 다른 피처 변환하여 **변수(열) 추가**

#### 스케일링
서로 다른 변수의 값 범위를 일정 수준으로 맞추는 작업 -> 숫자의 상대적 크기 차이로인한 결과 왜곡 방지.

➕변수 변환 = 변수의 단위 변경 OR 변수 분포가 편항되어 관계가 잘 안보이는 경우 -> Log 함수. 제곱근 취하기

a. Standard Scaler (표준화): 각 feature의 평균 = 0, 분산 = 1. 모든 피처들이 같은 스케일 가짐. -> 정규분포를 따른다고 가정하는 기술에 적합<br>
`StandardScaler()`

b. MinMaxScaler: 모든 피처가 0과 1사이에 위치하도록 만듦. <br>
-> 데이터가 서로 다른 비율의 속성으로 구성되어 있을 때 같은 비율로 속성 맞춤.<br>
-> 연산 속도를 높이고 알고리즘 최적화
`MinMaxScaler()` -> `skearn` 라이브러리 = 스케일러 제공

## **4. 시각화**
: 데이터를 시각적인 형태로 표현하는 과정 또는 결과물.<br>
그래프, 차트, 그림, 도표 => 패턴, 관계, 추세 파악

`Matplotlib`(Python 수학적 확장 & 그래프를 그리기 위한 라이브러리) 과 `Seaborn`(Matplotlib 기반 통계 데이터시각화 최적화 라이브러리) 사용

데이터를 바라보는 방식에 따라 사용하는 그래프가 다름.

### 파라미터
`hue=` -> '범형 변수를 넣어 내가 원하는 변수를 기준으로 데이터를 구분, 그래프에 표시 가능. 'hue'에 따른 'column'(변수)


### 그래프의 종류는 매우 다양하다! 
- boxplot: 사분위수와 이상치 확인
- countplot: 범주형 변수의 빈도수 제공
- hisplot: 도수분포표 -> 그림. 수치형 변수의 구간별 빈도수
- distplot, kdeplot (커널밀도추정 그래프) -> 히스토그램을 연속적으로 곡선 연결한 그래프
- barplot, pointplot: x(범주형 데이터 값)에 따른 y(수치형 데이터 값)의 평균 값. pointplot은 막대 그래프와 모양만 다르고 동일
- scatterplot (산점도 그래프), regplot(+회귀산): 두 변수 간의 관계 시각화. 각각의 데이터 포인트는 두 변수의 값 표현. x, y축에 데이터 포인트를 분산하여 그림. 변수 간의 패턴, 상관관계, 분포 파악 용이. regplot은 거기에 회귀선을 함께 표시
- catplot: categoryplot의 준말. 수치형 데이터 & 범주형 데이터 관계
- pieplot: 비율 강조할 때. 모든 데이터가 합쳐서 전체를 이루는 경우에 효과적.
- heatmap: 변수간 상관계수 파악. corr() 메서드로 변수간의 상관계수 구하고 히트맵에 표현. ❓상관계수는 두 수치형 변수 사이의 상관 관계의 정도를 수치적으로 나타낸 계수. 두 변수간 서로 영향을 주는 정도를 나타내는 값. -1 ~ 1 사이. 양극단에 가까울수록 큰 상관계수를 가짐.
  - 음의 상관계수 : 반비례 관계
  - 양의 상관계수 : 비례 관계
- pairplot: 여러 그래프 한 번에 찍기.여러 변수간의 산점도를 한 번에 보여주는 그래프.
