# 회귀 심화_과제: 클론 코딩 마크다운

집값을 설명하는 독립 변수를 살펴보기.

## 과정
1. 변수에 관한 설명 잘 읽기
2. 결측값 확인
3. df.describe().T(행과 열을 보기 좋게 뒤집어서 보여줌), df.info()로 데이터 통계 요약 및 데이터 형 파악하기 -> describe() 데이터 프레임에 우리가 원하는 정보(왜도, 첨도, 결측값 여부 등)를 추가로 넣어서 볼 수 있음.
4. 데이터 시각화 (hist, pairplot, ...)
5. 범주형 & 수치형 데이터를 분리해서 따로 관리할 수 있음.
6. df.apply(lambda x: <func>, axis=1)을 사용해서 컬럼의 값을 원하는 형태로 가공할 수 있음.
7. 산점도를 통해서 종속변수와 독립변수 사이의 선형관계를 대략적으로 파악할 수 있음.
8. 히트맵을 통해서 종속변수와 상관관계가 높은 독립 변수들을 파악할 수 있음. + 독립변수들 간 상관관계가 높은 것들에는 무엇이 있는지 살펴볼 수 있음. -> sort_values()를 이용하면 간편하게 확인 가능.
9. ❓"종속 변수와 높은 상관관계를 지닌 데이터 각각의 분포가 비슷하면 높은 다중공선성이 있다"고 해석해도 되는지 궁금함.

1. 회귀 분석할 때 항상 상수항 추가하기
2. fit()으로 선형 모델에 적합시키기.
3. summar() 찍어보기
4. 다중선형회귀분석의 경우, 다중공선성을 꼭 확인해야 함 -> VIF 확인
5. 모형의 유의성과 계수의 유의성이 확보되었을지라도, "오차에 대한 가정"을 만족하는지 꼭 확인해야 함.
- 정규성 검정: QQPlot, 잔차 패턴 확인: residual에 대한 산점도 찍어보기, 샤피로-윌크 검정 => p값이 0.05 미만일 때 정규성 만족한다고 볼 수 있음.
- 독립성 검정: 잔차 그래프, ACF(0시차 이후 구간을 벗어나는 시차 존재 X 경우에 독립성 만족)
- 등분산성: 잔차 그래프를 통해서 확인

## 느낀 점
데이터 분석을 시작할 때 각 변수의 의미를 명확히 이해하는 것의 중요성을 다시금 느꼈습니다. 산점도가 변수 간 선형관계의 유무를 파악하기에 참 좋은 도구인 것 같습니다. 범주형 변수와 수치형 변수를 EDA 때 분리해서 별도의 데이터프레임으로 유지하는 발상이 유용해 보였습니다. 다중회귀분석에서 다중공선성 파악을 마치면 끝이라고 생각했지만, 오차가 가정을 잘 따르고 있는지까지 적절한 통계지표를 가지고 평가하는 과정이 꼭 필요하다는 점도 기억에 남습니다.
