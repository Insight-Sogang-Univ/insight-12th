# 통계_사전과제

## 1. 통계란?
데이터 수집, 분석, 해석을 통해 패턴, 관계, 변동을 이해하고 결론을 도출하는 학문 및 방법론
-> 우리 주변에서 일어나는 일들에 관한 정보를 수집하고 이를 분석하여 패턴을 찾거나 미래를 예측하는 데 도움이 됨.

- 기술 통계학 : 데이터의 주요 특성을 요약 및 설명. `평균, 중앙값, 표준편차` 등의 개념이 이에 해당.<br>
=> 측정이나 실험을 통해 수집한 자료를 요약/정리하여 특성을 기술하는 절차에 관한 개념.

>수치 기술 통계<br>
숫자 형태로 데이터 요약

>표와 그래프 기술 통계<br>
표/그래프 형태로 정보 시각화

- 추론 통계 : 수집한 데이터 -> 미래 또는 수집하지 못한 정보를 추론, 결론을 도출하는 방법을 다루는 분야.

> 모집단에서 추출한 표본 집단을 통해 모집단에 대한 정보를 추측하는 과정으로, `통계적 가설 검정, 신뢰구간` 개념이 여기에 속함.

## 2. 기술 통계 (수치 기술 통계)
### 2.1 중심 위치 척도
1) 평균(mean)<br>
: `(데이터 값의 총합)/(데이터 개수)` -> **데이터의 중심 경향성을 나타냄**<br>
평균은 총합치를 사용. 따라서 이상치에 큰 영향을 받을 수 있음.<br>
`ex) series.mean()`

2) 중앙값(median)<br>
: 데이터를 크기 순으로 나열했을 때 가운데 위치한 값.<br>
평균과 달리 이상치에 덜 민감함. 데이터의 중심 경향성을 나타내는 대안적 지표임.
`ex) series.median()`

3) 최빈값(mode)<br>
: 데이터에서 가장 자주 나타나는 값.<br>
주로 **이산형 데이터(Discrete Data)** 에서 사용. **데이터의 분포** 이해에 도움을 줌.<br>
`ex) series.mode()`

평균, 중앙값, 최빈값을 통해 **왜도(Skewness)** 확인 가능. ❓왜도: 데이터 분포의 비대칭성을 나타내는 통계적 지표. 데이터 분포의 특징 기술에 사용.
> **Positive Skewness(Right-skewed)**<br>
왼쪽 꼬리 > 오른쪽 꼬리. 평균과 중앙값이 최빈값보다 큼.<br>
>**Negative Skewness(Left-skewed)**<br>
왼쪽 꼬리 < 오른쪽 꼬리. 평균과 중앙값이 최빈값보다 작음.<br>
>**Normal Distribution(정규 분포)**<br>
왼쪽 꼬리와 오른쪽 꼬리가 대칭. 평균 == 중앙값 == 최빈값<br>
`ex) series.skew()`

### 2.2 변동성 척도
1) 사분위수(quartile)<br>
: 데이터의 표본을 네 개의 동일한 크기로 나눈 값을 의미.<br>
- Q1 = 하위 25퍼센트에 위치한 값. '하위 사분위수'
- Q2 = 하위 50퍼센트에 위치한 값. '중앙값'
- Q3 = 하위 75퍼센트(상위 25퍼센트)에 위치한 값. '상위 사분위수'

2) 사분위간 범위(IQR)<br>
: Q3에서 Q1 값을 제외한 범위(데이터의 중앙 50%가 퍼져있는 범위) -> **이상치** 탐지에 유용
- 아래쪽 이상치: Q1 - 1.5 * IQR 보다 작은 값
- 위쪽 이상치: Q3 +1.5 * IQR 보다 큰 값

3) 범위(range)<br>
: 데이터의 최대, 최소 차이. 데이터의 전체 변동 범위를 알려줌.
- `ex) np.max(arr) - np.min(arr)`

4) 분산(variance)<br>
: 데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지 나타내는 측정 지표.<br>
- 분산이 높다 = 데이터가 고르게 퍼져있다.<br>
- `편차(각 데이터 값 - 평균)의 제곱의 평균`<br>
- `ex) np.var(arr)`

5) 표준편차(standard deviation)<br>
: 데이터 값들이 평균으로부터 얼마나 퍼져 있는지를 나타내는 중요한 지표.<br>
- 표준편차가 높다 = 데이터가 고르게 퍼져있다.<br>
- `분산의 제곱근`<br>
- ❓표준편차를 사용하는 이유?<br>
분산의 단위는 **관측값 단위의 제곱**임. -> 대상 데이터와 일치하지 않음. 표준편차는 분산의 제곱근이므로 원래 데이터와 동일한 값을 가짐. 단위의 일치성 덕분에 표준편차는 데이터를 해석하고 비교하는 데 **직관적**임.

6) 변동계수(coefficient of variation)<br>
: 데이터의 상대적 변동성을 나타내는 통계적 지표.<br>
- `(변동계수) = (표준편차 / 평균) * 100`
- ❓변동계수를 사용하는 이유?<br>
표준편차는 단위가 존재. 따라서 평균값이나 단위가 다른 두 표본집단을 비교하기에 적절하지 않음. 이때 표준편차를 평균으로 나누면 단위가 사라짐. 따라서 집단의 퍼진 정도를 동일 기준으로 비교할 수 있음.
- `ex) np.std(arr) / np.mean(arr) * 100`

### 2.3 연관성 척도
1) 공분산(covariance)<br>
: 두 변수가 **함께 어떤 방향으로 변화하는지와 그 크기** 표현
- Cov(X,Y) > 0 // X와 Y는 양의 관계. X가 커지면 Y도 커짐.
- Cov(X,Y) < 0 // X와 Y는 음의 관계. X가 커지면 Y는 작아짐.
- Cov(X,Y) = 0 // X와 Y는 양도 음도 아닌 관계 -> 선형적(직선적)인 관계는 없으나 **비선형적 관계(곡선, 꺾인 직선 등)** 가 존재할 수 있음. (X가 Y의 제곱에 비례하는 경우)<br>
`np.cov(df['col1'], df['col2'])`
`np.cov(df['col1'], df['col2'])[0][1]`
`

2) 상관계수(correlation)<br>
: 두 변수 간의 **선형적인 관계**를 나타내는 통계적 지
-1과 1사이의 값
- 상관관계 != 인과관계 => ❓상관관계: 두 변수가 수치적으로 연관 | 인과관계: 한 변수가 다른 변수를 발생시키는 관계.
- 공분산을 각 변수의 표준편차로 나누어 -1과 1사이로 표준화 한 것. => Corr(X,Y) > 0, 양의 상관관계 | Corr(X,Y) < 0, 음의 상관관계
`df[['col1', 'col2']].corr()`` df[['col1', 'col2']].corr().iloc[0,1`]

## 3. 추론 통계1(모집단 분포 추정)
- 모집단(Population)과 표본(Sample)<br>
-> 추출된 N개의 데이터를 **표본집단**이라고 하며, 표본과 **표본 통계량**을 통해 모집단 추정 가능.

- 대수의 법칙: 표본의 크기가 커질수록 표본 평균이 모평균에 근사하는 현상. 데이터는 다다익선.
- 중심극한정리: 표본의 크기가 크면 표본 평균의 분포가 **모집단의 분포와 상관 없이 정규분포에 근사**함.

### 3.1 확률 분포
- 확률 분포(Probability Distribution): 확률 변수가 특정한 값을 가질 확률을 나타내는 함수
- 확률 변수: 어떤 사건 또는 실험의 가능한 결과를 숫자에 대응시키는 함수.
> 이산 확률 분포: 확률 변수가 이산적인 값을 가지는 경우<br>
연속 확률 분포: 확률 변수가 연속적인 값을 가지는 경우

### 3.2 확률 분포의 종류
1) 정규분포(Normal Distribution)
- 종 모양
- 평균 기준 대칭
- 표준편차가 클수록 곡선이 평평해짐. 데이터가 넓게 퍼져있음.
- 매개변수: 평균, 표준편차
- 가장 많이 알려진 확률 분호. 많은 자연 현상과 사회 현상이 이를 따르는 것으로 알려짐.

2) T분포: 표본 평균이 어떤 분포를 가지는가?
- 종 모양
- 평균 기준 대칭
- **자유도**가 커짐에 따라 표준정규분포 N(0,1)에 수렴
- 매개변수: **자유도(df)** ❓자유도: 통계에서 "얼마나 많은 정보가 독립적으로 변할 수 있는가" <br>
-> T분포의 자유도는 표본 크기에 비례함. 자유도가 높을수록, T분포는 정규분포에 근사하고, 통계적 분석의 정확도가 높아질 가능성이 있음.<br>
->❗너무 높은 자유도를 가지면 **오버피팅** 문제 발생 ❓오버피팅: 모델이 학습 데이터에 너무 맞춰져서 새로운 데이터에 대한 일반화 성능이 저하되는 현상.
- 표본으로 모평균(혹은 모평균간의 차이)를 추정할 때 사용되는 분포
- 표본 크기가 작을 때 정규분포보다 더 정확한 결과를 제공
- **모집단의 분산을 모를 때**, 표본의 분산을 사용해 모집단 분산을 추정하여 만들어짐. 따라서 표본의 크기가 작을수록 이 불확실성이 커지고, 이 불확실성 때문에 **t분포는 표준정규분포보다 중심에서 더 퍼져 있게 됨.** -> **극단적인 값이 나올 가능성을 반영**

3) 카이제곱분포(Chi-Square Distribution)
- 0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭.
- **자유도가 증가함**에 따라 중심이 오른쪽으로 이동, 더 대칭적, 더 정규분포에 가까워짐.
- 독립적인 표준정규분포를 따르는 변수들의 제곱 합.
- 주로 독립성 검정, 적합도 검정 등에 사용됨.

## 4. 추론 통계2(가설 검정)
### 4.1 가설 검정의 절차
> 가설 설정 - 유의 수준 설정 - 검정 통계량 산출 - 가설 기각/채택 판단

1) 가설 설정
: 귀무가설(H0), 대립가설(H1) 설정. => 귀무가설은 주로 **기각하고자 하는 명제**.<br> 대립 가설은 **내가 주장하고자 하는 명제**.

- 1종, 2종 오류
-> 1종 오류 𝛼: 귀무가설이 실제로 참이었는데 기각할 확률 ❗**더 치명적**<br>
-> 2종 오류 𝛽: 귀무가설이 실제로 거짓이었는데 기각하지 못할 확률

2) 유의 수준 설정
: 정해진 수치를 얼마나 벗어났을 때 귀무가설을 기각할 것인가를 판단하는 기준. 결과 판단 시 허용할 수 있는 오류의 최대치.

3) 검정 통계량 산출
: 가설 검정의 결과를 판단하는 데 사용되는 값. -> 데이터의 분포나 특징에 따라 통계량 결정. <br>
- Z값
> - 표준 정규분포를 사용하여 얻은 값. 특정 데이터가 평균으로부터 얼마나 떨어져있는 지를 표준편차 단위로 나타냄.<br>
> - $Z=(X−μ)/σ$<br>
> - 큰 표본 크기(30 이상)에 대한 가설 검정에 사용 e.g) 모집단의 평균에 대한 가설 검정할 때 사용<br>
> - |Z|가 크다 = 평균으로부터 멀리 떨어져 있으며 이상치일 가능성.<br>
- T값
> - T분포를 사용하여 얻은 값. 작은 표본 크기에 대한 가설 검정(T검정)에 사용
> - T분포는 정규분포와 유사. 하지만 꼬리 부분이 더 두꺼워 작은 표본에서 발생할 수 있는 불확실성을 더 잘 반영함.
> - $T = \frac{X - \mu}{s / \sqrt{n}}$> - 모집단의 평균에 대한 가설 검정, 두 표본 평균 간의 차이 검정 시에 사용
> - |T|가 크다 = 해당 데이터가 모집단 평균으로부터 멀리 떨어져 있으며 평균과 유의미한 차이 있을 가능성. <br>

- F값
> - F분포를 사용하여 얻은 값. 두 표본 분산 사이의 차이를 비교하는 데 사용
> - 두 개의 독립적인 카이제곱 분포의 변수들의 비율로 정의됨.
> - F = 분산1/분산2. (각 분산은 두 독립적인 표본의 분산)
> - 주로 분산 분석(ANOVA)에서 두 개 이상의 그룹 간 평균의 차이를 검정하는 데 사용
> - F값이 1에 가까울 수록 두 표본 분산이 서로 비슷(퍼진 정도가 비슷)하다고 볼 수 있음.값

4) 가설 기각/채택 판단
- 유의 확률(p-value)
: 가설 검정에 쓰일 수 이씨는 또 다른 척도. 유의 확률은 검정통계량에서 유도하며 **유의 확률 p가 유의 수준 𝛼보다 작을 시에 귀무가설을 기각**, 클 시에는 귀무 가설을 채택해야 함.

### 4.2 여러 검정 방법
#### 수치형-수치형: 피어슨 상관분석<br>
: 두 변수간 피어슨 상관계수가 얼마나 **유의미한 지**를 검정하는 방법.<br>
H0: 두 변수간 선형 상관관계가 존재하지 않는다.<br>
H1: 두 변수간 선형 상관관계가 존재한다.<br>
(스피어만 상관계수 -> 데이터를 순위화해서 피어슨 상관계수를 구한 것.)<br>

시각화: scatter plot을 봐도 잘 모르는 상황에서 상관 분석을 이용함.

상관계수(r)에 대해서,
> |r|<0.3 -> 두 변수간 상관관계 약함.<br>
|r|<0.5 -> 두 변수간 상관관계가 중간.<br>
|r|>0.5 -> 두 변수간 상관관계가 강함.

p-value는 보통 0.05이하를 귀무가설 기각의 기준으로 사용함.<br>
❗이는 참고 지표이며 절대적 기준이 아님. 상관 계수는 선형 관계만 파악(직선의 기울기 고려 X)

#### 범주형-수치형: F-test(ANOVA), T-test
- 시각화: `sns.barplot`을 사용. (검은 선 = 각 평균 95퍼센트의 신뢰구간 == 95퍼센트 확률로 신뢰구간 안에 모평균이 포함됨)<br>
해석: 두 평균의 차이가 크고, 신뢰구간이 거의 겹치지 않다 -> 대립가설을 지지할 가능성이 높다.

- 수치화(범주 2개): **T-test** 사용. T-test 수행 시 `t = (두 평균 간 차이) / (표준 오차)`
|t| > 2 이면서 클수록 차이가 크다고 생각함. p값은 결과의 신뢰성. 0.05이하 => 귀무가설 기각(대립가설 채택)

❓ttest_ind: ind(independent) -> 두 범주가 서로 독립적인 경우(애초에 다른 집단) (<-> ttest_rel == 같은 그룹이 다른 상태의 2개 그룹으로 나누어진 경우)

양측 검정 == 두 그룹간 "차이" 존재?<br>
단측 검정 == 두 그룹 간 차이의 "한 쪽"만 보고 싶을 때 사용

t분포는 좌우 대칭. 따라서 코드는 똑같이 하되 결과 p 값을 2로 나누면 단측 검정의 결과로 볼 수 있음.

- 수치화(범주 3개 이상): 범주가 너무 많아지면 분산분석(ANOVA) 사용. ANOVA는 F-test의 일종.
=> One way ANOVA
> `F=(집단 간 분산) / 집단 내 분산<br>
F가 2~3 이상이면 차이가 있고 클수록 차이가 큼. p값 동일.

❗anova는 평균 대비 각 그룹간 차이가 있는지만 알 수 있음. 따라서 하나가 튀고 나머지가 비슷하더라도 귀무가설을 기각할 수 있음.

#### 범주형-범주형: Chi2-test
- 시각화: Mosaic Plot
- 수치화: 기대빈도와 관측빈도의 차이가 크다면 두 범주가 독립적이지 않을 가능성이 높음.<br>
`(관측빈도 - 기대빈도)^2 / 기대빈도`의 합. -> 범주 개수에 제한이 없음. 따라서 자유도의 2배 가량 기준으로 더 크면 차이 존재.

범주형 변수의 자유도 = 범주의 수 - 1(1개는 나머지가 정해지면 자동으로 결정됨)

카이제곱 변수에서의 자유도 = x 변수의 자유도 * y 변수의 자유도.

샘플 크기가 작을 때 (기대빈도가 5이하인 셀이 25퍼센트 이상인 경우) -> 피셔의 exact_test를 통해서 정확한 p값 획득 가능.

#### 수치형-범주형
- 시각화: sns.kdeplot -> 수치형이 변화할 때 범주형의 비율이 어떻게 변화하는가? => 관계가 없다면 개형이 비슷함.
- 수치화: 수치형을 구간별로 잘라서 범주화하지 않는 이상 검정 방법이 마땅치 않음. => 범주형을 target, 수치형을 feature로 두고 모델을 만들고, 그 모델의 계수를 유의성 검정해서 그에 대한 p값을 보고 판단.
