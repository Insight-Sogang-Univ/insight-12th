# 머신러닝

- 머신러닝이란 과거의 경험에서 학습을 통해 얻은 지식을 미래의 결정에 이용하는 학문 및 분야를 뜻합니다.
- 관측된 패턴을 일반화하거나 새로운 규칙을 생성하는 것을 목표로 합니다.
- 학습 방식에 따라 지도학습, 비지도학습, 강화학습으로 분류할 수 있습니다.


- 지도학습
    - 목표변수와 다른 변수간의 관계를 기계가 학습하도록 가르치는 것, 즉 정답을 아는 상태에서 학습을 시키는 방식입니다.
        - ex) 회귀, 분류
- 비지도학습
    - 지도나 목표변수를 설정하는 과정 없이 알고리즘이 스스로 학습하도록 하는 것, 즉 정답을 모르는 상태에서 학습을 시키는 방식입니다.
        - ex) 차원 축소, 클러스터링
- 강화학습
    - 기계가 지도를 받지 않고 스스로 결정을 내린 후에 그 결과에 따라 보상(피드백)을 받고, 보상값에 따라 행동을 재평가하며 학습하는 방식입니다.



### 데이터 분할

- 머신러닝 모델을 만들 때, 성능을 정확히 평가하기 위해서는 데이터 분할이라는 과정을 거쳐야 합니다.
- 데이터 분할시에는 보통 모델을 학습시킬 때 사용할 훈련 데이터(train data)와 모델을 평가할 때 사용할 테스트 데이터(test data)로 나눕니다.
- 만약 테스트 데이터를 따로 뽑지 않은 상태에서 모델 평가를 한다면, 그것이 학습 데이터에 대한 단순 암기의 결과일 수도 있기 때문에 정확성이 떨어집니다.
- 따라서 테스트 데이터는 반드시 학습에 사용되지 않은 것이어야 하며, 일반적으로 학습 데이터와 테스트 데이터의 비율은 7:3 ~ 8:2 정도로 선택됩니다.
<hr>



#### 과소적합(under-fitting)
- 머신러닝 모델의 성능을 평가할 때, 적합도가 낮다면 과소적합 현상이 일어났다고 볼 수 있습니다.
- 모델의 설명력이 충분하지 않다는 뜻이므로 훈련 데이터를 통해 추가적인 학습을 진행하여 문제를 해결할 수 있습니다.

#### 과대적합(over-fitting)
- 그러나 학습이 지나치게 많이 이루어질 경우, 과(대)적합이 발생할 우려가 있습니다.
- 모델 학습이 과도하게 이루어져 훈련 데이터에 대해서는 거의 완벽하게 설명하지만, 새로운 데이터에 대한 예측력은 떨어지는 경우입니다.
- 데이터 분할 시 훈련, 테스트 데이터와 함께 검증 데이터도 뽑아 모델의 초매개변수를 튜닝하는 방식으로 문제를 해결할 수 있습니다.
- 머신러닝 모델을 학습시킬 때는 적절한 훈련 지점을 찾아 과소적합과 과대적합을 모두 방지하는 것이 관건이라 할 수 있습니다.




# 회귀

- 회귀분석은 독립변수들이 종속변수에 미치는 영향을 파악하는 분석방법입니다.
- 데이터에서 패턴을 찾아내 미래의 값을 예측할 수 있고, 회귀식을 통해 그 내용을 쉽게 이해할 수 있어서 머신러닝에도 유용하게 활용됩니다.
- 일반적으로 연속형 변수들 사이의 관계를 파악할 때 사용됩니다.

## 선형회귀
- 독립변수와 종속변수 간의 관계를 가장 잘 설명하는 회귀선을 찾아 예측하는 과정입니다.
- 데이터를 분석할 때 선형회귀를 사용하기 위해서는 4가지 기본 가정이 지켜져야 하는데, 내용은 다음과 같습니다.
    - 선형성 : 종속변수와 독립변수는 선형관계
    - 등분산성 : 잔차의 분산이 고르게 분포
    - 정규성 : 잔차가 정규분포의 특성을 가짐
    - 독립성 : 독립변수들은 서로 상관관계가 없음
- 독립변수의 개수에 따라 크게 단순회귀와 다중회귀로 구분할 수 있습니다.
 
### 단순선형회귀분석
- 오직 하나의 독립변수만이 존재하는 선형 회귀입니다.
- 회귀선은 y = ax + b와 같이 직선의 형태로 표현됩니다.
    - 위 일차식에서 a는 회귀 계수, b는 절편을 의미합니다.
    - 단순회귀는 데이터를 가장 잘 설명하는 회귀계수와 절편을 찾는 과정이며, 이를 위해 최소제곱법을 사용합니다.
        - **최소제곱법**이란, 잔차의 제곱합이 최소가 되는 회귀계수와 절편을 구하는 방법입니다.
            - 잔차는 계산값과 예측값의 차이를 뜻하며, 오차와 비슷한 개념입니다.
                - 다만 오차는 모집단 기준, 잔차는 표본집단 기준으로 사용된다는 점에서 약간의 차이가 있습니다.
            - 잔차가 최소가 된다는 것은 회귀선과 실제 데이터 간의 차이도 최소가 된다는 뜻이기 때문에 예측력이 높아지게 됩니다.


### 다중선형회귀분석
- 독립변수가 2개 이상 존재하는 선형 회귀입니다.
- 독립변수가 2개일 때, 회귀선은 3차원 공간에서 평면으로 표시됩니다.
- 다중회귀에서도 마찬가지로 회귀계수를 구하기 위해 최소제곱법을 사용하며, 이때 잔차는 관측치와 평면간의 수직거리가 됩니다.


#### 다중공선성
- 한편, 독립변수가 여러 개인 다중회귀에서는 각 독립변수들 간 상관관계가 클 때 분석이 정확도가 떨어지고 모델이 불안정해진다는 다중공선성 문제가 발생합니다.
    - 선형회귀의 기본가정 중 독립성에 위배되는 경우라고 볼 수 있습니다.
- 종속변수에 대한 설명력이 낮은 변수들을 제거하거나, 중요한 변수들만을 선택하는 방식으로 다중공선성 문제를 해결할 수 있습니다.
    - 선형 회귀에 규제를 가하는 릿지(L2)/라쏘(L1) 모델을 사용하는 방법도 있습니다.
        - 설명력이 낮은 설명변수에 대해 영향력을 줄이거나 아예 제거하는 방식이며, 과대적합 방지에도 유용합니다.
    - 주성분 분석(PCA)을 통해서도 다중공선성을 해결할 수 있습니다.
        - 상관성 높은 변수들의 선형결합을 통해 차원을 축소하여 새로운 변수를 생성하는 방식으로, 데이터의 복잡성을 낮출 수 있습니다.

## 모델 평가
- 회귀 모델의 성능은 데이터에 대한 설명력, 예측력과도 같습니다.
- 회귀 모델의 성능을 평가할 때는 분산분석(ANOVA)표를 참고해 볼 수 있습니다.

### 분산분석표

- SSR은 Regression Sum of Square의 약자로, (예측값 - 평균)^2의 총합입니다.
    - (예측값의 분산과 유사, 설명 가능한 부분)
- SSE는 Residual(Error) Sum of Square의 약자로, (실제값 - 예측값)^2의 총합입니다.
    - (잔차의 분산과 유사, 설명 불가능한 부분)
    - MSE는 잔차제곱의 평균으로, SSE를 잔차의 자유도로 나눠 표준화한 값입니다.
- SST는 Total Sum of Square의 약자로, (실제값 - 평균)^2의 총합입니다.
    - (실제값의 분산과 유사, SSR + SSE)
- 성능을 평가할 땐 결정계수(R-Square)를 사용합니다.
    - SSR/SST (= 1 - SSE/SST)로 계산하며, 1에 가까울수록 성능이 좋다고 판단합니다.
    - 다중회귀에서는 Adjusted R-square를 평가지표로 사용하며, 계산식은 1-(n-1)MSE/SST입니다.
 
### 회귀모형의 검정
- 다중회귀 모델의 유의성을 판단할 때는 두 단계의 검정을 거칩니다.
- 먼저 모형 자체가 통계적으로 유의미한지 판단하기 위해 F검정을 합니다.
    - 이때 귀무가설은 '모든 회귀계수는 0이다'가 되고, 귀무가설을 기각하는 것을 목표로 합니다. (대립가설 : 모든 회귀계수가 0은 아니다)
- 만약 위 귀무가설을 기각하는 것에 성공했다면, 다음으로 t검정을 통해 각각의 회귀계수들이 유의미한지 확인합니다.
    - 이때 각각의 회귀계수에 대한 귀무가설은 '회귀계수는 0이다'가 되고, 이것까지 모두 기각할 수 있다면 해당 모델을 활용할 수 있습니다.
- 이후 모형의 설명력에 대해 평가할 때는 앞서 보았던 결정계수(R square) 값을 통해 판단합니다.
