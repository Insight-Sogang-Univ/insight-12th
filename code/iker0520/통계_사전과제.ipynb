# 통계
- 통계학은 데이터를 수집 및 분석하여 그로부터 패턴을 찾아내고 결론을 도출해내는 학문이자 방법론입니다.
- 역사는 반복된다는 대전제 하에, 통계학적 지식을 활용한 패턴탐색은 미래 예측에도 도움이 될 수 있습니다.
- 오늘날 정보기술의 발전 등으로 인해 데이터를 쉽게 구할 수 있지만, 데이터 자체만으로는 의미 있는 정보를 얻을 수 없기 때문에 통계학적 지식을 활용해야 합니다.
- 통계는 크게 기술통계와 추론통계로 구분할 수 있습니다.

## 기술통계
- 기술 통계란 표본의 요약 통계량을 이용해 데이터를 분석하는 것을 말합니다.
- 기술통계와 관련있는 대표적인 값으로 평균, 중앙값, 최빈값이 있습니다.
- 자료가 수치형 척도이고 정규분포를 따른다면 산술평균과 표준편차를, 비정규분포라면 중앙값 또는 사분위수(IQR)을 주로 사용합니다.
- 자료가 범주형 척도일 땐 최빈값이 유용합니다.
- 표본의 특성을 한눈에 알 수 있다는 점에서 시각화 기법과도 잘 어울립니다.

## 추론통계
#### 전수조사
- 만약 어떤 집단의 특성에 대해 파악하고 싶다면, 가장 확실한 방법은 집단의 전체 인원에 대해 조사를 시행하는 것입니다.
    - 그러나 시간, 비용적인 측면에서 이는 다소 비현실적인 방법입니다.
        - (ex 대한민국 국민의 소득수준을 파악하기 위해 5000만 전국민에 대해 조사 수행)
#### 표본조사
- 때문에 모집단에 대해 일부만, 즉 표본집단을 추출하여 분석하고 그 결과를 바탕으로 모집단의 특성을 파악하는 방법을 주로 사용합니다.
    - 물론 전수조사에 비해 모집단에 대한 정확성이 떨어질 수 있지만, 표본과 표본통계량을 통해 허용가능한 오차범위 내에서 추론이 가능합니다.
- 이렇게 표본을 통해 모집단의 분포를 추정하는 것이 바로 추론 통계의 영역입니다.

### 분포 모형
- 표본집단을 추출하기 위해 무작위로 데이터를 샘플링했다면, 그 표본집단은 여러 확률분포모형 중 하나를 따르게 될 것입니다.
    - 확률분포는 확률변수가 특정한 값을 가질 확률을 나타내는 함수입니다.
    - 확률변수는 random한 세계에서 어떤 특정 사건이 발생할 때 나타날 수 있는 모든 경우의 수를 값으로 가집니다.
        - 알려진 분포모형들을 따른다는 것은 random sampling이 됐다는 뜻으로도 해석할 수 있습니다.
        - 만약 분포모형과 일치하지 않는 부분이 있다면 인위적인 개입을 의심해볼 수 있다는 점에서 분포모형들을 숙지하는 것은 의미가 있습니다.
- 확률분포모형은 크게 확률변수가 이산형이냐, 연속형이냐에 따라 이산확률분포와 연속확률분포로 나뉩니다.
    - 대표적인 이산확률분포로는 베르누이분포, 이항분포, 푸아송분포 등이 있습니다.
    - 대표적인 연속확률분포로는 정규분포, T분포, F분포 등이 있습니다.

#### 정규분포


- 가장 널리 알려진 확률분포 모형입니다.
- 많은 자연, 사회현상이 정규분포를 따르고 있다고 알려져 있습니다.
- 기술통계에서 산술평균을 대표치로 사용할지, 중앙값을 대표치로 사용할지에 대한 기준점으로 쓰이기도 합니다.
- 중심극한정리에 따르면, 표본의 크기가 커질수록 표본평균의 분포는 **모집단의 분포와 상관없이** 정규분포에 가까워집니다.
    - **즉, 모집단의 분포를 몰라도 표본의 크기가 충분히 크다면(n>=30) 근사적으로 정규분포를 따른다고 가정할 수 있습니다.**

그러나 표본의 크기가 충분히 크지 않거나 **모분산이 알려져 있지 않은 경우** 추정 및 검정시에 정규분포를 사용하기 어려운데, 이때 그대신 t분포를 쓸 수 있습니다.
### t분포


- t분포는 모분산을 모를 때 표본분산을 활용해 모분산을 추정하는 방식으로 만들어집니다.
- 따라서 정규분포에 비해 불확실성이 높고, 이는 중심으로부터 보다 멀리 퍼져 있는 형태로 반영됩니다.
- 종(bell) 모양이고, 평균을 기준으로 대칭된다는 점에서 정규분포와 형태가 비슷하지만, 정규분포보다 꼬리가 두껍습니다.
- 평균과 표준편차를 매개변수로 사용하는 정규분포와 달리, 자유도를 매개변수로 씁니다.
    - t분포의 자유도는 표본의 크기와 비례하기 때문에 자유도가 커질수록 t분포는 정규분포에 가까워지고, 분석의 정확성 또한 높아집니다.
    - 그러나 지나친 자유도는 오버피팅(overfitting)을 유발할 우려가 있습니다.

### 가설검정

- 모집단의 특성에 대한 주장을 가설로 세우고 표본조사로 가설의 채택여부를 판정하는 것을 말합니다.
- 귀무가설(H0) : 일반적으로 통용되는 가설입니다. (차이가 없다)
- 대립가설(H1) : 귀무가설을 기각하는 가설입니다. (차이가 있다)
    - 내가 주장/증명하고자 하는 가설이기도 합니다.
- 유의수준(a) : 귀무가설이 참임에도 불구하고 잘못 기각하는 1종 오류를 범할 확률의 허용 한계입니다. (일반적으로 5%)
- 유의확률(p-value) : 귀무가설을 지지하는 정도를 나타내는 확률입니다.
    - 가설검정을 통해 얻은 유의확률(p-value)이 내가 임의로 설정한 유의수준보다 높다면 귀무가설을 채택하고, 낮다면 귀무가설을 기각할 수 있습니다. 


#### 가설검정 과정
1. 귀무가설/대립가설 설정
2. 유의수준 설정
3. 데이터 수집 및 검정통계량(z값, t값, f값 등) 계산
4. p-value 계산 및 해석
5. 귀무가설 채택 및 기각 여부 결정
