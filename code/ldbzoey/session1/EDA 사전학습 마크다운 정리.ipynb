0. 데이터 분석 프로세스
   데이터 수집 -> 데이터 탐색 -> 데이터 전처리 -> 데이터 모델링
   - 데이터 탐색(EDA) 
   - 데이터 전처리

1. EDA의 개념 
    - 데이터 분석(모델링)을 위해 데이터를 여러 방식으로 파악하는 모든 과정
    - 탐색적 데이터 분석 과정 
    - 시각화 등의 도구를 통한 패턴 발견/ 데이터의 특이성 확인

    (1) EDA 대상
   
        - 일변량 Univariate
            EDA 통해 한번에 파악하려는 변수가 한 개
            데이터를 설명하고 존재하는 패턴을 찾는 것이 목적 
        - 다변량 Multi-variate
            여러개 일 때
            변수 간 관계를 보는 것이 주요 목적 
    (2) EDA 종류
   
        - 시각화 Graphic
            대략적 형태 파악
        - 비시각화 Non-Graphic
            주로 Summary Statistics
            정확한 값 필요할 때 
    (3) EDA 유형
   
        - Uni-Non Graphic (일변량 비시각화)
        - Uni-Graphic (일변량 시각화)
        - Multi-Non Graphic (다변량 비시각화)
        - Multi-Graphic (다변량 시각화)


2. 데이터 확인

   - 데이터를 불러오는 것 - 읽기
   - 데이터를 만드는 것 - 쓰기

    (1) 절대경로와 상대경로
   
        - 절대 경로 : URL 등 전체적인 경로
        - 상대경로 : 작업중인 디렉터리를 기준
    (2) 데이터 입출력
   
       - 데이터 파일 형태따라 읽고 쓰는 함수 다름
    (3) CSV 와 EXCEL

       - CSV : 데이터의 크기가 작고 압축이 용이
       - EXCEL : 행과 열이 데이터프레임의 행, 열
       - 두 형식 모두 pandas를 통해 DataFrame 형식으로 읽어올 수 있음 
   

3. 데이터셋 파악하기
   
   (1) 데이터 프레임 보기
   
   (2) 데이터 변수 확인
   
       - 어떤 정보를 가지는지, 유형은 어떠한지, 형식은 어떠한지
       1. 범주형
           - 명목형
           - 순서형
       2. 수치형
           - 이산형
           - 연속형
   
    (3) 데이터 분포 확인
   
       평균값, 최빈값, 중간값 등을 통해 전처리 아이디어를 얻을 수 있음 

4. 전처리
   
    데이터 정제, 변환, 필터링, 통합 등의 과정을 거치며 데이터 품질을 높이고 머신러닝 등 데이터 분석의 정확도 개선에 기여

    # 결측값 처리

    데이터 수집과정에서 측정되지 않거나 누락된 데이터
    
    1. 결측값 확인 
    
        info(), value_counts(dropna=False) 등의 함수 통해 결측값 확인 
        
    2. 결측값 처리 
    
        a. 삭제
        
            행삭제 : 관측값 (레코드) 삭제
            열삭제 : 특성 (변수) 삭제 

            전체 삭제 : 결측값 발생한 모든 관측치 삭제
            부분 삭제 : 변수 증 결측값 발생한 모든 관측치 삭제 -> 모델따라 변수 제각각 될 수도 

        b. 대체 

            일괄 대체 : 모든 변수를 일괄 같은 값으로 대체
            유사 유형 대체 : 범주형 변수 활용하여 유사한 범주에 따라 다른 값으로 대체
            : fillna()

            

    #이상치 처리 

    데이터 범위에서 많이 벗어난 값 = outlier 

    1. 이상치 확인 
        통계, 시각화, z-score, Tukey Fences 통해 
    2. 이상치 제거 
        전체삭제, 다른값으로 대체, 변수화, 리샘플링 

    #피처 엔지니어링 (변수가공) 

        피처(변수)의 형태를 변형하거나 적절하게 처리하는 과정 

    1. 레이블 인코딩, 원핫인코딩 : 텍스트로 주어지는 값을 숫자로 바꾸는 작업 

        - lable Encoding : 범주형 변수를 0 ~ N-1 의 숫자로 
        - One-Hot Encoding : 범주형 변수를 이진 벡터로 
        
    2. 구간화 : 연속적인 값을 일정한 구간으로 구분
    
        pd.cut(데이터프레임['컬럼명'], bins=[나누는 기준 리스트], labels=[지정할 label])
    
    3. 변환 : 기존의 피처를 변환해 새로운 피처 생성
    
    4. 스케일링 : 서로 다른 변수 값 범위를 일정 수준으로 맞추는 작업 

        - 표준화 : 모든 feature의 평균을 0, 분산을 1로
        - MinMaxScaler : 모든 feature가 0과 1사이에 오도록
