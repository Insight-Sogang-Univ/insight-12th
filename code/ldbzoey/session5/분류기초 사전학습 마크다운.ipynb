### 1. 지도 학습과 비지도 학습의 차이에 대해 설명해주세요.
지도 학습과 비지도 학슴은 모두 머신러닝의 일종
- 머신 러닝이란?
  - 인공지능의 한 분야로 컴퓨터가 스스로 학습할 수 있도록 도와주는 알고리즘/ 기술 개발 분야

- 지도학습
    문제와 정답을 모두 알려주고 공부시키는 방법
    즉 정답에 해당하는 label가 있는 데이터를 사용하여 기계를 학습시키는 것
    이에 입력 값과 결과 값을 같이주어 학습시켜 이 학습의 예측값이 정답과 같아지도록 

- 비지도학습
    답을 가르쳐주지 않고 공부시키는 방법
    정답이 없는 데이터를 사용하여 기계 학습하도록 유도

  
### 2. 회귀와 분류의 차이에 대해 설명해주세요.
지도학습 내에 대표적인 개념으로 있는 회귀와 분류
- 회귀
주어진 데이터(X)를 기반으로 정답(Y)를 잘 맞추는 함수
데이터가 연속형 변수를 예측하기 위해 사용될 때


- 분류
기존 데이터가 어떤 레이블에 속하는지 패턴을 알고리즘으로 인지한 뒤에 새롭게 관측된 데이터에 대한 레이블을 판별하는 것
데이터가 범주형 변수를 예측하기 위해 사용될 때

### 3. 분류 모델의 네 가지 종류와, 각 모델이 무엇인지 간단하게 정리해주세요.
분류 모델에는 로지스틱 회귀, 결정 나무, 서포트 벡터 머신, KNN이 있음
1. 로지스틱 회귀 : 이진 분류 문제를 푸는 알고리즘, 샘플이 특정 클래스에 속할 확률을 추정, 출력값을 0에서 1사이로 변환
2. 결정 나무 : 조건에 따라 데이터를 분류하며 최종적으로 데이터가 순수한 lable 집합으로 구성될 때까지 분류를 반복, 불순도(지니계수)가 낮아지는 방향으로 임계값 나누어야 함
3. 서포트 벡터 머신 : 클래스를 분류할 수 있는 다양한 경계선 중 최적의 라인을 찾아내는 알고리즘
4. KNN : 데이터로부터 거리가 가까운 k개의 다른 데이터 레이블을 참조하여 분류하는 알고리즘 (데이터 준비-> k값 설정 -> 거리계산 -> 가까운 k개 이웃선택 -> 분류)
   
### 4. 분류 평가 지표에는 무엇이 있는지 작성해주세요.
1. 혼동행렬
   분류 모델의 예측 결과를 정확한 예측 잘못된 예측으로 구분하여 나타낸 표(행렬)
   예측 성공여부와 예측값으로 구분하여 총 4종류의 경우의 수가 있음
2. F1 score
   정밀도와 재현율의 조화 평균, 그 둘의 균형을 효과적으로 평가하기 위함임
3. ROC/ AUC Curve
   ROC curve : 얼마나 분류가 잘 되어있는가를 보여주는 그래프
   AUC curve : ROC와 x축 사이의 면적(적분값)
4. 다중 분류 평가 지표
   상위 지표는 모두 이진 분류 평가 지표
   다중 분류 평가 지표는 이진 평가지표 사용하여 클래스별로 점수 구한뒤 적절히 평균을 내리는 방식
### 5. 하이퍼파라미터 최적화가 무엇인지 쓰세요.
- 하이퍼파라미터 : 학습 시작 전에 사용자가 직접 설명하는 변수
  모델 학습 과정에 반영되는 값
  학습시작 전에 사용자가 직접 설정하는 변수
- 하이퍼파라미터 최적화 : 적절한 하이퍼파라미터를 찾아 모델 성능을 향상시키는 것
  그 과정
    1. Hyperparameter 탐색 범위 설정 (최적 값을 찾고 싶은 하이퍼파라미터의 범위를 설정)
    2. 평가 지표 계산 함수(성능 평가 함수) 정의(탐색하려는 Hyperparameter를 인수로 받아 평가지표 값을 계산해주는 함수를 정의)
    3. 1단계에서 샘플링한 Hyperparameter 값을 사용하여 검증 데이터로 정확도 평가
    4. 위 단계를 특정 횟수 반복하며, 정확도 결과를 보고 하이퍼파라미터의 범위 좁힘
