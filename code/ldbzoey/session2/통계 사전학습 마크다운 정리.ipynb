## 1.통계

데이터를 수집, 분석, 해석하는 과정을 통해 패턴, 관계, 변동 등을 이해하고 결론을 도출하는 학문 및 방법론

    (1)기술통계학
        데이터의 주요 특성을 요약하고 설명하는 방법
        평균, 중앙값, 표준편차
        측정이나 실험을 통해 수집한 자료를 정보로써 전환하기 위하여 요약하고 정리하여 그 특성을 기술하는 절차에 관한 개념
        - 수치기술통계
            숫자 형태로 데이터를 요약
        - 표와 그래프 기술통계
            표와 그래프 형태로 정보를 시각화
    (2)추론통계학
        수집한 데이터를 기반으로 미래의, 혹은 조사하지 못한 전 정보를 추론하고 결론을 도출
        전체 집단(=모집단)에서 일부를 뽑아 만든 표본 집단을 통해 전체 집단에 대한 정보를 추측하는 과정
        통계적 가설검정과 신뢰구간 

## 2. 기술통계
    2.1 중심 위치 척도 
        - 평균
            데이터 값 총합/ 데이터 개수
            outlier에 민감할 수 있음
        - 중앙값
            크기 순서대로 나열했을 때 가운대 위치
        - 최빈값
            가장 나누 나타난 값
            이산형 데이터에 주로 나타남 

        이를 통해 자료의 개형 (왜도 : Skewness 확인 가능)

    2.2 변동성 척도
        - 사분위수 (Quartile)
            데이터 표본을 네 개 동일한 크기로 나눈 값
            Q1, Q2, Q3
        - 사분위간 범위 (IQR)
            Q3에서 Q1 값 제외한 범위 
            데이터의 중앙 50%가 퍼져있는 범위 
            이상치 탐지하는 데에 유용함 
            :: 네모박스에서 중앙 박스에 해당하는 값 ! 
        - 범위 (Range)
            데이터 최대값 최소값 차이 나타냄 
        - 분산 (Vaiance)
            데이터가 평균으로부터 얼마나 멀리 퍼져있는지 
            분산이 높을수록 데이터가 퍼져있음 
        - 표준편차 (Standard Deviation)
            분산의 제곱근 
            분산은 관측값 단위의 제곱인 반해 이를 제곱근 함으로써 더 많이 사용할 수 있음 
        - 변동계수 (Coefficient of variance)
            상대적 변동성을 나타내는 통계적 지표 
            표준편차에는 단위를 사용, 단위가 다른 데이터 비교할 때 용이함

    2.3 연관성 척도
        - 공분산 (covariance)
            두 변수가 함께 어떤 방향으로 변화하는지, 그 크기 
            - Cov (X,Y) > 0이면 X와 Y는 양의 관계, X가 커지면 Y도 커짐
            - Cov (X,Y) < 0이면 X와 Y는 음의 관계, X가 커지면 Y는 작아짐
            - Cov (X,Y) = 0이면 X와 Y는 양도 음도 아닌 관계
                - 선형적인 관계는 없으나, 비선형적인 관계가 존재할 수 있음 (꺾인 직선, 곡선, 파형 등)
        - 상관계수 (correlation)
            두 변수 간 선형적 관계를 나타내는 통계적 지표
            인과관계와 구별됨
            - Corr (X,Y)이 양수면 양의 상관관계 (1이면 완전한 양의 상관관계)
            - Corr (X,Y)이 음수면 음의 상관관계 (-1이면 완전한 음의 상관관계)


## 3.추론 통계 1 (모집단 분포 추정)

    - 모집단
    - 표본집단 (이를 통해 모집단 추정)
        1. 대수의 법칙 
            표본의 크기가 커질수록 통계적 정확도 증가
        2. 중심극한정리
            표본의 크기가 크면 표본 평균의 분포가 모집단의 분포와 상관없이 정규분포에 가까워짐

    3.1 확률분포
        확률변수가 특정 값을 가질 확률 
        확률 변수란 어떤 사건이나 실험의 가능한 결과를 숫자로 대응시키는 함수 
            동전의 앞면 1, 뒷면 0
        - 이산확률분포
        - 연속확률분포
    3.2 확률분포 종류
        1. 정규분포
            종모양
            평균(μ), 표준편차(σ)
        2. T분포
            종모양, 정규분포보다 꼬리가 두꺼움
            자유도(df) : 얼마나 많은 정보가 독립적으로 변할 수 있는지 
                이미 정해진 다른 조건 떄문에 자유롭게 변할 수 없음 
                T분포의 자유도는 포본 크기에 비례 
            표본크기 작을 떄 정규분포보다 더 정확 
        3. 카이제곱분포
            0에서 시작하여 오른쪽으로 긴꼬리를 가진 비대칭 
            자유도가 증가함에 따라 중심이 오른쪽으로 이동, 더 대칭, 더 정규분포에 가까워짐 
            자유도(df)
            독립적인 표준 정규분포 따르는 변수 제곱합 
            주로 독립성 검정, 적합도 검정에 사용됨 


## 4. 추론통계 2 (가설검정)

    4.1 가설검정의 절차

        - 가설설정
            귀무가설 H0, 대립가설 H1설정 
            H0 - 기각하고자 하는 명제 
            H1 - 내가 주장하고자 하는 명제 (위험성 비교적 강함)
        - 1종, 2종 오류
            - 1종 오류(𝛼) : 귀무가설이 실제로 참이었는데 기각할 확률 -> 더 치명적
            - 2종 오류(𝛽) : 귀무가설이 실제로 거짓인데 기각하지 못할 확률
        - 유의 수준 설정
            얼마나 변했을 때 귀무가설을 기각할 수 있을까 
            허용할 수 있는 오류의 최대치 
            유의 수준을 0.05로 설정한다는 것은, "귀무가설이 실제로 참인데도 불구하고, 5%의 확률로 귀무가설을 잘못 기각할 수 있다"는 것
        - 검정통계량 산출
            가설 검정의 결과를 판단하는 데 사용
            데이터 분포, 특징에 따라 T값, F값 등 사용할 통계량 결정 
            (1) Z score
                표준 정규 분포 사용
                특정 데이터가 평균으로부터 얼마나 떨어져 있는지를 표준편차 단위로 표현
                큰 표본크기에 대한 가설 검정에 사용 
                ex. 모집단 평균에 대한 가설 검정할 때 
                절댓값 클수록 평균으로부터 멀리 떨어짐
                임계점 - 양측 검정 시 유의수준 기준
            (2) T score
                T분포 사용 
                작은 표본에서 발생할 수 있는 불확실성을 보다 잘 반영함 
                모집단 평균에 대한 가설 검정, 두 표본 평균 간 차이 검정 
                절댓값 클수록 모집단 평균으로부터 멀리 떨어짐 
                임계점 - 자유도와 유의수준에 따른 t분포표 
            (3) F score
                F분포 사용 
                두 표본 분산 사이 차이 비교, 카이제곱 분포된 변수의 비율로 정의 
                분산분석에서 두 개 이상의 그룹 간 평균의 차이 검정 
                1에 가까울수록 두 표본 분산이 서로 비슷 
        - 가설 기각/ 채택 판단
            (1)  유의확률(p-value)
                검정통계량에서 유도가능 
                p-value가 유의수준보다 작을 시에 귀무가설을 기각
