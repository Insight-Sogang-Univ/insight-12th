# 회귀 심화

- 지도학습 : 정답으로 간주할 데이터 있음
- 비지도학습 : 정답으로 간주할 데이터 없음

설명 변수의 개수만 차이, 기본적 원리는 같음

### 0-3) 최소 제곱법 (최소 자승법)

OLS 라고도 부름

# 선형회귀


### 1-1) 다중선형회귀의 수식적 이해 

- 우리가 관심있는 y에 대해 영향을 줄 것 같은 여러 설명변수를 채택하여 선형적으로 식을 작성
- 상수항(베타)과 오차항(입실론)도 포함된 개념
- 선형회귀식 : 종속변수 y에 대한 설명변수(feature)의 가중평균 : 설명변수에 가중치 곱해 더함으로써 y를 얼마나 잘 설명하는지 나타냄
![](./스크린샷_2024-08-23_오후_2.27.16.jpg)


### 1-2) 다중선형회귀의 기본가정 

- 다음 가정 만족하지 않으면 신뢰성이 떨어질 수밖에 없고 심한 경우 분석이 아예 불가능함

- 가정1) 선형성

    - 종속변수 - 설명변수 간 관계가 선형적이어야 함


- 가정2) 오차항의 평균은 0이다.

    - 오차항(잔차) = 백색잡음


- 가정3) 독립성

    - 각각의 설명변수가 서로 선형 독립적이어야 함 -> 그렇지 않으면 다중공선성

      
- 가정4) 등분산성

    - 오차항의 분산이 일정해야함 


- 가정5) 오차항은 자기상관되어 있지 않다

    - 오차항의 공분산은 항상 0이어야 한다


- 가정6) 정규성

    - 오차항이 정규분포를 따른다
    - 이는 위배되어도 결과에 큰 영향을 주지는 않음



### 1-3) 회귀분석 평가방법 

- 시각화 
        선형관계를 파악 but 어느정도로 더 잘 요약한 회귀선인지 알기 위해선 통계지표 확인 필요


- 통계지표

  (1)결정계수 : 변동(평균과의 차이) 중에서 설명된 비율을 나타냄 \
                    총 편차(SST) = 회귀선 설명 가능 편차(SSR) + 불가능 편차(SSE)\

  
  (2)조정된 결정계수 : R-square 값 높을수록 데이터 잘 설명함 \
            but 설명변수 개수 많을수록 R-square도 높아지기 때문에 개수 증가에 덜 민감하도록 조정\
            Adjusted R^2\

  
  (3) 정보기준 (information criteria) : **AIC, BIC**\
            낮을수록 좋음
        

# 비선형 회귀

- 데이터의 분포가 선형적이지 않을 때

### 2-1) 다항식 회귀모델 

- 선형모델의 feature 다항식으로 만들어 선형회귀 사용
- 기존 변수를 다항식으로 만들고
- scikit-learn 클래스 이용

### 2-2) 지수/로그 회귀모델 

- 특정 증/감소 형태가 지수적으로 or 그 역인 경우가 있음
- 이때의 회귀선은 지수/로그 식으로 하면 됨
- 다항식과 유사하게 기존식에 지수/로그 변형 후 선형회귀 사용
