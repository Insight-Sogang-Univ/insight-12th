통계 마크다운 정리
---

### 1. 통계란?
데이터를 수집, 분석, 해석하여 패턴, 관계, 변동 등을 이해하고 결론을 도출하는 것

#### 통계의 종류
기술 통계학
- 측정, 실험을 통해 수집한 자료를 통해 데이터의 특성을 파악하는 것
- 수치 기술 통계 & 표와 그래프 기술 통계

추리 통계학
- 표본 집단의 데이터를 기반으로 전체 집단의 특성을 추정하는 것
- 통계적 가설 검정

### 2. 기술 통계

#### 중심 위치 척도
평균 (mean)
- 중심 경향성
- 데이터 총합 / 데이터 개수
- 이상치에 민감

중앙값 (median)
- 크기 순으로 나열했을 때 중간에 위치한 값
- 이상치에 덜 민감

최빈값 (mode)
- 가장 자주 나타난 값
- 주로 이산형 데이터에서 사용

** 왜도 (skew)
- 데이터 분포의 비대칭성을 나타내는 통계적 지표
- 최<중<평 (Right-skewed), 최=중=평 (Normal distribution), 평<중<최 (Left-skewed)


#### 변동성 척도
사분위수 (quantile)
- 데이터의 표본을 네 개의 동일한 크기로 나눈 값
- Q1 (하위 25%), Q2 (하위 50%=중앙값), Q3 (하위 75%, 상위 25%)
- np.percentile(arr, 25/50/75)

사분위간 범위 (IQR)
- 이상치 탐지 위해 사용
- 이상치 범위 : Q1-1.5*IQR보다 작고 Q3+1.5*IQR보다 큰 값

범위 (range)
- 데이터 최대값 - 최소값 차이
- np.max(arr) - np.min(arr)

분산 (variance)
- 데이터가 평균으로부터 퍼져 있는 정도를 나타내는 측정 지표
- 높을 수록 많이 퍼져 있음
- 편차의 제곱의 합
- np.var(arr)

표준편차
- 데이터가 평균으로부터 퍼져 있는 정도를 나타내는 통계적 지표
- 높을 수록 많이 퍼져 있음
- 분산의 제곱근
- np.std(arr)
- 기존 데이터와 동일한 단위로 해석 및 비교에 용이

변동계수 (coefficient of variation)
- 데이터의 상대적인 변동성
- (표준편차/평균)*100
- 표준편차의 단위를 없앰으로써 집단의 퍼진 정도를 동일한 수준으로 비교하기 위해 사용
- 높을 수록 변동성이 큼


#### 연관성 척도
공분산 (covariance)
- 두 변수가 변화하는 방향+크기
- Cov(X,Y)>0 : 양의 관계 (선형)
- Cov(X,Y)<0 : 음의 관계 (선형)
- Cov(X,Y)=0 : 양도 음도 아닌 관계
- np.cov(df['col1'], df['col2'])

상관계수 (correlation)
- 두 변수 간 선형적인 관계를 나타내는 통계적 지표
- -1~1 사이의 값
- 공분산/표준편차
- Corr(X,Y)>0 : 양의 상관관계
- Corr(X,Y)<0 : 음의 상관관계
- df[['col1','col2']].corr()

### 3. 추론 통계 1 (모집단 분포 추정)

- 모집단 : 전체 집단
- 표본집단 : 모집단으로부터 추출된 일부
- 표본과 표본통계량을 통해 모집단 추정

** 대수의 법칙 : 표본크기가 커질수록 표본평균이 모평균에 근사
** 중심극한정리 : 표본크기가 크면 표본평균의 분포가 모집단의 분포와 상관없이 정규분포에 근사

#### 확률분포
- 확률변수가 특정한 값을 가질 확률을 나타내는 함수
- 이산확률분포 : 확률 변수가 이산적인 값 가지는 경우 (유한한 값)
    - 베르누이분포, 이항분포, 기하분포, 음이항분포, 초기하분포, 포아송분포
- 연속확률분포 : 확률 변수가 연속적인 값 가지는 경우 (무한한 값)
    - 정규분포, t분포, f분포, 감마분포, 지수분포, 카이제곱분포, 베타분포, 균일분포
 
#### 종류
정규분포
- 평균을 기준으로 대칭인 종 모양
- 매개변수 : 평균(μ), 표준편차(σ)
- 표준편차 클 수록 곡선 평평

t분포
- 정규분포보다 꼬리가 두꺼운 종 모양
- 매개변수 : 자유도(df)
- 자유도 클 수록 표준정규분포 N(0,1)에 수렴
  ** 표준정규분포는 모집단의 평균, 분산을 정확히 알고 있을 때, t분포는 모집단의 분산을 모를 때 사용

카이제곱분포
- 0에서 시작, 오른쪽으로 긴 꼬리를 가진 비대칭
- 매개변수 : 자유도(df)
- 자유도 클 수록 정규분포에 가까워짐 (대칭적으로 변화)
- 독립성 검정, 적합도 검정 등에 사용

### 4. 추론 통계 2 (가설 검정)

#### 가설검정 절차
가설 설정 - 유의수준 설정 - 검정통계량 산출 - 가설 기각/채택 판단

##### 1) 가설 설정
- 귀무가설(H0) : 기각하고자 하는 명제 (ex. ~는 차이가 없다) 
- 대립가설(H1) : 주장하고자 하는 명제 (ex. ~는 차이가 존재한다)
- 1종오류(𝛼) : H0가 참인데 기각할 확률
- 2종오류(𝛽) : H0가 거짓인데 기각하지 못할 확률

##### 2) 유의수준 설정
- 정해진 수치를 얼마나 벗어났을 때 H0을 기각할 것인지 판단하는 기준
- 유의수준 0.05인 경우 ) 실제로 효과가 없더라도 5%의 확률로 효과가 있다고 결론 내릴 가능성 수용하는 것

##### 3) 검정통계량 산출
- 가설검정의 결과를 판단하는 데 사용되는 값
- 데이터의 분포, 특징에 따라 사용되는 통계량 다름

Z값
- (X−μ)/σ
- 표준정규분포를 사용하여 얻은 값
- 큰 표본 크기(n>=30) 에서 주로 사용
- 절댓값 크면 평균으로부터 멀리 떨어져 있음을 의미

T값
- t분포를 사용하여 얻은 값
- 작은 표본 크기에서 주로 사용
- 모집단의 평균에 대한 가설 검정, 두 표본 평균 간의 차이 검정
- 절댓값 크면 표본평균이 모집단평균과 유의미한 차이가 있을 가능성 높음을 의미

F값
- f분포를 사용하여 얻은 값
- 분산분석(ANOVA)에서 두 표본분산 간의 차이 비교하는 데 사용
- 두 개의 독립적인 카이제곱 분포된 변수들의 비율로 정의 (분산1/분산2)
- 1에 가까울 수록 두 표본분산 서로 비슷 (퍼진 정도 비슷)

##### 4) 가설 기각/채택 판단
- 유의확률(p-value) < 유의수준(𝛼) : H0 기각
- 유의확률(p-value) > 유의수준(𝛼) : H0 기각 실패


#### 여러 검정 방법

- 수치형-수치형 : (피어슨) 상관분석
- 범주형-수치형 : ANOVA(F검정), T검정
- 범주형-범주형 : 카이제곱검정
