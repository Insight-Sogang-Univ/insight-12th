[session 2] EDA 사전과제

0. 데이터 분석 프로세스

데이터 수집 - 데이터 탐색 - 데이터 전처리 - 데이터 모델링

- 데이터 탐색 (EDA) : 수집한 데이터의 특성을 여러 가지 방식으로 파악하는 것
- 데이터 전처리 : 데이터 분석을 위해 데이터를 가공하는 것 (결측치, 이상치 처리)
- 데이터 모델링 : 데이터로부터 정보를 추출하기 위해 목적에 맞는 모델을 구축하는 것

-----
1. EDA의 개념

*** 핵심 키워드 : 탐색, 이해

(1) 대상
- 일변량 : 변수 1개
- 다변량 : 변수 여러 개 (변수들 간 관계 파악)
  
(2) 종류
- 시각화 : 차트, 그림 (장점 : 형태 파악 용이)
- 비시각화 : 수치 통계 (장점 : 정확한 값 파악 가능)
  
(3) 유형
- 일변량 비시각화 : 빈도표, 기술 통계량
- 일변량 시각화 : 파이차트, 막대그래프, 히스토그램, 박스플롯
- 다변량 비시각화 : 교차표, 상관계수
- 다변량 시각화 : 모자이크플롯, 박스플롯, 평행좌표, 산점도
  
    - 시각화 - '전체적으로' 살펴보는 것
    - 일변량 - 하나의 데이터 살펴보는 것
    - 다변량 - 둘 이상의 변수 간 관계 살펴보는 것

-----
2. 데이터 확인

- 읽기 : 데이터를 불러오는 것
- 쓰기 : 새로운 데이터를 만드는 것

(1) 절대경로와 상대경로

- 절대경로 : 해당 파일까지의 모든 경로 (처음~목적지)
    - ex) C:/users/juu/desktop/insight-12th
- 상대경로 : 현재 작업 중인 디렉터리 기준으로 상대적으로 표현되는 경로 (상하위 디렉터리)
    - ex) 현재(./) 상위(../) 최상위(/) 하위(이름)

(2) 데이터 입출력
- 파일 형식에 따라 읽고 쓸 때 사용하는 함수 다름
- 읽기는 read_( ), 쓰기는 to_( )
- 파일 형식 : csv, json, html, excel, sql 등

(3) CSV와 EXCEL
- CSV : 데이터를 쉼표(,)로 구분, 크기 작음, 압축 용이 -> 가장 많이 사용
- EXCEL : 데이터프레임 형태 (행/열), 특정 시트 선택하여 불러오기 가능, 여러 시트는 list로 불러오기 가능

  -> 둘 다 dataframe 형식으로 불러오기 가능

3. 전처리

데이터를 손질하는 과정

- 정제 : 결측값 보완, 이상값 제거
- 변환 : 일관성 확보, 중복 최소화
- 필터링 : 오류 삭제 및 보정
- 통합 : 유사한 성질의 데이터 연계

결측값 : 데이터 수집 과정에서 측정되지 않거나 누락된 데이터 (NaN, ?, 0) 
- 처리해야 하는 이유 : 변수 간 관계 왜곡 방지


(1) 결측값 확인

- info() : 결측값 제외한 개수 보여줌
- value_counts(dropna=False) : 결측값 포함한 전체 데이터 보여줌
- isnull() : 누락이면 True, 유효하면 False
- notnull() : 유효하면 True, 누락이면 False
- replace() : '0'이나 '?'을 NaN으로 변환

(2) 결측값 처리

- 삭제
    - dropna() : 결측값 포함 레이블 제거
    - 행 삭제, 열 삭제, 전체 삭제, 부분 삭제
    - 결측값이 무작위로 발생한 경우에 사용
    
- 대체
    - fillna() : 결측값 대체
    - 일괄 대체, 유사 유형 대체
    - 평균값, 최빈값 등을 활용

이상치 : 관측된 데이터의 범위에서 많이 벗어난 값
- 처리해야 하는 이유 : 데이터 분석에 큰 영향 미칠 수 있기 때문

(1) 이상치 확인

- describe() : 통계
- boxplot() : 시각화
- Z-score : 표준편차
- IQR : 사분위 범위

(2) 이상치 제거
- 삭제 (단순 오타, 비현실적인 응답, 처리 과정에서의 오류)
- 대체 (절대적인 숫자 작은 경우)
- 변수화 (자연 발생한 경우 - 삭제 X, 변수화 통한 파악)
- 리샘플링 (자연 발생한 경우 - 분리하여 모델 형성)

피처 엔지니어링 : 피처(변수)들의 형태를 변경하거나 처리하는 것
- 데이터 전처리의 마지막 단계
- 새로운 변수의 추가 없이, 기존 변수를 더 유용하게 만드는 것

(1) 레이블인코딩 / 원핫인코딩
- [텍스트->숫자] 변경 작업
- 레이블인코딩 : 범주형 변수를 0부터 n-1까지의 숫자로 변환
- 원핫인코딩 : 범주형 변수를 이진 벡터로 변환
- pd.get_dummies(데이터프레임, columns=[컬럼 리스트])

(2) 구간화
- 연속 데이터를 일정 구간으로 구분
- pd.cut(데이터프레임['컬럼명'], bins=[나누는 기준 리스트], labels=[지정할 label])

(3) 변환
- 기존의 피처를 다른 피처로 변환하여 변수 추가

(4) 스케일링
- 서로 다른 변수의 값 범위를 일정 수준으로 맞추는 것
- 표준화 : 평균 0, 분산 1 (정규분포)
- MinMaxScaler : 모든 피처가 0~1 사이

-----
4. 시각화

데이터의 패턴, 관계, 추세 등을 쉽게 파악하기 위해 시각적 요소로 변환하는 것

- matplotlib, seaborn 라이브러리 사용
- 파라미터 : 프로그래밍된 함수의 입력값
    - data=data, x='Survived' 에서 data, x
    - hue='Sex' 에서 hue (범주형 변수)

그래프 종류

(1) boxplot
- 상자 수염 그림
- 사분위수, 이상치 확인

(2) countplot
- 범주형 변수의 빈도수 확인

(3) histplot
- 수치형 변수의 구간별 빈도수 확인
- 도수분포표를 그래프로 나타낸 것

(4) distplot, kdeplot
- 커널밀도추정 그래프
- 히스토그램을 연속적인 곡선으로 연결한 것

(5) barplot, pointplot
- 범주형 데이터(x)에 따른 수치형 데이터(y)의 평균 값
- 차이점 : 모양 (막대 & 점)

(6) scatterplot
- 산점도 그래프
- 두 변수 간의 관계 (패턴, 상관관계, 분포) 시각화

(7) regplot
- 회귀선이 추가된 산점도 그래프
- 두 변수 간 산점도 + '회귀선'

(8) catplot
- 수치형 데이터-범주형 데이터 관계

(9) pieplot
- 원형 그래프 (for 비율 확인)
- 모든 데이터를 합쳤을 때 전체를 이루는 경우

(10) heatmap
- 변수 간 상관계수 직관적으로 파악
    - 상관계수 : 두 수치형 변수 간 상관관계의 정도를 나타낸 것 (서로 영향을 주는 정도)
        - -1~1 사이의 값
        - corr()
        - -1, 1에 가까울수록 큰 상관관계 (음/양)

(11) violinplot
- 박스플롯 + 커널밀도추정 함수 그래프

(12) pairplot
- 여러 변수 간 산점도 한번에 파악
