## 회귀기초 마크다운 정리
### 회귀란?
회귀는 데이터에서 패턴을 찾아내어 미래 값을 예측하는 것으로 통계나 머신러닝의 주요 개념 </br>
집값 예측, 주식 가격 예측 등과 같이 연속적인 숫자를 다룸
### 회귀의 종류
선형회귀, 비선형 회귀, 로지스틱회귀, 릿지회귀, 라쏘회귀, 다항회귀 등 다양한 회귀 기법이 존재
### 선형 회귀
선형 회귀는 데이터를 가장 잘 설명하는 회귀선을 찾아 예측하는 과정 </br>
예측하고자 하는 것을 종속 변수, 예측을 위해 사용하는 변수를 독립 변수
* 단순 선형 회귀: 데이터를 직선 형태로 표현하는 경우
* 다중 선형 회귀: 데이터를 두 개 이상의 독립 변수를 사용하여 표현하는 경우

#### 단순선형회귀분석
* 하나의 독립변수만 존재하는 선형 회귀
* 데이터를 가장 잘 설명하는 직선 y=aX+b 을 찾는 과정에 해당
* y=aX+b에서 a는 회귀계수, b는 y절편을 의미
* 적절한 회귀계수와 절편을 찾는 과정
#### 어떻게 a와 b를 잘 찾을 수 있을까?
최소 제곱법, 최소자승법!!
* 잔차의 제곱의 합이 최소가 되는 지점으로 최적의 회귀선을 구하는 방식
* 잔차: 실과 예측값의 차이
* 핵심: 회귀선과 실제 데이터와의 차이를 최소화하기 위한 시도
* 최소 제곱법을 통해 구한 함수가 복잡하여 최소값을 구하기 어려울 때 보통 경사하강법 사용

#### 다중선형회귀분석
* 독립 변수가 2개 이상인 경우의 선형회귀
* 독립 변수 xi가 2개인 경우 3차원 공간에 표현됨. 회귀선은 평면이 된다
* 단순과 마찬가지로 최소제곱법을 사용, 관측치와 평면간의 차이가 잔차

#### 다중공선성
다중공선성은 회귀 분석에서 독립 변수들 간에 상관관계가 큰 경우 발생함

- 다중공선성이 높은 경우 어떤 독립변수가 종속변수에 얼마나 영향을 미치는지를 정확하게 구분하기 어렵기 때문에 회귀모델은 어떤 변수의 영향을 반영해야 할지 불확실해지고, 그 결과 회귀분석의 정확도가 낮아짐

#### 다중공선성 확인 방법
* 상관계수 : 상관계수가 -1, 1에 가까울수록 두 변수의 상관성이 높음
* VIF지수: 분산 팽창 인수로 값이 높으면 다중공선성이 존재한다고 판단
#### 다중공선성 대처 방법
1. 변수 제거: 독립 변수로서 사용할 변수를 선택하는 방법
2. 변수 변환: 변수를 더하거나 빼서 새로운 변수 생성
3. 규제 선형 모델 활용: 릿지, 라쏘, 엘라스틱넷 등의 방법을 통해 모델의 복잡도를 줄이는 방법
4. PCA: 데이터의 차원을 축소하는데 사용되는 통계적 기법
