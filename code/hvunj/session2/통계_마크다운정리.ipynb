## 통계 마크다운 정리
### 1. 통계란?
데이터를 수집, 분석, 해석하는 과정을 통해 패턴, 관계, 변동 등을 이해하고 결론을 도출하는 학문 및 방법론
* 기술 통계학 </br>
: 데이터의 주요 특성을 요약하고 설명하는 방법을 다루는 분야, 평균, 중앙값, 표준편차 등의 개념이 이에 해당 </br>
-> 측정이나 실험을 통해 수집한 자료를 정보로써 전환하기 위하여 요약하고 정리하여 그 특성을 기술하는 절차
  - 수치 기술 통계
  - 표와 그래프 기술 통계
* 추론 통계학 </br>
: 수집한 데이터를 기반으로 미래의, 혹은 조사하지 못한 전 정보를 추론하고 결론을 도출하는 방법 </br>
-> 전체 집단(=모집단)에서 일부를 뽑아 만든 표본 집단을 통해 전체 집단에 대한 정보를 추측하는 과정으로, 통계적 가설검정과 신뢰구간 등 </br>
 

### 2. 기술 통계 (수치 기술 통계)
#### 2.1 중심 위치 척도
* 평균 </br>
데이터 값의 총합을 데이터의 개수로 나눈 값으로 데이터의 중심 경향성을 나타냄
* 중앙값 </br>
데이터 값들을 크기 순서대로 나열했을 때 가운데 위치한 값
* 최빈값 </br>
데이터에서 가장 자주 나타나는 값
##### 평균, 중앙값, 최빈값을 통해 자료의 개형(왜도 등)을 확인할 수 있음
- Positive Skewness </br>
: 오른쪽 꼬리가 왼쪽보다 더 길 때를 의미, 평균과 중앙값이 최빈값보다 큼
- Negative Skewness </br>
: 왼쪽 꼬리가 오른쪽보다 더 길 때를 의미, 평균과 중앙값이 최빈값보다 작음
- Normal distribution </br>
: 정규분포는 왼쪽과 오른쪽이 대칭이며 평균과 중앙값, 최빈값이 같음

#### 2.2 변동성 척도
* 사분위수 </br>
데이터의 표본을 네 개의 동일한 크기로 나눈 값
- Q1 = 하위 25%에 위치한 값
- Q2 = 하위 50%에 위치한 값
- Q3 = 하위 75%, 즉 상위 25%에 위치한 값
* 사분위간 범위 (IQR) </br>
Q3에서 Q1 값을 제외한 범위를 의미, 즉 데이터의 중앙 50%가 퍼져있는 범위 </br>
-> 이상치를 탐지하는 데에 유용
* 범위 </br>
데이터의 최대값과 최소값의 차이를 나타내는 지표로, 데이터의 전체 변동 범위를 알려줌
* 분산 </br>
데이터 값들이 평균으로부터 얼마나 멀리 퍼져있는지를 나타내는 측정 지표
* 표준편차 </br>
데이터 값들이 평균으로부터 얼마나 퍼져 있는지를 나타내는 중요한 통계적 지표
* 변동계수 </br>
데이터의 절대적인 변동성이 아닌 상대적인 변동성을 나타내는 통계적 지표

#### 2.3 연관성 척도
* 공분산 </br>
두 변수가 함께 어떤 방향으로 변화하는지와 그 크기를 표현하는데 사용 </br>
  Cov(X,Y) > 0이면 X와 Y는 양의 관계 </br> Cov(X,Y) < 0이면 X와 Y는 음의 관계 </br> Cov(X,Y) = 0이면 X와 Y는 양도 음도 아닌 관계
* 상관계수 </br>
두 변수 간의 선형적인 관계를 나타내는 통계적 지표 </br>
-> 상관계수는 -1과 1 사이의 값을 가지며, 상관관계는 인과관계가 다름

### 3. 추론 통계 1 (모집단 분포 추정)
* 대수의 법칙 </br>
표본의 크기가 커질수록 표본 평균이 모평균에 근사하는 현상을 말하며, 취합하는 표본의 수가 많을수록 통계적 정확도가 올라감. 따라서 데이터가 많을 수록 좋음
* 중심극한정리 </br>
표본의 크기가 크면 표본 평균의 분포가 모집단의 분포와 상관없이 정규분포에 가까워짐
#### 3.1 확률분포
: 확률변수가 특정한 값을 가질 확률을 나타내는 함수 </br>

확률분포는 다양한 종류가 있으며, 대표적으로는 이산확률분포와 연속확률분포가 있음 
이산확률분포는 확률 변수가 이산적인(discrete) 값을 가지는 경우에 해다. 이산적인 값은 셀 수 있으며 무한한 개수가 아닌, 유한하거나 셀 수 있는 수의 값 가짐 </br>
 연속확률분포는 확률 변수가 연속적인(continuous) 값을 가지는 경우에 해다. 연속적인 값은 무한히 많은 값의 범위를 가지며 실제로는 무한한 값 중 하나를 선택하는 것이 불가.
#### 3.2 확률분포 종류
* 정규분포 </br> 종모양, 평균을 기준으로 대칭, 표준편차가 클수록 곡선이 평평해짐
* T분포 </br> 종모양, 정규분포보다 꼬리가 두꺼움, 평균을 기준으로 대칭, 자유도가 커집에 따라 표준정규분포 N(0,1)에 수렴
* 카이제곱분포 </br> 0에서 시작하여 오른쪽으로 긴 꼬리를 가진 비대칭, 자유도가 증가함에 따라 중심이 오른쪽으로 이동하고 모양이 더 대칭적으로 변화하며, 정규분포에 가까워짐 

### 4. 추론통계2 (가설 검정)
#### 4.1 가설 검정의 절차
#### #1 가설 설정
귀무가설 H0(기각하고자 하는 명제)와 대립가설 H1(내가 주장하고자 하는 명제) 설정
* 1종 오류 : 귀무가설이 실제로 참이었는데 기각할 확률
* 2종 오류 : 귀무가설이 실제로 거짓인데 기각하지 못할 확률 </br>
--> 가설검정은 1종 오류를 더 치명적인 것으로 판단
#### #2 유의 수준 설정
정해진 수치를 얼마나 벗어났을 때 귀무가설을 기각할 것인가를 판단하는 기준을 유의수준이라고 함
#### #3 검정통계량 산출 
검정통계량은 가설 검정의 결과를 판단하는 데 사용하는 값
#### #4 가설 기각/ 채택 판단
유의 확률이 유의 수준보다 작을 시에 귀무가설을 기각
#### 4.2 여러 검정 방법
[수치형 - 수치형] 피어슨 상관분석 </br>
[범주형 - 수치형] F-test(ANOVA), T-test </br>
[범주형 - 범주형] Chi2-test </br>
[수치형 - 범주형] </br>
