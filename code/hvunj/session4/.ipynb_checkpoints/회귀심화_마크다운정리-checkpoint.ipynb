## 회귀심화 마크다운 정리
### 1. 선형회귀
#### 1-1) 다중선형회귀의 수식적 이해
선형회귀식은 종속변수에 대한 설명변수의 가중평균 </br>
설명변수에 가중치를 곱해 더함으로써 y를 얼마나 잘 설명하는지를 나타냄
#### 1-2) 다중선형회귀의 기본가정
* 가정 1: 선형성 </br>
종속변수와 설명변수 간의 관계가 선형적이어야 함
* 가정 2: 오차항의 평규은 0 </br>
오차항은 다른 말로 백색 잡음이라고도 하는데, 백색 잡음의 성질을 만족하기 위해서 필요한 가정
* 가정 3: 독립성 </br>
각각의 설명 변수가 서로 선형독립적이어야 함 </br>
-> 선형독립적이지 않은 경우 다중공선성이 있다고 함
* 가정 4: 등분산성 </br>
오차항의 분산이 일정해야 한다는 가정, 일정하지 않은 경우 이분산성이 있다고 함
* 가정 5: 오차항은 자기상관되어 있지 않다 </br>
오차항의 공분산은 항상 0이어야 한다는 가정, 0이 아닌 경우 자기상관이 있다고 함
* 가정 6: 정규성 </br>
오차항이 정규분포를 따른다는 가정, 위배되어도 다중선형회귀분석의 결과에 큰 영향x
#### 1-3) 회귀분석 평가 방법
회귀 분석의 결과가 얼마나 유효한지 평가하기 위해선, 회귀선을 데이터와 함께 시각화하거나 통계지표를 이용
#### * 시각화
#### * 통계 지표
* 결정 계수 </br>
R^2는 결졍 계수의 약어로, 회귀 분석에서 모델이 설명하는 데이터의 총 변균(평균과의 차이) 중에서 설명된 비율을 나타냄, 0과 1사이의 값으로 나타남
* 조정된 결정 계수 </br>
R^2는 설명변수의 개수가 증가하면 증가할수록 자연스레 증가 -> 데이터와 큰 관련성이 없는 변수를 추가해도 높아짐 </br>
-> 따라서 변수의 개수 증가에 덜 민감하도록 조정된 결정 계수 사용
* ABI, BIC </br>
두 지표 모두 정보기준이라고도 불리며, 값이 낮을수록 좋다고 평가
### 2. 비선형 회귀
#### 2-1) 다항식 회귀모델
다항식을 이용한 회귀 모델은 선형 모델의 feature를 다항식으로 만들어 선형 회귀 사용
* 장점: 선형 모델에 비해 회귀선을 잘 fit하도록 그릴 수 있음
* 단점: 너무 많은 feature를 이용하면 결과가 overfitting될 수 있음 </br>
 -> 이를 방지하기 위해 릿지, 라쏘 규제 이용
#### 2-2) 지수/로그 회귀모델
데이터를 분석하다보면 어떤 칼럼의 증/감소 형태가 지수적으로 혹은 그 역인 경우가 있음. 이때 회귀선을 긋고 싶다면 지수/로그를 사용하여 변형시킨 후, 선형 회귀를 사용
