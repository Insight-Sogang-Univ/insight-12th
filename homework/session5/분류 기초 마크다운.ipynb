## 분류기초

### 분류란?

(1) 머신러닝
- 컴퓨터가 스스로 학습할 수 있도록 도와주는 알고리즘이나 기술을 개발하는 분야
- 지도학습
 - 문제와 답을 모두 알려주고 공부시키는것
- 비지도학습
 - 답을 안가르쳐주고 학습시키는 것
- 강화학습
 - 보상을 통해 상을 최대화/ 벌을 최소화하는 법
(2) 지도학습: 회귀와 분류 
회귀: 주어진 데이터를 기반으로 정답을 잘 맞추는 함수를 찾는것
분류: 기존 데이터가 어떤 레이블에 속하는지 패턴을 알고리즘으로 인지한 뒤 새롭게 관측된 데이터에 대한 레이블 판별하는 것

회귀/분류 차이?
- 회귀: 데이터가 연속형 변수를 예측하기 위해 사용될때
- 분류: 데이터가 범주형 변수를 예측하기 위해 사용될때

분류에 쓰이는 대표적 머신러닝 알고리즘
- 로지스틱 회귀
- 결정트리
- 서포트 백터 머신
- 최소 근접
(3) 이진 분류와 다중 분류
- 이진 분류 : 참/ 거짓
- 다중 분류: 변수가 가질 수 있는 값 3개 이상

### 분류모델
(1) 로지스틱 회귀 
- 이진 분류 문제를 푸는 대표적인 알고리즘으로, 샘플이 특정 클래스에 속할 확률을 추정
(2) 결정 나무
- 조건에 따라 데이터를 분류하며, 최종적으로 데이터가 순수한 label의 집합으로 구성될 때까지 분류를 반복하는 모델 
(3) 서포트 벡터 머신(SVM)
- 클래스를 분류할 수 있는 다양한 경계선 중 최적의 라인을 찾아내는 알고리즘
(4) KNN 
- 데이터로부터 거리가 가까운 k개의 다른 데이터 레이블을 참조하여 분류하는 알고리즘

### 분류평가지표
(1) 혼동 행렬
- 분류 모델의 예측 결과를 정확한 예측과 잘못된 예측으로 구분하여 나타낸 표(행렬)
(2) F1-Score
- 정밀도(precision)와 재현율(recall)의 조화 평균
(3) ROC/ AUC Curve
- 얼마나 분류가 잘 되었는가를 보여주는 그래프
(4) 다중 분류 평가 지표

### 하이퍼 파라미터 최적화
(1) 하이퍼파라미터: 학습 시작 전에 사용자가 직접 설정하는 변수
- 하이퍼파라미터 최적화: 적절한 하이퍼파라미터를 찾아 모델 성능을 향상시키는 것
- Hyperparameter
    모델 학습 과정에 반영되는 값으로, 학습 시작 전에 사용자가 직접 설정하는 변수
- Hyperparameter Optimization
    하이퍼파라미터 최적화란, 튜닝을 거쳐 적절한 하이퍼파라미터를 찾음으로써 모델 성능을 향상시키는 것
(2) 하이퍼파라미터 최적화 방법
- grid search
- random search
- bayesian optimization
